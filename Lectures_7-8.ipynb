# Устойчивость термодинамического равновесия

Пусть наша система находиться в теплоизолированной оболочке ($\delta Q \;=\; 0$), а также в состоянии равновесия с максимальной энтропией $S$. (Любое другое состояние, в которое можно перейти с $\delta Q \;=\; 0$ будет иметь меньшую энтропию). Пусть (допустим) система перешла в состояние с меньшей энтропией $S_2\;\leq \; S_1$. НО! Мы получали 

$S_2\;-\;S_1\; \geq \int\limits_1^2 \dfrac{\delta Q}{T}$,  т.к. $\delta Q \;=\; 0 \Rightarrow \; S_2\; \geq \; S_1$.

Таким образом, находясь только в теплоизолированной оболочке, система будет оставаться в этом состоянии (если энтропия максимальна в этом состоянии). Сформулировали условие термодинамической устойчивости равновесия для случая, когда $\delta Q \;=\; 0$.

Рассмотрим другие случаи:
* Пусть $V\;=\;V_0$, $T\;=\;const$.

т.е. поместили систему в жесткую оболочку и температура окружающей среды - fixed. Следовательно, система не совершает работу.

$$Y\;=\; U\;-\;T_0S,$$
$$Y_1\;-\;Y_2\; \geqslant \; 0.$$

**Nota bene:** Процессы могут идти только с убылью $Y$.

Если же температура среды совпадает с температурой системы, то:

$$Y\;=\;F\;=\;U\;-\;TS.$$

**Nota bene:** 

$$ dU \;=\; T dS \;-\; P dV,$$
$$ dF \;=\; -S dT \;-\; P dV,\; \text{т.е. $dF\;=\;0$}.$$

* Пусть $T_0$, $P_0$ $\;-\;$ фиксированные параметры.

$$Z\;=\;Y\;+\;P_0V.$$
$$A\;<\;Y_1\;-\;Y_2.$$
Работа может быть только $P_0 (V_2\;-\;V_1)$.
$$0\;=\;A_{\text{полезное}}\;=\;A\;-\;P_0 (V_2\;-\;V_1)\; \leqslant \;\underbrace{Y_1\;+\;P_0V_1}_{Z_1}\;-\;(\underbrace{Y_2\;+\;P_0V_2}_{Z_2}),$$
$$\Rightarrow \; Z_1\;-\;Z_2\; \geqslant \; 0.$$
т.е. процесс идет с убылью $Z$.

Если $P_0$, $T_0$ свопадают со значениями $P$, $T$ системы, то $Z\;=\;G\;=F\;+\;PV$, т.е. $G_1\;-\;G_2\; \geqslant \; 0$.

* Помним, что:

$$S_2\;-\;S_1 \; \geqslant \int \dfrac{dU \;+\; \delta A}{T}.$$

Пусть $S$ и $V$ - константы, тогда $\delta A\;=\; 0$.
$$\Rightarrow \;0\; \geqslant \; \int \dfrac{dU}{T} \; \Rightarrow \boxed{\;0\; \geqslant \;\int\limits_1^2 dU}.$$

т.е. если поддерживаем $S$ и $V$ постоянными, то процесс идет с убылью $U$.

т.е. если $S$ и $V$ - константы и $U$ достигла минимума (в некоторовом равновесном состоянии), следовательно, состояне термодинамически устойчиво.

* Рассмотрим следующее:

$$H\;=\;U\;+\;PV,$$
$$ dH\;=\; T dS \;+\; V dP,$$
$$ dU \;=\; dH.$$

Если $S$ и $P$ - константы, то:
$$0\; \geqslant \; \int \dfrac{d U \;+\; P d V}{T}\;=\; \int \dfrac{d H}{T},$$
$$\int d H\; \leqslant \;0.$$

Процессы идут с убылью энтальпии.


## Принцип Ле Шателье — Брауна и устойчивости термодинамического равновесия

Пусть система выведна из положения равновесия. Будем считать, что в системе существует состояние устойчивого равновесия.
Если система находится в устойчивом равновесии то всякий процесс, который вызван внешним воздействием, направлен таким образом, чтобы уничтожить изменения, вызванные внешним взаимодействием. 

**Математически.**

Условие равновесия формируется как минимум некоторой функции состояния.
Параметры системы разделим на те, кторые меняются и нет (свободные).

Если мы имеем дело с изотропной и однородной системой, то свободных параметров будет 2 (независимых).

Итак, рассмотрим $f = f(x,y):$

<br>

\begin{align}
    &d f = \Big(\frac{\partial f}{\partial x}\Big)_yd x + \Big(\frac{\partial f}{\partial y}\Big)_xd y
    \\
    &d f = 0, \text{т.е.} \Big(\frac{\partial f}{\partial x}\Big)_y = \Big(\frac{\partial f}{\partial y}\Big)_x = 0
\end{align}

<br>

Однако, есть ещё и второй дифференциал:

<br>

\begin{align}
    d ^2 f = \dfrac{\partial^2 f}{\partial x^2} d x^2 + 2 \dfrac{\partial^2 f}{\partial x \partial y} d x d y + \dfrac{\partial^2 f}{\partial y^2} d y^2
\end{align}

<br>

(В случае функции одной это просто вторая производная (для min должны иметь $\dfrac{\partial^2 f}{\partial x^2} > 0$)).

<br>

\begin{align}
    d x^2 >0, \; d y^2 >0
\end{align}

<br>

Что можно сказать про $d f$?
Перепишем:

<br>

\begin{align}
    & d ^2 f = \dfrac{\partial^2 f}{\partial x^2} \left( d x + \dfrac{\dfrac{\partial^2 f}{\partial x \partial y} d y}{\dfrac{\partial^2 f}{\partial x^2}}\right)^2 - \dfrac{\left(\dfrac{\partial^2 f}{\partial x \partial y}\right)^2 d y^2}{\dfrac{\partial^2 f}{\partial x^2}} + \dfrac{\partial^2 f}{\partial y^2} d y^2
    \\
    & \Rightarrow \dfrac{\partial^2 f}{\partial x^2} > 0\;  \text{и} \; \bigg( \dfrac{\partial^2 f}{\partial y^2} \dfrac{\partial^2 f}{\partial x^2} - \left(\dfrac{\partial^2 f}{\partial y \partial x}\right)^2 \bigg) >0
    \\
    & \Rightarrow \text{det} \begin{vmatrix}
  \partial^2_y f & \partial^2_{xy} f\\
  \partial^2_{xy} f & \partial^2_{x} f
\end{vmatrix} >0 \; \text{и} \; \partial^2_{x} f >0.
\end{align}

<br>

Пусть:

<br>

\begin{align*}
    &\Big(\frac{\partial f}{\partial x}\Big)_y\;=\;X(x,y),\\
    &\Big(\frac{\partial f}{\partial y}\Big)_x\;=\;Y(x,y).\\
\end{align*}

<br>

Помним, что:

<br>

\begin{align*}
    &\Big(\frac{\partial X}{\partial x}\Big)_y\;>\;0& \Big(\frac{\partial Y}{\partial y}\Big)_x\;>\;0.
\end{align*}

<br>

Получим ещё условия:

<br>

\begin{align*}
    &\Big(\frac{\partial X}{\partial x}\Big)_y\Big(\frac{\partial Y}{\partial y}\Big)_x\;-\;\Big(\frac{\partial Y}{\partial x}\Big)_y\Big(\frac{\partial X}{\partial y}\Big)_x\;>\;0,\\
    &\text{(из определителя матрицы > 0)}\\
    &\Big(\frac{\partial Y}{\partial y}\Big)_x\;-\;\Big(\frac{\partial x}{\partial X}\Big)_y\Big(\frac{\partial Y}{\partial x}\Big)_y\Big(\frac{\partial X}{\partial y}\Big)_x\;>\;0,\\
    &\Big(\frac{\partial x}{\partial X}\Big)_y\Big(\frac{\partial X}{\partial y}\Big)_x\;=\;\dfrac{\partial(x,y)}{\partial(X,y)}\dfrac{\partial(X,x)}{\partial(y,x)}\;=\;-\Big(\frac{\partial X}{\partial y}\Big)_x.
\end{align*}

<br>

Помним, что $X\;=\;0$

<br>

\begin{align*}
    &\Big(\frac{\partial Y}{\partial y}\Big)_x\;+\;\Big(\frac{\partial x}{\partial y}\Big)_{X=0}\Big(\frac{\partial Y}{\partial x}\Big)_y\;>\;0,\\
    &d Y\;=\;\Big(\frac{\partial Y}{\partial x}\Big)_yd x\;+\;\Big(\frac{\partial Y}{\partial y}\Big)_xd y\;=\;\Big(\frac{\partial Y}{\partial x}\Big)_yd X\;+\;\Big(\frac{\partial Y}{\partial y}\Big)_xd y,\\
    &\Big(\frac{\partial Y}{\partial y}\Big)_x\;+\;\Big(\frac{\partial Y}{\partial x}\Big)_y\Big(\frac{\partial X}{\partial y}\Big)_x\;=\;\Big(\frac{\partial Y}{\partial y}\Big)_x,\\
    &\implies \boxed{\Big(\frac{\partial Y}{\partial y}\Big)_x>\;0}.\\
    &\text{Аналогично: }\;\;\boxed{\Big(\frac{\partial x}{\partial X}\Big)_y>\;0}.
\end{align*}

<br>

Ещё соотношения.

Уже было:

<br>

\begin{align}
    &\Big(\frac{\partial Y}{\partial y}\Big)_x\;=\;\Big(\frac{\partial Y}{\partial y}\Big)_x\;+\;\Big(\frac{\partial Y}{\partial x}\Big)_y\Big(\frac{\partial X}{\partial y}\Big)_x =\nonumber\\
    &=\;\Big(\frac{\partial Y}{\partial y}\Big)_x\;-\;\Big(\frac{\partial Y}{\partial x}\Big)_y\Big(\frac{\partial x}{\partial X}\Big)_y\Big(\frac{\partial X}{\partial y}\Big)_x =\nonumber\\
    &=\;\Big(\frac{\partial Y}{\partial y}\Big)_x\;-\;\dfrac{\Big(\frac{\partial X}{\partial y}\Big)_x^2\geqslant 0}{\Big(\frac{\partial x}{\partial X}\Big)_y > 0}\nonumber\\
    &\implies \boxed{\Big(\frac{\partial Y}{\partial y}\Big)_x\geqslant\Big(\frac{\partial Y}{\partial y}\Big)_{X=0} \ },\\
    &\implies \boxed{\Big(\frac{\partial x}{\partial X}\Big)_y\geqslant\Big(\frac{\partial X}{\partial x}\Big)_{Y=0} \ }.
\end{align}

<br>

Ещё соотношения:

<br>

\begin{align}
    \Big(\frac{\partial X}{\partial y}\Big)_x\;=\;\dfrac{\partial(X,x)}{\partial(y,x)}\dfrac{\partial(X,y)}{\partial(X,y)}\;=\;-\Big(\frac{\partial X}{\partial y}\Big)_x \Big(\frac{\partial x}{\partial X}\Big)_y.
\end{align}

<br>

Из $x, y, X, Y$ только 2 независимо могут меняться, если $X$=const $\implies$ то из 3 величин только одна может меняться.

<br>

\begin{align}
    &\Big(\frac{\partial X}{\partial y}\Big)_x\;=\;\Big(\frac{\partial X}{\partial y}\Big)_x\Big(\frac{\partial Y}{\partial y}\Big)_x,\\
    &\implies \Big(\frac{\partial X}{\partial y}\Big)_x\;=\;-\underbrace{\Big(\frac{\partial x}{\partial X}\Big)_y}_{>0}\Big(\frac{\partial X}{\partial y}\Big)_x\underbrace{\Big(\frac{\partial Y}{\partial y}\Big)_x}_{>0}.\\
    &\qquad\qquad\qquad\qquad\huge{||}\nonumber\\
    &\;\;\quad\quad\Big(\frac{\partial Y}{\partial x}\Big)_y\;=\;-\underbrace{\Big(\frac{\partial Y}{\partial y}\Big)_x}_{>0}\Big(\frac{\partial Y}{\partial x}\Big)_y \underbrace{\Big(\frac{\partial x}{\partial X}\Big)_y}_{>0}.
\end{align}

<br>

$\implies$ совпадают знаки у:

<br>

\begin{align}
    \Big(\frac{\partial X}{\partial y}\Big)_x,\Big(\frac{\partial Y}{\partial x}\Big)_y,-\Big(\frac{\partial X}{\partial y}\Big)_x,-\Big(\frac{\partial Y}{\partial x}\Big)_y, -\Big(\frac{\partial x}{\partial y}\Big)_{X=0},-\Big(\frac{\partial y}{\partial x}\Big)_{Y=0}.
\end{align}

<br>

Рассмотрим пример:

<br>

\begin{align}
    &\text{Пусть:}\;x\;\to\;x\;+\;\Delta_1X\nonumber\\
    &y\;\to\;y\nonumber\\
    &\text{Однако:}\;\nonumber\\
    &\Delta Y\;=\;\Big(\frac{\partial Y}{\partial x}\Big)_y\Delta_1x,\\
    &\Delta_2 x\;=\;\Big(\frac{\partial x}{\partial y}\Big)_{X=0}\Delta Y\;=\;\Big(\frac{\partial x}{\partial y}\Big)_{X=0}\Big(\frac{\partial Y}{\partial x}\Big)_y\Delta_1x,\\
    &\implies\boxed{\text{если}\;\Delta_1x>0\implies\Delta_2x<0},\\
    &\text{Аналогично для:}\;\Delta X\;=\;\Big(\frac{\partial x}{\partial X}\Big)_y\Delta_1x.
\end{align}

<br>

Рассмотрим процесс постоянного давления и температуры: $P_0,\;T_0 $ $\;-\;$ const

<br>

\begin{align}
    Z = U - T_0S + P_0 V = Z(\underset{x}{S},\underset{y}{V})
\end{align}

<br>

Функция Z $\;-\;$ функция переменных $S$ и $V$.

<br>

\begin{align}
    &X = \left(\frac{\partial Z}{\partial S}\right)_V\!\!\! = \left(\frac{\partial U}{\partial S}\right)_V\!\!\! - T_0 = T - T_0, \nonumber \\
    &Y = \left(\frac{\partial Z}{\partial V}\right)_S\!\!\! = \left(\frac{\partial U}{\partial V}\right)_S\!\!\! + P_0 = P_0 - P,\nonumber \\
    &d U = T d S - P d V.
\end{align}

<br>

Необходимое условие равновесия:

<br>

\begin{align}
    &X=0, \; \; Y = 0 \nonumber \\
    &\implies T = T_0, \; \; P = P_0, \nonumber \\
    &\left(\frac{\partial X}{\partial x}\right)_y\!\!\! = \left(\frac{\partial T}{\partial S}\right)_V\!\!\! > 0 \implies T \left(\frac{\partial S}{\partial T}\right)_V\!\!\! = C_V > 0,\nonumber
\end{align}

<br>

\begin{align}
    & \left(\frac{\partial Y}{\partial y}\right)_x\!\!\! = - \left(\frac{\partial P}{\partial V}\right)_S\!\!\! > 0, \nonumber \\
    &\Big(\frac{\partial Y}{\partial y}\Big)_{X=0}\!\!\! = -\Big(\frac{\partial P}{\partial V}\Big)_T > 0,\nonumber \\
    &\Big(\frac{\partial X}{\partial x}\Big)_{Y = 0}\!\!\! = \Big(\frac{\partial T}{\partial S}\Big)_P > 0 \implies T \Big(\frac{\partial S}{\partial T}\Big)_P = C_P > 0.
\end{align}

<br>

## Третий закон термодинамики
Энтропия химически однородного тела конечной плотности при стремлении абсолютной температуры к нулю стремится к предельному значению, которое не зависит от давления, плотности или фазы.\\
Пусть при $T=0\;\text{К}\;S_0 = 0$, тогда

<br>

\begin{align}
    \boxed{\lim\limits_{T\rightarrow 0} S  = S_0 = 0}.
\end{align}

<br>

То есть абсолютный ноль нельзя достичь.

При выходе молекулы из области этой сферы возникает пристеночный слой газа, толщина которого равна радиусу сферы молекулярного действия (цифра 2). Молекулы этого слоя подвергаются действию некоторой средней силы $f$:

<br>

\begin{align}
    &P = \dfrac{\nu R T}{V} - P_\text{слоя},\nonumber \\

    &\left(P + \dfrac{a\nu^2}{V^2}\right) = \nu R TV.
\end{align}

Собираем всё вместе:

<br>
\begin{align}
    \boxed{\left(P + \dfrac{a \nu^2}{V^2}\right)(V - \nu b) = \nu R T}.
\end{align}
