# Явления переноса. Средняя длина свободного пробега. Диффузия

## Флуктуация в случае многих переменных

Теперь займемся случаем многих переменных:

Пусть $W(\Delta x_1, ..., \Delta x_{n})\;=\; \tilde{c} \cdot \exp{-\dfrac{1}{2} \sum\limits_{i, j} \alpha_{ij} \Delta x_{i} \Delta x_{j}}$, при этом $\alpha_{ij}\;=\; \alpha_{ji}$.

Нормировка:

$$\int W(\Delta x_1, ..., \Delta x_{n}) d \Delta x_1 ... d \Delta x_{n}\;=\;1,$$
$$\int \tilde{c} \cdot \exp{-\dfrac{1}{2} \sum\limits_{i, j} \alpha_{ij} \Delta x_{i} \Delta x_{j}} d \Delta x_1 ... d \Delta x_{n}\;=\;1.$$

Как его найти?

Необходимо диагонализировать формулу:
$$\left( \hat{\alpha} \vec{\Delta x}, \vec{\Delta x} \right) \;=$$

Преобразование: пусть $\Delta x_{i}\;=\; \sum \limits_{k} \alpha_{k}' \Delta x_{k}'$
\begin{align}
    &\sum \limits_{ij} \alpha_{ij} \sum \limits_{k} \alpha_{ik} \Delta x_{k}' \sum \limits_{l} \alpha_{jl} \Delta x_{l}'\;= \nonumber \\
    &=\; \sum \limits_{kl} \Delta x_{k}' \Delta x_{l}' \cdot \sum \limits_{ij} \alpha_{ij} a_{ik} a_{jl}\;= \nonumber \\
    &=\; \sum \limits_{k} \gamma_{k} (\Delta x_{k}')^2.
\end{align}

\begin{align}
    &\sum \limits_{ij} \alpha_{ij} a_{ik} a_{jl} \;=\; \sum \limits_{i} a_{ik} (\hat{\alpha} \cdot \hat{a})_{il}\;= \nonumber \\
    &=\; \sum \limits_{i} (\hat{a}^{T})_{ki} (\hat{\alpha} \hat{a})_{il} \;=\; (\hat{a}^{T} \hat{\alpha} \hat{a})_{kl}\;=\; \gamma_{k} \delta_{k,l}.
\end{align}

$$d \Delta x_1 ... d \Delta x_{n}\;=\; \left( \frac{\partial (\Delta x_1, ..., \Delta x_{n})}{\partial \Delta x_1', ..., \Delta x_{n}'} \right) d \Delta x_1'...d \Delta x_{n}'.$$

$$J\;=\; \begin{Vmatrix}
\frac{\partial \Delta x_1}{\partial \Delta x_1'} & ... & \frac{\partial \Delta x_1}{\partial \Delta x_{n}'} \\
\vdots & dots & \vdots \\
\frac{\partial \Delta x_{n}}{\partial \Delta x_1'} & ... & \frac{\partial \Delta x_n}{\partial \Delta x_n'} 
\end{Vmatrix} \;=\; |\det \hat{a}|,$$
$$\hat{a}\;=\; \begin{pmatrix}
a_{11} & ... & a_{1n} \\
\vdots & dots & \vdots \\
c_{n1} & ... & c_{nn} 
\end{pmatrix}.$$

\begin{align}
    &\Rightarrow \; 1\;=\; \int \hat{c} |\det a| \exp{-\dfrac{1}{2} (\hat{\text{Г}} \Delta \vec{x}', \Delta \vec{x}')}\;= \nonumber \\
    &=\; \hat{c} |\det a| \bigcap\limits_{i=1}^{n} \left[ \int d \Delta x_{j}' \exp{-\dfrac{\gamma_{i}}{2} (\Delta x_{i}')^2} \right] \;= \nonumber
\end{align}

$$\det \hat{\text{Г}} \;=\; \det \hat{a}^{T} \hat{\alpha} \hat{a}\;=\; (\det \hat{a})^2 (\det \hat{\alpha}) \Rightarrow$$
$$\dfrac{|\det \hat{a}|}{\sqrt{\det \hat{\text{Г}}}}\;=\; \dfrac{1}{\sqrt{\det \hat{\alpha}}}; \; \det \hat{\text{Г}}\;=\; \bigcap\limits_{i=1}^{n} \gamma_{i}.$$

\begin{align}
    &=\; \hat{c} |\det a| \bigcap\limits_{i=1}^{n} \left[ \dfrac{\sqrt{2}}{\sqrt{\gamma_{i}}} \int \limits_{\infty}^{\infty} d u \exp{-u^2} \right]\;= \nonumber \\
    &=\; \hat{c} \dfrac{|\det a|}{\bigcap\limits_{i=1}^{n} \sqrt{\gamma_{i}}} (2 \pi)^{\frac{n}{2}}\;=\; \dfrac{\hat{c}}{\sqrt{\det \hat{\alpha}}}(2 \pi)^{\frac{n}{2}}\;=\;1 \Rightarrow \nonumber \\
    &\dfrac{\sqrt{\det \hat{\alpha}}}{\sqrt{(2 \pi)^{n}}}\;=\; \hat{c}.
\end{align}

$$W(\Delta x_1, ..., \Delta x_{n})\;=\; \;=\; \dfrac{\sqrt{\det \hat{\alpha}}}{(2\pi)^{\frac{n}{2}}} \exp{- \dfrac{1}{2} \sum \limits_{ij} \alpha_{ij} \Delta x_{i} \Delta x_{j}}.$$

$$\left\langle\Delta x_k \Delta x_l \right\rangle = \alpha_{kl}^{-1},$$ где  $\widehat{\alpha}^{-1}$ - обратная матрица к $\widehat{\alpha}$.

Теперь вернемся к физическим параметрам. Мы рассматриваем какую-то часть системы. 

Найдем флуктуации параметров. Мы начинали с разложения энтропии, вернемся к этому.

\begin{align}
    & S(E, V)=S(\overline{E}, \overline{V}) + \frac{1}{2} \left[ \frac{\partial^2 S}{\partial E^2} \left( \Delta E \right)^2 +2\frac{\partial S}{\partial E \partial V}\Delta E \cdot \Delta V + \frac{\partial^2 S}{\partial V^2} \left( \Delta V \right)^2  \right]= \\
    & = S(\overline{E}, \overline{V}) + \frac{1}{2} \left[ ... \right] \overset{*}{=}
\end{align}
\begin{align}
    & \left( \frac{\partial S}{\partial E} \right) = \frac{\partial S}{\partial E} \left( \overline{E}, \overline{V}\right)+\left( \frac{\partial^2 S}{\partial E^2} \right) \cdot \Delta E + \frac{\partial S}{\partial V \partial E} \Delta V, \\
    & \Delta \left( \frac{\partial S}{\partial E} \right) = \left( \frac{\partial S}{\partial E} \right) \Delta E + \left( \frac{\partial S}{\partial V \partial E} \right) \Delta V, \\
    & ( \Delta E) \cdot \left( \Delta \left( \frac{\partial S}{\partial E} \right) \right) = \left( \frac{\partial S}{\partial E} \right) \Delta \exp2 + \left( \frac{\partial S}{\partial V \partial E} \right) \Delta V \Delta E, \\
    & \left( \frac{\partial S}{\partial V} \right) = \left( \frac{\partial S}{\partial V} \right) \cdot (\overline{E}, \overline{V}) + \left( \frac{\partial S}{\partial V \partial E} \right) \Delta V \Delta E + \left( \frac{\partial^2 S}{\partial V^2} \right)\Delta V, \\
    & \Delta \left( \frac{\partial S}{\partial V} \right) = \left( \frac{\partial S}{\partial V \partial E} \right) \Delta E + \left( \frac{\partial^2 S}{\partial V^2} \right) \Delta V, \\
    & \Delta V \Delta \left( \frac{\partial S}{\partial V} \right) = \left( \frac{\partial S}{\partial V \partial E} \right) \Delta E \Delta V + \left( \frac{\partial^2 S}{\partial V^2} \right) \Delta V^2, \\
    & d E = T d S - p d V,\\
    & \frac{\partial S}{\partial V}_{E} = \frac{p}{T}, \frac{\partial S}{\partial E}_{V} = \frac{1}{T},\\
    & \overset{*}{=} \text{const} + \frac{1}{2} \left[ \Delta V \Delta \left( \frac{p}{T} \right) + \Delta E \Delta (\frac{1}{T}) \right] =\\
    & \Delta E = T \Delta S - p \Delta V, \\
    &= \text{const} + \frac{1}{2} \left[ \frac{\Delta V \Delta p}{T} - \frac{\Delta V p \Delta T}{T^2} + \frac{T\Delta S}{T^2}\left( - \Delta T \right) + \frac{\Delta V p \Delta T}{T^2} \right] = \\
    &= \text{const} + \frac{1}{2} \left[ \frac{\Delta V \Delta p}{T} - \frac{\Delta T \Delta S}{T} \right].
\end{align}

С учетом размерностей

$$ \boxed{w = \text{const} \cdot \exp{-\frac{1}{2} \left[ \frac{\Delta S \Delta T}{kT} - \frac{\Delta V \Delta p}{kT} \right]}}.$$

Здесь вcе параетры. Они не являются. как мы помним, все разом независимыми. Расмотрим пару $(T, V)$:

\begin{align}
    & \Delta S = \frac{\partial S}{\partial T}_{V} \Delta T + \frac{\partial S}{\partial V}_{T} \Delta V, \\
    & \frac{\partial S}{\partial T}_{V} = \frac{C_v}{T};  \frac{\partial S}{\partial V}_{T} = \frac{\partial p}{\partial T}_{V}.\\
    & \Delta S = \frac{C_v}{T} \Delta T + \frac{\partial p}{\partial T}_{V} \Delta V, \\
    & \Delta p = \frac{\partial p}{\partial T}_{V} \Delta T + \frac{\partial p}{\partial V}_{T} \Delta V \Rightarrow\\
    & \Rightarrow \frac{\Delta S \Delta T}{kT} - \frac{\Delta V \Delta p}{kT} = \\
    & \frac{C_v \Delta T^2}{kT^2} + \frac{\Delta T \Delta V}{kT} \frac{\partial p}{\partial T}_{V}-\frac{\Delta V \Delta T}{kT} \frac{\partial p}{\partial T}_{V} - \frac{\Delta V}{kT}\Delta V \frac{\partial p}{\partial V}_{T}=
\end{align}
\begin{align}
    & = \frac{C_v}{kT^2} \left( \Delta T \right)^2 - \frac{\partial p}{\partial V}_{T} \frac{1}{kT} \left( \Delta V \right)^2 \Rightarrow \\
    & \Rightarrow w(\Delta T, \Delta V) = const\cdot \exp{-\frac{1}{2} \left[ \frac{C_v}{kT^2} \Delta T^2-\frac{\partial p}{\partial V}_{T} \frac{\Delta V^2}{kT} \right]}.
\end{align}

Получается, что $\left\langle\Delta T \Delta V \right\rangle=0$,

\begin{align}
    & \Rightarrow \left\langle\left( \Delta T\right)^2 \right\rangle = \frac{kT^2}{C_v}, \\
    & \left\langle\left( \Delta V \right)^2 \right\rangle = - kT \cdot \frac{\partial V}{\partial p}_{T}.
\end{align}

Теперь через $p, S$:

\begin{align}
\end{align}

Вместе:

\begin{align}
    &\left\langle \left(\Delta T\right)^2\right\rangle\;=\;\dfrac{kT^2}{C_v}\;\sim\;\dfrac{1}{V},\\
    &\left\langle \left(\Delta V\right)^2\right\rangle\;=\;-kT\Big( \frac{V}{p} \Big)_T\;\sim\;V.
\end{align}

Это мы смотрели при фиксированном числе частиц. Пусть теперь объём фиксированный. Тогда:
\begin{align}
    d E\;=\;Td S\;-\;pd V\;+\;\mu d N.\nonumber
\end{align}
То есть $p\to -\mu$, $V\to N$.
\begin{align}
    \implies\;W\;=\;const\cdot\exp{-\dfrac{1}{2}\left[\dfrac{\Delta S\Delta T\;+\;\Delta N\Delta\mu}{kT}\right]}.
\end{align}
Рассмотрим $N, T$.
\begin{align}
    &\Delta\mu\;=\;\Big( \frac{\partial \mu}{T} \Big)_N\Delta T\;+\;\Big( \frac{\mu}{N}\Big)_T\Delta N,\\
    &\Delta S\;=\;\Big( \frac{\partial S}{\partial T} \Big)_N\Delta T\;+\;\Big( \frac{\partial S}{\partial N} \Big)_T\Delta N\;=\\
    &=\;\dfrac{C_{V, N}}{T}\Delta T\;+\;\Big( \frac{\partial S}{\partial N} \Big)_T\Delta N\;=\;\dfrac{C_{V, N}}{T}\Delta T\;-\;\Big( \frac{\partial \mu}{\partial T} \Big)_N\Delta N.\\
    &d F\;=\;-Sd T\;+\;\mu d N\;\implies\;\Big( \frac{\partial S}{\partial N} \Big)_T\;=\;-\Big( \frac{\partial \mu}{\partial T} \Big)_N,\\
    &\Delta T\left(\dfrac{C_{V, N}}{T}\Delta T\;-\;\Big( \frac{\partial \mu}{\partial T} \Big)_N\Delta N\right)\;+\;\Delta N\Big( \frac{\partial \mu}{\partial T} \Big)_N\Delta T\;+\;\Delta N^2\Big( \frac{\partial \mu}{\partial N} \Big)_T\;=\\
    &=\;\dfrac{C_{V, N}}{T}\Delta T^2\;+\;\Delta N^2\Big( \frac{\partial \mu}{\partial N} \Big)_T,\\
    &\implies W\left(\Delta N, \Delta T\right)\;=\;const\cdot\exp{-\dfrac{1}{2}\left[\dfrac{C_{V, N}\Delta T^2}{kT}\;+\;\Big( \frac{\partial\mu}{\partial N} \Big)_{T}\dfrac{\Delta N^2}{kT}\right]},\\
    &\implies \left\langle \left(\Delta N\right)^2\right\rangle\;=\;kT\Big( \frac{\partial N}{\partial \mu} \Big)_{T,V},\\
    &d\mu\;=\;\dfrac{d G}{N}\;=\;\dfrac{Vd P\;-\;Sd T}{N}.
\end{align}
Для идеального газа:
\begin{align}
    &\mu\;=\;\mu\left(T\right)\;+\;\int\underset{T=const}{\dfrac{kTd P}{P}}\;=\;\mu^0\left(T\right)\;+\;kT\ln{P},\\
    &PV\;=\;NkT.\\
    &\mu\;=\;\mu^0\left(T\right)t\;+\;kT\ln{\dfrac{NkT}{V}}\;=\;kT\ln{N}\;+\;\ldots\\
    &\Big( \frac{\partial N}{\partial \mu} \Big)_{T,V}\;=\;\dfrac{1}{\Big( \frac{\partial \mu}{\partial N} \Big)_{T,V}}\;=\;\dfrac{N}{kT},\\
    &\left\langle \left(\Delta N\right)^2\right\rangle\;=\;kT\dfrac{N}{kT}\;=\;N,\\
    &\dfrac{\sqrt{\left\langle \left(\Delta N\right)^2\right\rangle}}{N}\;=\;\dfrac{1}{\sqrt{N}}.
\end{align}

## Длина свободного пробега
Мы уже знаем среднее значение скорости движения молекул и знаем, что это значение довольно высокое. Следовательно, когда мы убираем крышку с чашки кофе, запах по комнате должен распространяться моментально, однако этого не происходит. Всё дело в столкновениях частиц. Молекуллы газа не всё время движутся свободно, траектория движения молекул часто изменяется.

Как же охарактеризовать частоту столкновений? Введём длину свободного пробега. Для оценки воспользуемся моделью твёрдых шаров.

Для простоты рассмотрим путь одной частицы, заморозив все остальные. Между двумя столкновениями имеем цилиндр. Тогда число столкновений можно записать как:
\begin{align}
    &z=n\cdot V, \qquad V = \sigma \cdot v \cdot t,\nonumber \\
    \implies&z = n\cdot \sigma \cdot v \cdot t.
\end{align}
Пройденный молекулой путь составляет $vt$, и тогда длина свободного пробега:
\begin{align}
    \lambda = \dfrac{vt}{z} = \dfrac{1}{n\sigma}.
\end{align}
Также есть более точное выражение Максвелла, где в знаменателе ещё стоит $\sqrt{2}$, но это не принципиально.

Что будет с длиной пробега, если имеем дело со смесью разных газов? Тогда зоны ограждения будут иметь другой вид, и надо переписать сечения:
\begin{align}
    &\sigma_{11}=\pi (2r_1)^2, \nonumber \\
    &\sigma_{12}=\sigma_{21}=\pi(r_1+r_2)^2,\nonumber \\
    &\sigma_{22}=\pi(2r_2)^2.
\end{align}
Тогда число столкновений запишутся в виде:
\begin{align}
    z_1=(n_1\sigma_{11}+n_2\sigma_{12})< v >_1,\nonumber \\
    z_2=(n_1\sigma_{21} + n_2\sigma_{22})< v >_2.
\end{align}
И, длины свободного пробега:
\begin{align}
    \lambda_1 = \dfrac{1}{n_1\sigma_{11}+n_2\sigma_{12}}, \quad \lambda_2 = \dfrac{1}{n_1\sigma_{21}+n_2\sigma_{22}}.
\end{align}

## Эффективное сечение
Площадь сечения сферы ограждения молекулы по большому кругу называется эффективным сечением молекулы, или же газокинетическим эффективным сечением. Однако эффективное сечение не всегда стоит связывать с настоящим диаметром молекулы. Сейчас определим его.

В процессе столкновений могут происходить разные процессы: рассеяние (как в нашем случае), поглощение, диссоциация, ионизация. До каждого из "успехов" необходимо какое-то число столкновений $z_i$, $i$  $\;-\;$ тип процесса. $z_i=n\sigma_i v$, откуда 
\begin{align}
    \sigma_i \overset{def}{=} \dfrac{z_i}{nv}.
\end{align}
Размерность площади так же остаётся.

Рассмотрим такую задачу: параллельный пучок молекул летит через газ. Пусть $I_0$  $\;-\;$ интенсивность пучка в начальный момент $x=0$. Какова будет интенсивность $I$ в точке с координатой $x$?

Рассмотрим слой $d x$: число молекул-мишеней в нём составляет $nsd x$, где $n$  $\;-\;$ концентрация в дифференциальном слое.

Для справки: $I=n\cdot s\cdot v\cdot t$, или через единичную площадь за единицу времени: $I = nv$.

<figure>
    <center>
    <img src="images/2lecture23-78.png" width="400" align="center"/>
        <figcaption>Рис. 23.3: Поток налетающих молекул.</figcaption>
    </center>
</figure>

\begin{align}
    & d p \text{ ― вероятность налетающей частицы столкнуться в промежутке } d x. \\
    & d p = \dfrac{\sigma n S d x}{S} = \sigma n d x. \\
    & d I = - \underbrace{n_{\text{налетающих}} v}_I \sigma n d x, \\
    & \dfrac{d I}{I} = - \sigma n d x = - \dfrac{d x}{\lambda}, \\
    & I = I_0 \exp \left( - \dfrac{x}{\lambda} \right). \\
\end{align}
Получаем выражение для числа частиц, дошедших до координаты без столкновений:
\begin{align}
    & N = N_0 \exp \left( - \dfrac{x}{\lambda} \right). \\
    & |d N| = \dfrac{1}{\lambda} N_0 \exp \left( - \dfrac{x}{\lambda} \right) d x. \\
    & \text{Вероятность рассеяния } \dfrac{|d N|}{N_0} = \dfrac{1}{\lambda} \exp \left( - \dfrac{x}{\lambda} \right) d x.
\end{align}
Как и следовало ожидать, средний путь, пройденный частицей без столкновения
\begin{align}
    & \langle x \rangle = \int \limits^{x = + \infty}_{x = 0} x \dfrac{|d N|}{N_0} = \dfrac{1}{\lambda} \int \limits^{+ \infty}_0 x \exp \left( - \dfrac{x}{\lambda} \right) d x = \lambda.
\end{align}

Можем записать, сколько долетает до каждого из атомов решётки:
\begin{align}
    & N_1 = N_0 \exp \left( - \dfrac{x_1}{\lambda} \right), \\
    & N_2 = N_0 \exp \left( - \dfrac{x_2}{\lambda} \right), \\
    & ... \\
    & \dfrac{N_1}{N_2} = \exp \left( \dfrac{x_2 - x_1}{\lambda} \right), \\
    & \ln \dfrac{N_1}{N_2} = \dfrac{x_2 - x_1}{\lambda}. \\
    & \lambda = \dfrac{x_2 - x_1}{\ln \dfrac{N_1}{N_2}} = \dfrac{x_2 - x_3}{\ln \dfrac{N_2}{N_3}}.
\end{align}
Стоит заметить, что здесь $\lambda$ это характеристика рассеяния атомов серебра на атомах воздуха.

## Внутреннее трение и теплопроводность.

В вязкой среде одна пластинка при движении увлекает другую за собой ― это показывает наличие внутреннего трения между слоями жидкости или газа.

Выдвинем ряд предположений о природе внутреннего трения:
* Скорости всех молекул в ходе теплового движения равны.
* Движение молекул можно разбить на шесть равнозначных потоков.
* В обмене импульса между слоями участвует по потоку каждого из двух участвующих слоёв.
* Скорость слоя мало меняется на длине свободного пробега.
* Можно считать, что относительные скорости слоёв много меньше скорости теплового движения молекул.
