   * с одно стороны, это то, что долетало хотя бы до $x_1$ (если бы летело $N_0$ от $x\;=\;0$) и т.д.; 
   * с другой стороны, это то, что сформировалось в пучок в промежутке $(\infty ;\; x)$, который долетит до $x\;=\;0$ без столкновений!
   
<br>
   

$$dN = \dfrac{n v}{6\lambda } \exp{-\dfrac{x}{\lambda}}dx,$$

где $v\;=\; \dfrac{N_0}{\lambda}$.

<br>

Молекулы в каждом слое получают боковой импульс, который перенесут вниз: $p(x)$

$$P_{+}\;=\;\int \limits_{0}^{\infty} p(x) dN(x),$$

(весь через единичную площадку в единицу времени).

<br>

Сделаем разложение:

(скорости макроскопические мало меняются на длине свободного пробега).


<br>

\begin{align}
    &p(x)\;=\;p(0)\;+\; \dfrac{d p}{d x} \bigg|_{x\;=\;0}x 
\end{align}
\begin{align}
    &\Rightarrow \; P_{+}\;=\;\int \limits_{0}^{\infty} p(0) d N(x)\;+\; \dfrac{n v}{\lambda 6 } \left( \dfrac{d p}{d x} \right) \cdot \int \limits_{0}^{\infty} x \exp{-\dfrac{x}{\lambda} d x}\;= \nonumber \\
    &=\; P_0 \;+\; \dfrac{n v}{\lambda 6 } \left( \dfrac{d p}{d x} \right)_{x\;=\;0} \nonumber \\
    &\approx \; \dfrac{1}{6} n v p(\lambda).
\end{align}

<br>

$$P_{-}\;=\; P_0 \;-\; \dfrac{n v}{\lambda 6 } \left( \dfrac{d p}{d x} \right)_{x\;=\;0}.$$

Тогда суммарный переносимый импульс снизу вверх:
\begin{align}
    &P\;=\;P_{-}\;-\;P_{+}\;=\; - \dfrac{1}{3} n v \lambda \left( \dfrac{d p}{d x} \right)_{x\;=\;0}.
\end{align}


<br>

$$\boxed{P\;=\; - \dfrac{1}{3} n v \lambda \left( \dfrac{d p}{d x} \right)_{x\;=\;0}}.$$

Это все было в единицу времени и единицу площади, т.е. по размерности это напряжение (давление).

$$\dfrac{P}{\Delta t \Delta S}\;=\; \dfrac{F}{\Delta S}\;=\; \tau \;=\; \eta \dfrac{d U}{d x},$$

эта величина называется касательное напряжение, где $\eta$ \$-\$ коэффициент внутреннего трения. Закон Ньютона.

## Теплопроводность

<br>

Теперь передается энергия. 

Энергия частицы: $\epsilon \;=\; m_0 c_{v} T$, где $c_{v}$ \$-\$ удельная теплоемкость при постоянном $v$.

Пусть у нас есть две пластинки:

<br>


Пусть $T_2 \;>\; T_1 \; \Rightarrow \; \dfrac{d T}{d x}\;>\; 0 \; \Rightarrow \; q(x) \;<\; 0.$

$$c_{v} \cdot m \;=\; c_{\text{част.}},$$
$$\xi \;=\; \dfrac{1}{3} n v c_{\text{част.}} \lambda .$$

\begin{align}
    &q(x)\;=\; - \dfrac{1}{6} n v m c_{v} \left( T_0 \;+\; \lambda \dfrac{d T}{d x}\;-\; \left( T_0\;-\; \lambda \dfrac{d T}{d x} \right) \right) \;= \nonumber \\
    &=\; -\dfrac{1}{3} n v m c_{v} \lambda \dfrac{d T}{d x}.
\end{align}

$$\boxed{q(x)\;=\; -\dfrac{1}{3} n v m c_{v} \lambda \dfrac{d T}{d x}}.$$


$$\boxed{\vec{q}\;=\; -\dfrac{1}{3} n v m c_{v} \lambda \vec{\nabla} T}.$$

$$\Rightarrow \boxed{\vec{q}=- \xi \nabla T} - \text{закон Фурье для стационарной теплопроводности}.$$

## Самодиффузия в газах

<br>


Представим сосуд с перегородкой. По каждую сторону от перегородки находится газ. У них одинаковое $T$ и $P$. Если газы разные, то это взаимная диффузия, если газы одинаковые, то это самодиффузия. 


<br>

Представим мы промаркировали молекулы по каждую сторону от перегородки. Убираем перегородку и наблюдаем, что происходит. 

<br>


Пусть $n$ - общая концентрация.

<br>


\begin{align*}
    n = n_1+n_2.  
\end{align*}

Потоки будут относительной концетраций: $\frac{n_1}{n}$ и $\frac{n_2}{n}$.

\begin{align}
    & \Rightarrow \Gamma_1 = - \frac{1}{3} n v \lambda \frac{d (\frac{n_1}{n})}{d x}=-\frac{v \lambda}{3} \frac{d n_1}{d x}, \\
    & \Gamma_2 = - \frac{1}{3} n v \lambda \frac{d (\frac{n_2}{n})}{d x}=-\frac{v \lambda}{3} \frac{d n_2}{d x}, \\
    & \Gamma_1 = -D\frac{d n_1}{d x};  \Gamma_2=-D\frac{d n_2}{d x}.
\end{align}

Наблюдение: $\Gamma_1=-\Gamma_2$.

\begin{align}
    \vec{\Gamma}=-F\vec{\nabla}n,
\end{align}
закон Фика. 

Диффузионный поток $\sim \nabla n$.

<br>

D - коэффициент диффузии. 

<br>

В случае взаимной диффузии одна концентрация должна быть мала по сравнению с другой. 

## Подвижность частицы

<br>

Рассмотрим молекулы/частицы, скажем, движущиеся в жидкости.

Пусть на них действует постоянная сила $\vec{F}$ (например электрическое поле и ионы). 

При движении частицы испытывают соударение с молекулами жидкости и другими ионами.

<br>

Из-за $\vec{F}$ появляктся преимущественное направдение скорости иона. появляется регулярное движение (частиц/ионов/молекул) со скоростю $\vec{u}$.

<br>

Дадим определение подвижности частицы:
\begin{align}
    &\vec{U}\;=\;B\vec{F},\\
    &\nonumber(m\ddot{x}\;=\;F-\alpha v,\;\;\ddot{x}\;=\;0:\;F\;=\;\alpha v).
\end{align}
где $B$ \$-\$ подвижность частицы.

<br>

Рассматриваем систему сильно разреженного газа. Пусть $x$ \$-\$ ось направления силы $F$, тогда из-за действия силы:
\begin{align*}
    \varepsilon_P\;=\;-xF.
\end{align*}

<br>

Следовательно частицы должны быть распределены по Больцману:
\begin{align*}
    n\;=\;n_0\cdot\exp{-\dfrac{\varepsilon_P}{kT}}\;=\;n_0\cdot\exp{\dfrac{Fx}{kT}}.
\end{align*}

<br>

Но такое распределение должно быть согласовано с диффузией и подвижностью. Поток из-за диффузии:
\begin{align*}
    -D\cdot\dfrac{d n}{d x}.
\end{align*}

<br>

С другой стороны поток из-за силы:
\begin{align*}
    B\cdot F\cdot n.
\end{align*}

<br>

В сумме они должны быть равны нулю. Тогда можно перейти к равенству:
\begin{align}
    &-D\cdot n_0\cdot\dfrac{F}{kT}\;+\;B\cdot F\cdot n_0\;=\;0,\\
    &\boxed{D\;=\;kTB}.
\end{align}

<br>

<br>

<br>

Значит, если $m_2>m_1\;\implies\;m_2$ идёт к $T_1$.
Если же $d_2>d_1\;\implies\;d_2$ идёт к $T_1$.
Это **термическая диффузия**.

<br>

Направление термодиффузии зависит от характера взаимодействия:
\begin{align*}
    F\sim\dfrac{1}{r^\nu}.
\end{align*}
При $\nu\;=\;s$ будет граничный случай.


Происходит частичное разделение газовой смеси, следовательно возникают градиенты концентраций. Если при этом разность температур поддерживается постоянной, то устанавливаются градиенты $T$ и $n$:
\begin{align}
    \Gamma_1\;=\;-nD_{12}\dfrac{d c_1}{d x}\;+\;n\cdot\dfrac{D_T}{T}\dfrac{d T}{d x}.
\end{align}
В стационарном случае $\Gamma_1\;=\;\Gamma_2\;=\;0$.

<br>

Пусть:

<br>

\begin{align*}
    K_\tau\;=\;\dfrac{D_\tau}{D_{12}},
\end{align*}
тогда:
\begin{align}
    &\dfrac{d c_1}{d x}\;=\;\dfrac{D_\tau}{D_{12}\cdot T}\dfrac{d T}{d x},\\
    &d c_1\;=\;K_\tau\cdot\dfrac{d T}{T},\\
    &c_1(T)\;-\;c_1(T_0)\;=\;K_\tau\ln{\dfrac{T}{T_0}}.
\end{align}

<br>
У более мелких частиц больше шансов долететь в какую-то область. Если мы сравним, как большая и маленькая молекулы одинаковой массы попадают в область холодной стенки, то поймём, что крупная сталкивается чаще: ей тяжело вылететь; частота столковений в горячей области у большой выше, чем у маленькой.

## Явления в разреженных газах
### Вакуум

<br>

**Вакуумом** называется такая область пространства, когда длина свободного пробега сопоставима с размером сосуда ($\lambda\sim L$). Вакуум бывает **низкий**, **средний** и **высокий**.

<br>

### Эффузия разреженного газа

<br>

<figure>
    <center>
    <img src="images/lecture_25-26_88.png" width="400" align="center"/>
        <figcaption>Эффузия разреженного газа.</figcaption>
    </center>
</figure>

<br>

Рассмотрим $s$:
\begin{align}
    &N\;=\;\dfrac{1}{4}nvs\;=\;ns\sqrt{\dfrac{8kT}{\pi m16}}\;=\;ns\sqrt{\dfrac{kT}{nm2}},\\
    &N\;=\;C\dfrac{Ds}{\sqrt{mT}},\\
    &C\;=\;\dfrac{1}{\sqrt{2k\pi}}.
\end{align}

<br>

Проделаем отверстие в перегородке. Чему будет равно число молекул, вылетающих из A в B? 

Всё зависит от длины свободного пробега, толщины перегородки и отверстия? Давление обычное, $\implies \lambda \ll D$, следовательно вблизи отверстия возникает упорядоченное движение.

Если же сделать очень малое отверстие, то это мало скажется на распределении концентрации, и применима обычная формула.

<br>

Поток молекул газа называется **эффузионным потоком**, если размеры отверстия и толщина перегородки малы по сравнению с длиной свободного пробега.

Рассмотрим различные варианты постановки.

<br>

Одинаковые газы, но разные $P$ и $T$. Тогда возникают два эффузионных потока из 1 в 2 и из 2 в 1.
\begin{align}
    N = \dfrac{cS}{\sqrt{m}}\left(\dfrac{P_A}{\sqrt{T_A}} - \dfrac{P_B}{\sqrt{T_B}}\right).
\end{align}
Чем характеризуется равновесие? Равенством встречных потоков.
\begin{align}
    N=0 \implies \dfrac{P_A}{\sqrt{T_A}} = \dfrac{P_B}{\sqrt{T_B}}.
\end{align}

<br>



Пусть теперь имеются разные газы с разными $P$ и $T$. Тогда есть $m_1$, $m_2$, $P_{1,A}$, $P_{1,B}$, $T_A$, $T_B$, $P_{2,A}$, $P_{2,B}$.

<br>

\begin{align}
    &N_{A\rightarrow B} = \dfrac{cS}{\sqrt{m_1}}\left(\dfrac{P_{1,A}}{\sqrt{T_A}} - \dfrac{P_{1,B}}{\sqrt{T_B}}\right), \nonumber \\
    &N_{B\rightarrow A} = \dfrac{cS}{\sqrt{m_2}}\left(\dfrac{P_{2,B}}{\sqrt{T_B}} - \dfrac{P_{2,A}}{\sqrt{T_A}}\right).
\end{align}

<br>

В начальный момент времени

<br>
\begin{align}
    \dfrac{N_{A\rightarrow B}}{N_{B\rightarrow A}} = \sqrt{\dfrac{m_2}{m_1}}\left(\dfrac{P_{1,A}}{P_{2,B}} \sqrt{\dfrac{T_B}{T_A}}\right).
\end{align}

<br>

Если равны давления и температуры, то
\begin{align}
    \dfrac{N_1}{N_2} = \sqrt{\dfrac{m_2}{m_1}}.
\end{align}

В случае крупной трубки тепловое равновесие двух газов нами хорошо изучено: а именно, у газов в обоих сосудах попарно $P_1 = P_2$ и $T_1 = T_2$.

<br>

Если же трубка тонкая, то равновесие может достигаться при разнице давлений и температуры! А именно, фиксируя постоянное количество частиц в каждом из сосудов, получаем

<br>

\begin{align}
    & \dfrac{P_1}{\sqrt{T_1}} = \dfrac{P_2}{\sqrt{T_2}}.
\end{align}

## Общая задача теплопроводности

<br>

Рассмотрим передачу тепла за счёт теплообмена, пренебрегая при этом другими формами теплопередачи (излучением и конвекцией), каким-либо перемещением вещества. Кроме того, будем рассматривать одномерный случай.

<br>


Введём величину плотности потока тепла $j$ ― количество тепла через единичную площадь за единицу времени.

<br>

<figure>
    <center>
    <img src="images/2lecture26-92.png" width="400" align="center"/>
        <figcaption>Тепловой поток в кусочек тела и из него.</figcaption>
    </center>
</figure>

<br>



В каждый конкретный момент времени энергия должна сохраняться ― то, что приходит в кусочек тела в виде потока, должно или уноситься потоком дальше, или нагревать этот кусочек тела.

<br>

\begin{align}
    & j(x, t) S d t - j(x + dx, t) S d t - c_V d M (d T)_x = 0. \\
    & c_V d M (d T)_x = - (d j)_t S d t. \\
    & d M = \rho(x) S d x, \\
    & c_V \rho(x) S d x (d T)_x = - (d j)_t S d t, \\
    & \boxed{c_V \rho(x) (\dfrac{\partial {T}}{\partial t})_{x} = - (\dfrac{\partial {j}}{\partial x})_{t} \ \ } \label{eq:thermal_conductivity}.
\end{align}

<br>

<figure>
    <center>
    <img src="images/2lecture26-93-1.png" width="400" align="center"/>
        <figcaption>Теплопроводность однородного кубика толщины $l$.</figcaption>
    </center>
</figure>

<br>

Можно заметить, что этого уравнения недостаточно, чтобы получить зависимость $j(T)$. В качестве первого приближения можно взять эмпирическую зависимость

<br>

\begin{align}
    & j = ae \; \dfrac{T_1 - T_2}{l} = - ae \; \dfrac{\Delta T}{l}. \\
    & l \to 0, \quad j = - ae (\dfrac{\partial T}{\partial x})_{t}.
\end{align}

<br>

<figure>
    <center>
    <img src="images/2lecture26-93-2.png" width="400" align="center"/>
        <figcaption>Предельный переход от кубика к дифференциально тонкой пластине.</figcaption>
    </center>
</figure>

<br>

Подставляя это в уравнение, получаем

<br>

\begin{align}
    \boxed{\rho(x) c_V(x) (\dfrac{\partial T}{\partial t})_{x} \ = \dfrac{\partial}{\partial x} \left( ae(x) \dfrac{\partial T}{\partial x} \right)_t}.\label{eq:heat}
\end{align}

<br>

Таким образом, уравнение является уравнением теплопроводности.

<br>


В частности, когда $\rho$, $c_V$, $ae$ постоянны,
\begin{align}
    & \chi = \dfrac{ae}{\rho c_V}, \\
    & (\dfrac{\partial T}{\partial t})_{x} = \chi \left(\dfrac{\partial^2 T}{\partial x^2}\right)_t.
\end{align}
