# Статистическая физика. Заключение

В случае идеальной жидкости мы получали:
$$\dfrac{v^2}{2}\;+\; gh\;+\;\dfrac{P}{\rho}\;=\; const.$$

Константа будет одной и той же вдоль одной линии тока. Например, это касается течения жидкости в трубе, т.к. вы не можете выйти жидкостью за пределы трубы.

Рассмотрим частный случай: пусть высота не меняется. Жидкость течет по трубе переменного сечения (не сжимается)
\begin{align}
    &\dfrac{v^2}{2}\;+\; gh\;+\;\dfrac{P}{\rho}\;=\; const, \nonumber
\end{align}
чем выше скорость, тем ниже давление.

С другой стороны: $v_1 S_1 \;=\; v_2 S_2$, т.е. чем меньше площадь, тем больше скорость.

Тогда, чем выше площадь, тем выше и давление, и наоборот.

Можно дать оценку на условие, когда можно считать жидкость несжимаемой.

Общее утверждение: $|\Delta \rho|\;\ll \;\rho \; \Rightarrow \; \dfrac{|\Delta \rho|}{\rho}\;\ll \;1.$
\begin{align}
    &\dfrac{v^2}{2}\;+\; gh\;+\;\dfrac{P}{\rho}\;=\; const, \nonumber \\
    &\dfrac{\rho \left( v_1^2 \;-\; v_2^2 \right)}{2}\;=\; P_2\;-\;P-1\;=\;\Delta P.
\end{align}

С другой стороны: $\Delta \rho \;=\; \dfrac{d \rho}{d P} \Delta P.$

\begin{align}
    &\dfrac{d \rho}{d P}\;=\; \dfrac{1}{c^2}, \\
    &\Delta \rho \cdot c^2\;=\; \dfrac{\rho}{2} \left( v_1^2\;-\;v_2^2 \right), 
\end{align}
\begin{equation*}
\begin{cases}
    \dfrac{\Delta \rho}{\rho}\;=\; \dfrac{\left( v_1^2\;-\;v_2^2 \right)}{2 c^2},\\
    \dfrac{|\Delta \rho|}{\rho}\;\ll \;1.
\end{cases}
\end{equation*}

\begin{align}
|v_2^2\;-\;v_1^2|\;\ll \;c^2.
\end{align}

Только вклад от:
\begin{align}
    &\eta\left(\frac{\partial v_x}{\partial y}+\frac{\partial v_y}{\partial x}\right)=\eta\left(\frac{\partial v_\varphi}{\partial r}+\frac{\partial v_r}{\partial \varphi}\right),\\
    &\tau_{r\varphi} = \eta\left(\frac{\partial v_\varphi}{\partial r}-\dfrac{v_\varphi}{r}\right), \\
    &d v_r=-v_\varphi d \varphi, \\
    &v\varphi = \omega r,\\
    &\tau_{r\varphi} = \eta r \frac{\partial \omega}{\partial r}.
\end{align}
Из этого можно сделать вывод, что если жидкость вращается как целое, то вязкого напряжения не будет, то есть если руками держать какой-то цилиндр, то всегда будет напряжение.

Рассмотрим пример:

Рассмотрим слой на расстоянии $r$: как мы уже получили, момент силы вязкости будет:
\begin{align}
    &M = 2 \underbrace{\pi r l}_{s} \cdot \underbrace{r}_{\text{плечо}} \cdot \underbrace{r\eta \frac{\partial \omega}{\partial r}}_{\text{напряжение}}, \\
    &M = 2\pi lr^3\eta \frac{\partial \omega}{\partial r}.
\end{align}
В случае установившегося режима эта величина не должна быть функцией от $r$, поскольку слои бы разгонялись, а этого не может быть. Пусть для определенности снаружи цилиндр разгоняет, а внутри тормозит. Должно быть $M = const$.
\begin{align}
    &r^3 \frac{\partial \omega}{\partial r} = const, \\
    &\omega = \dfrac{A}{r^2} + C.
\end{align}
Граничные условия:
\begin{align}
    \begin{cases}
        &\Omega_1 = \dfrac{A}{R_1^2} + C, \\
        &\Omega_2 = \dfrac{A}{R_2^2} + C
    \end{cases}
\end{align}
Откуда
\begin{align}
    &\Omega_2 - \Omega_1 = A \left(\dfrac{1}{R_2^2}-\dfrac{1}{R_1^2}\right),\\
    &A = \dfrac{\Omega_2 - \Omega_1}{R_1^2 - R_2^2}R_1^2R_2^2,\\
    &C = \dfrac{\Omega_2 R_2^2 - \Omega_1R_1^2}{R_2^2 - R_1^2}.
\end{align}
\begin{align}
    &\omega = \dfrac{\Omega_2 - \Omega_1}{R_1^2 - R_2^2}R_1^2R_2^2\dfrac{1}{r^2} + \dfrac{R_2^2\Omega_2 - R_1^2 \Omega_1}{R_2^2 -R_1^2}.
\end{align}
Момент сил вязкости, действующий на внутренний цилиндр:
\begin{align}
    2 \pi l \eta (-2A) = 4 \pi l \eta \dfrac{\Omega_1 - \Omega_2}{R_1^2 - R_2^2}R_1^2R_2^2 =const.
\end{align}
Если $\Omega_1 \; > \; \Omega_2$, то торможение. Если наоборот, то разгон.

Покажем, как измерить вязкость. Подвесим цилиндр на нитке:
\begin{align}
    &M = f \varphi,\\
    \implies&\eta = \dfrac{f\varphi(R_1^2-R_2^2}{4\pi l R_1^2R_2^2(\underset{=0}{\Omega_1} - \Omega_2)},\\
    \implies&\eta = \dfrac{f\varphi(R_2^2 - R_1^2)}{4\pi l R_1^2 R_2^2\Omega_2}.
\end{align}

## Стационарное течение жидкости по прямолинейной трубе. Течение Пуазёйля

Рассмотрим трубу постоянного сечения, а в ней ― вязкую несжимаемую жидкость. Допустим, что скорость меняется вдоль радиуса трубы (не зависит от прохождения жидкости вдоль трубы).


### Распределение скоростей

Запишем силы, которые действуют на маленький цилиндр жидкости радиуса $r$ и протяжённости $dx$ ― а именно, силы давления окружающей жидкости и внутреннего трения с другими слоями жидкостями. Если поток стационарен, то каждый такой кусочек жидкости в равновесии.
\begin{align}
    & d F_p = 2 \pi r d x \cdot \eta \;  \dfrac{d v}{d r}, \\
    & d F_\eta = \pi r^2 \Bigl(p(x) - p(x + d x) \Bigr) = - \pi r^2 \; \dfrac{d P}{d x} \, d x. \\
    & d F_p + d F_\eta = 0, \\
    & 2 \pi r \eta \; \dfrac{d v}{d r} \, d x = - \pi r^2 \; \dfrac{d P}{d x} \, d x, \\
    & 2 \eta \dfrac{d v}{d r} = r \dfrac{d P}{d x}, \\
    & \underbrace{2 \eta \; \dfrac{1}{r} \; \dfrac{d v}{d r}}_{\text{не зависит от } x} = \underbrace{\qquad \dfrac{d P}{d x} \qquad }_{\text{не зависит от } r} = \underbrace{\dfrac{P_2 - P_1}{l}}_{\text{не зависит от } x, r}. \\
    & \dfrac{d v}{d r} = \dfrac{r}{2 \eta} \, \dfrac{P_2 - P_1}{l} = - \dfrac{P_1 - P_2}{l} \, \dfrac{r}{2 \eta} = - \dfrac{\Delta P}{l} \, \dfrac{r}{2 \eta}. \\
    & v(R) - v(r) = - \dfrac{\Delta P}{4 \eta l} (R^2 - r^2).
\end{align}
У самой стенки вода должна быть неподвижна, поэтому $v(R) = 0$, что приводит к
\begin{align}
    & v(r) = \dfrac{\Delta P}{ 4 \eta l} (R^2 - r^2), \\
    & \boxed{v_0 = \max_{r} v(r) = v(0) = \dfrac{\Delta P R^2}{4 \eta l}}.
\end{align}

### Расход жидкости
Найдём расход (объём жидкости, протекающий через сечение за единицу времени) для пуазёйлевского течения, просуммировав кусочки изменения объёма, протекающие за единицу времени $\left(\dfrac{d^2 V}{d t}\right)$.

\begin{align}
    & Q = \dfrac{d V}{d t} = \int \limits^{r = R}_{r = 0} \dfrac{d^2 V}{d t} \int \limits^{r = R}_{r = 0} \dfrac{d x \, d S}{d t} = \int \limits^R_0 v(r) \cdot 2 \pi r d r \\
    & = 2 \pi \int \limits^R_0 \dfrac{\Delta P}{4 \eta l} (R^2 - r^2) \; d r = 2 \pi \int \limits^R_0 \dfrac{\Delta P}{4 \eta l} (R^2 - r^2) \; \dfrac{d (r^2)}{2} \\
    & = \dfrac{\pi \Delta P}{4 \eta l} \int \limits^{R^2}_0 (R^2 - x) d x = \dfrac{\pi  \Delta P}{4 \eta l} \left( R^4 - \dfrac{R^4}{2} \right) = \dfrac{\pi \Delta P}{8 \eta l} R^4. \\
    & \boxed{Q = \dfrac{d V}{d t} = \dfrac{\pi \Delta P}{8 \eta l} R^4} \text{ ― формула Пуазёйля}. \\
    & \boxed{\dfrac{d m}{d t} = \rho Q = \dfrac{\pi \rho \Delta P}{8 \eta l} R^4}. \\
    & Q = \pi R^2 \; \dfrac{v_0}{2}. \\
    & \langle v \rangle = \dfrac{\int \limits^{r = R}_{r = 0} v(r) d S }{S} = \dfrac{Q}{\pi R^2} = \dfrac{v_0}{2}, \\
    & Q = \pi R^2 \langle v \rangle.
\end{align}

### Энергетические характеристики течения

Разобьём сечение на маленькие кольца от $r$ до $r + d r$ и просуммируем мощность, которая нужна для поддержания течения через каждое из колец:
\begin{align}
    & P = \int^{r = 0 }_{r = 0} v(r) d F = \Delta P \int \limits^{R}_0 2 \pi r v(r) d r = Q \Delta P.
\end{align}

Вычислим и общую кинетическую энергию частиц, протёкших в потоке за единицу времени:
\begin{align}
    & K = \int \limits^{r = R}_{r = 0} \dfrac{\rho v^2}{2} \; v(r) d S = \int \limits^R_0 \dfrac{\rho v^3}{2} \; \cdot 2 \pi r \, d r \\
    & = \pi \rho \int \limits^R_0 \left( \dfrac{\Delta P}{4 \eta l} \right)^3 \Bigl(R^2 - r^2 \Bigr)^3 r \, d r \\
    & = \dfrac{\pi \rho}{8} \left( \dfrac{\Delta P}{4 \eta l} \right)^3 R^8 = \dfrac{1}{4} Q v_0^2. \\
    & \boxed{K = \dfrac{1}{4} Q v_0^2}.
\end{align}

Сравнивая эти две величины, можно сделать вывод о характере течения: так, если $|P| \ll |K|$, то в данном случае жидкость можно считать идеальной и пренебрегать её внутренним трением.
