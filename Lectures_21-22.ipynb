## Классическая теория теплоёмкости кристаллов


<br>
Рассмотрим простейшую модель **кристалла**: решётка, в узлах которой находятся атомы, принимаемые за материальные точки. Они совершают тепловые колебания около положения равновесия. Предположим, что колебания малы, тогда они будут гармоническими. 

<br>

Запишем энергию по теореме Кёнига: будет кинетическая энергия относительно положения равновесия и потенциальная энергия:

<br>

\begin{align}
    &\varepsilon_{i}\;=\;\dfrac{m_{i}\dot{\vec{x}}\;_{i}^2}{2}\;+\; \dfrac{k \dot{\vec{x}}\;_{i}^2}{2} \\
    &Z \; \sim \; \int \bigcap_{j=1}^{N} \left[ (d \vec{p}_{j} d \vec{x}_{i}) \exp{-\dfrac{m_{i}\dot{\vec{x}}\;_{i}^2}{2kT}\;-\; \dfrac{k \dot{\vec{x}}\;_{i}^2}{2kT}} \right]\\
\end{align}

<br>

Ранее мы получали, что теплоёмкость, получаемая в случае одного слагаемого в энергии вида $x^2/2$ равнялась $3/2kN$ . Теперь же у нас их два. Следовательно, молярная теплоемкость $3R$ (**закон Дюлонга-Пти**)

<br>

\begin{align}
    c\;=\;\dfrac{3R \nu}{m}\;=\; \dfrac{3R}{M}\;=\;\dfrac{3R}{m_0 N_{A}}.
\end{align}
То есть $c\cdot(m_0N_A)=3R=\text{const}.$

<br>

Если же речь идет о соединении, то теплоемкость в простейшем рассмотрении будет уже равно $3Rn$ (\textbf{закон Джоуля-Коппа}).

## Недостатки классической теории

<br>

Оказывается, классическая теория имеет ряд ограничений. В частности, она не описывает температурную зависимость теплоемкости. Аналогичные рассуждения можно провести для вращательных степеней свободы:
$$\varepsilon_{e} \sim \dfrac{\hslash}{I}\;l\;(l\;+\;1),$$

<br>

т. е. есть вращения с нулевой энергией есть (точнее допустимы отсутвия вращений).

<br>

Если $kT \ll \hslash^2 / I$, то можно не учитывать вращательные степени свободы. Момент инерции определяет, какие учитывать, а какие \$-\$ нет.

## Гармонический осцилятор

<br>

\begin{align}
    &\varepsilon_{kj}\;=\;\hslash \omega \left( k \;+\; \dfrac{1}{2} \right), 
\end{align}
\begin{align}
    &Z\;=\; \sum \limits_{{k}} \exp{- \dfrac{\sum \limits_{j\;=\;1} \varepsilon_{kj}}{kT}}\;=\; \left( \sum \limits_{k\;=\;0}^{\infty} \exp{- \dfrac{\hslash \omega \left( k \;+\; \dfrac{1}{2} \right)}{k_{\text{Б}}T}} \right)^{N}\nonumber\\
    &\;=\; \left( \dfrac{\exp{- \dfrac{\hslash \omega}{2k_{\text{Б}}T}}}{1\;-\; \exp{- \dfrac{\hslash \omega}{kT}}} \right)^{N}.
\end{align}

<br>


Не все степени свободы вносят вклад! Говорят, что они "заморожены". Классическая теория в своем рассмотрении накладывала связи, что эквивалентно заморозке некоторых степеней свободы.

<br>

**Пример:** гармонический осцилятор (2-х атомная молекула)
$$\varepsilon_{k}\;=\;\hslash \omega \left(k\;+\; \dfrac{1}{2} \right).$$
В чем особенность? Самое низкое по энергии состояние \textbf{не равно нулю}!
$$\varepsilon_0 \;=\; \dfrac{\hslash \omega}{2}$$
Рассмотренная система находится при низкой температуре: $$kT \ll \hslash \omega.$$
Тогда тепловой энергии не хватает для того, чтобы возбудить колебательную степень свободы, т. е. не участвует энергопоглощение, следовательно, не влияет на теплоемкость.
$$T_{\omega}\;=\; \dfrac{\hslash \omega}{k_{\text{Б}}},$$
где $\omega$ - характеристика вещества.

<br>

\begin{align}
    &\Rightarrow F=NkT\ln 2\sinh\frac{\hslash\omega}{2 k_\text{Б} T}, \\
    & S=-Nk\ln 2 \sinh \frac{\hslash\omega}{2k_\text{Б} T}+NkT\frac{\cosh \frac{\hslash \omega}{2 k_\text{Б} T}}{\sinh \frac{\hslash \omega}{2 k_\text{Б} T }} \frac{\hslash \omega}{2 k_\text{Б} T^2} \Rightarrow \\
    &\Rightarrow C=N\cdot\left(\frac{\hslash \omega}{2 k_\text{Б} T}\right)^2 k\cdot \frac{1}{\sinh^2\left(\frac{\hslash \omega}{2 k_\text{Б} T}\right)} \Rightarrow \\
    & \underset{T \rightarrow \infty}{\Rightarrow} C_{\text{частицы}}=k.
\end{align}

## Статистика Ферми-Дирака и Бозе-Эйнштейна

<br>


Два класса частиц: фермионы (${1}/{2}$), бозоны ($1\;+/-\;k$).

<br>


Классической (больцмановское) рассмотрение стоит на принципальной различности частиц. Если у нас даже есть 2 одинаковые частицы, каждая из которых находится в каком-то состоянии, то совокупное состояние будет отличаться от того, когда эти частицы поменяются между собой.

<br>


В квантовой статистике ситуация иная: тождественные частицы принципально неразличимы. 

<br>


**Разница:** Ф-Д: 0,1 (не могут быть 2 фермиона в 1-м  состоянии по **принципу Паули**), Б-Э: 0, 1, 2, ..., $\infty$.

### Канонический ансамбль

<br>

Рассмотрим закрытую систему (то есть частицы не обмениваются), тогда берём распределение для канонического ансамбля (считаем среднеквадратичную флуктуацию энергии):

<br>
\begin{align}
    &W_n=\dfrac{1}{z}\exp{-\dfrac{E_n}{kT}}\implies \nonumber 
\end{align}
\begin{align}
    &\overline{E} = \sum\limits_n W_nE_n = \dfrac{1}{z}\sum\limits_n E_n\exp{-\dfrac{E_n}{kT}} \nonumber \\
    &\implies \overline{E} = - \dfrac{1}{z} \dfrac{\partial z}{\partial \beta},
\end{align}
где $\beta = 1/kT.$\\Ещё надо: 
\begin{align}
    &\overline{E^2}=\sum\limits_n E_n^2 W_n = \dfrac{1}{z} \dfrac{\partial^2 z}{\partial \beta^2}\nonumber \\
    &\implies \overline{\Delta E^2}=\dfrac{1}{z}\dfrac{\partial^2z}{\partial\beta^2} - \dfrac{1}{z^2}\cdot (\dfrac{\partial{z}}{\partial \beta})^2.
\end{align}
<br>

С другой стороны:

<br>

\begin{align}
    & \dfrac{\partial \overline{E}}{\partial \beta} = \dfrac{\partial}{\partial \beta} \left( - \dfrac{1}{z} \, \dfrac{\partial z}{\partial \beta} \right) = \dfrac{1}{z^2} \left( \dfrac{\partial z}{\partial \beta} \right)^2 - \dfrac{1}{z} \, \dfrac{\partial^2 z}{\partial \beta^2}, \\
    & \overline{\Delta E^2} = - \dfrac{\partial \overline{E}}{\partial \beta} = kT^2 \left( \dfrac{\partial \overline{E}}{\partial T} \right) = kT^2 C_V, \\
    & d \beta = d \left( \dfrac{1}{kT} \right) = - \dfrac{1}{kT^2} d T.
\end{align}

### Большой канонический ансамбль

Рассмотрим среднеквадратичную флуктуацию числа частиц в большом каноническом ансамбле:
<br>
\begin{align}
    & \overline{\Delta N^2} = \overline{N^2} - \overline{N}^2.
\end{align}
<br>

Выразим среднее число частиц в системе, суммируя по состояниям со всеми возможными количествами частиц всех возможных энергий.
<br>
\begin{align}    
    & \overline{N} = \sum_n w_n N_n = \dfrac{\displaystyle \sum_n N_n \exp\left( - \dfrac{E_n - \mu N_n}{kT} \right)}{\displaystyle \sum_n \exp\left( - \dfrac{E_n - \mu N_n}{kT} \right)}, \\
    & z = \sum_n \exp\left( - \dfrac{E_n - \mu N_n}{kT} \right), \\
    & \overline{N} = \dfrac{kT}{z} \dfrac{\partial z}{\partial \mu}. \\
    & \text{Аналогично, } \overline{N^2} = \dfrac{\displaystyle \sum_n N_n^2 \exp\left( - \dfrac{E_n - \mu N_n}{kT} \right)}{\displaystyle \sum_n \exp\left( - \dfrac{E_n - \mu N_n}{kT} \right)} = \dfrac{(kT)^2}{z} \dfrac{\partial^2 z}{\partial \mu^2}. \\
    & \overline{\Delta N^2} = (kT)^2 \left(\dfrac{1}{z} \, \dfrac{\partial^2 z}{\partial \mu^2} - \dfrac{1}{z^2} \left( \dfrac{\partial z}{\partial \mu} \right)^2 \right).
\end{align}
<br>
С другой стороны:
<br>
\begin{align}
    & \dfrac{\partial \overline{N}}{\partial \mu} = \dfrac{\partial}{\partial \mu} \left( \dfrac{kT}{z} \, \dfrac{\partial z}{\partial \mu} \right) = - \dfrac{kT}{z^2} \left( \dfrac{\partial z}{\partial \mu} \right)^2 + \dfrac{kT}{z} \, \dfrac{\partial^2 z}{\partial \mu^2} = \dfrac{\overline{\Delta N^2}}{kT}, \\
    & \overline{\Delta N^2} = kT \dfrac{\partial \overline{N}}{\partial \mu}.
\end{align}

### Распределение вероятностей для флуктуаций
<br>
Вспомним выражение для энтропии на основе статвеса:
\begin{align}
    & S = k \ln \Gamma.
\end{align}
Для удобства далее опустим постоянную Больцмана $k$. При необходимости её всегда можно восстановить при помощи размерного анализа.
\begin{align}
    & S = \ln \Gamma, \\
    & \Gamma = \exp{S}.
\end{align}
<br>
Вероятность, как было установлено ранее, пропорциональна статвесу,
\begin{align}
    & w = C \Gamma = C \exp{S}.
\end{align}
<br>
Стоит заметить, что нет никакой нужды ограничивать $w$ дискретными значениями ― распределение вполне может быть непрерывным.
<br>

Рассмотрим произвольную замкнутую систему в равновесии. Пусть, есть какой-нибудь параметр ($E$, $V$, ...), которых характеризирует систему. Обозначим его символом $x$. Если мы знаем зависимость $S(x)$, то получаем и распределение для $x$ вида $w = C \exp{S(x)}$.
<br>

Равновесное значение энтропии соответствует сред. значению нашего параметра.
<br>
\begin{align}
    & S_0 = S(\overline{x}). 
\end{align}
<br>
Для малых отклонений от равновесия 
<br>
\begin{align}
    S(x) = S(\overline{x}) + \left. \dfrac{\partial S}{\partial x}\right|_{x = \overline{x}} (x - \overline{x}) + \dfrac{1}{2!} \, \left. \dfrac{\partial^2 S}{\partial x^2}\right|_{x = \overline{x}} (x - \overline{x})^2 + ...
\end{align}

<br>
Кроме того, в равновесном состоянии энтропия максимальна:

<br>
\begin{align}
    & \left. \dfrac{\partial S}{\partial x}\right|_{x = \overline{x}} = 0, \quad \left. \dfrac{\partial^2 S}{\partial x^2}\right|_{x = \overline{x}} < 0. \\
    & \text{Обозначим } \alpha = - \left. \dfrac{\partial^2 S}{\partial x^2}\right|_{x = \overline{x}}. \\
    & w(x) = C \exp\left(S(\overline{x}) - \dfrac{\alpha}{2} (x - \overline{x})^2 \right), \\
    &\text{Если } \tilde{C} = C \exp \Bigl( S(\overline{x}) \Bigr), \\
    & w(x) = \tilde{C} \exp \left( - \dfrac{\alpha}{2} (x - \overline{x})^2 \right).
\end{align}

<br>
Из этого можно получить распределение Гаусса для флуктуации $\Delta x = x - \overline{x}$:
<br>
\begin{align}
    & w(\Delta x) = \tilde{C} \exp \left( - \dfrac{\alpha}{2} \Delta x^2 \right). \\
    & \int \limits^{+ \infty}_{- \infty} w(\Delta x) d (\Delta x) = 1, \\
    & \tilde{C} \int \limits^{+ \infty}_{- \infty} \exp \left( - \dfrac{\alpha}{2} \Delta x^2 \right) d (\Delta x) = \tilde{C} \sqrt{\dfrac{2}{\alpha}} \sqrt{\pi} = 1, \\
    & \tilde{C} = \sqrt{\dfrac{\alpha}{2 \pi}}.
\end{align}
<br>
Итого:
<br>
\begin{align}
    & \boxed{w(\Delta x) = \sqrt{\dfrac{\alpha}{2 \pi}} \exp \left( - \dfrac{\alpha}{2} \Delta x^2 \right)}.
\end{align}
<br>
Рассмотрим среднеквадратичное отклонение $\overline{\Delta x^2}$ при $\overline{\Delta x} = 0$.
<br>
\begin{align}
    & \int \limits^{+ \infty}_{- \infty} \Delta x^2 \cdot \sqrt{\dfrac{\alpha}{2 \pi}} \exp \left( - \dfrac{\alpha}{2} \Delta x^2 \right) d (\Delta x), \\
    & = \dfrac{2 \pi}{\alpha \sqrt{\pi}}  \int \limits^{+ \infty}_{- \infty} \left( \sqrt{\dfrac{\alpha}{2 \pi}} \Delta x^2 \right) \exp \left( - \dfrac{\alpha}{2} \Delta x^2 \right) d \left( \sqrt{\dfrac{\alpha}{2 \pi}}\Delta x) \right),
\end{align}
<br>
\begin{align}
    & = \dfrac{2}{\alpha \sqrt{\pi}} \int \limits^{+ \infty}_{- \infty} y^2 \exp(- y^2) d y = \dfrac{2}{\alpha \sqrt{\pi}} \left. \left( - \dfrac{\partial}{\partial \gamma} \right) \int \limits^{+ \infty}_{- \infty} \exp{- y^2 \gamma} d y \right|_{\gamma = 1} \\
    & = \dfrac{2}{\alpha \sqrt{\pi}} \left. \left( - \dfrac{\partial}{\partial \gamma} \right) \dfrac{\sqrt{\pi}}{\sqrt{\gamma}} \right|_{\gamma = 1} = \dfrac{1}{ \alpha}. \\
    & \overline{\Delta x^2} = \dfrac{1}{\alpha}.
\end{align}
<br>
В целом, для известной функции флуктуации $f(\Delta x)$
<br>
\begin{align}
    \boxed{\overline{(\Delta f)^2} = \left. \left( \dfrac{d f}{d (\Delta x)} \right)^2 \right|_{\Delta x = 0} \!\!\overline{\Delta x^2}}.
\end{align}
