# Ансамбли

Остановились на:

$$N_1,\;N_2,\;...\; : \; N_1\;+\;N_2\;+\;...\;=\;N\;=\; const,$$
$$N_1 \varepsilon_1 \;+\; N_2 \varepsilon_2 \;+\;...\;=\;E\;=\; const.$$

$\left\langle  N_{i}\right\rangle\;=\;\exp${$-\alpha \varepsilon_i$} $\exp${$- \beta$}.

Приводим в контакт с термостатом:

\begin{align}
    &d E\;=\; T d S,\\
    &d E \;=\; \sum\limits_{i} \varepsilon_{i} d \left\langle N_{i}\right\rangle\;=\; T \left( -k \sum \ln{\overline{N_{i}}} d N_{i} \right), \\
    &=\;T \left( -k \sum -\alpha \varepsilon_{i} d N_{i}\;+\; const \right), \\
    &\approx kT \alpha \sum \varepsilon_{i} d N_{i}. \\
    &\boxed{\alpha \;=\; \dfrac{1}{kT}}.\\
    &\boxed{\left\langle N_{i}\right\rangle\;=\;\exp{-\beta} \exp{-\dfrac{\varepsilon_{i}}{kT}}}.\\
    &\sum\left\langle N_{i}\right\rangle\;=\; \exp{-\beta} \sum\limits_{i} \exp{-\alpha \varepsilon_{i}}\;=\; N \\
    &\Rightarrow \left\langle N_{i}\right\rangle\;=\; N \dfrac{\exp{-\frac{\varepsilon_{i}}{kT}}}{\sum\limits_{i} \exp{- \frac{\varepsilon_{i}}{kT}}}.\\
    &\dfrac{\left\langle N_{i}\right\rangle}{N}\;=\; \dfrac{\exp{-\frac{\varepsilon_{i}}{kT}}}{\sum\limits_{i} \exp{- \frac{\varepsilon_{i}}{kT}}}.
\end{align}

## Элементы квантовой статфизики

Как и раньше будем рассматривать равновесное состояние, то есть рассматриваем систему в течение времени, превышающего время её релаксации.

Для определения частицы в фазовом пространстве надо выбрать некий объем: $\delta r \cdot \delta p$.
В квантовом случае на одну степень свободы приходится объем: $2 \pi \hslash$. Тогда число состояний можно будет определить как: $d \Gamma\;=\; \dfrac{d r d p}{(2 \pi \hslash)^3}$, где Г  $\;-\;$ все число состояний.

(Сперва мы фиксируем энергию и число частиц). Уже получали. Вероятность каждого: $W_{i}\;=\; \dfrac{1}{\Gamma}$,
$$\sum \limits_{i}\;=\; \dfrac{\Gamma}{\Gamma}\;=\; 1.$$

Ну а сама энтропия: $S\;=\; -k \sum W_{i} \ln{W_{i}}.$ 

Пусть $k\;=\;1$ (везде, где после будет $T$ просто добавим $k$). 
Отметим, что в такой постановке, когда система находится в состояниях с равной энергией с одинаковой вероятностью, энтропия будет максимальна.

## Метод множителей Лагранжа
Рассмотрим пример, поясняющий метод множителей Лагранжа, который пригодится нам дальше. Допустим у нас есть плоскость, задающаяся уравнением $f(x,y)=x+y$, и мы хотим найти её минимум. Тогда необходимо задать дополнительные условия. Пусть наше условие будет: 

\begin{align}
   &x^2+y^2=\nonumber1=const.
\end{align}

Теперь рассмотрим новый **функционал**: 

 \begin{align}
     &\tilde{f}(x, y)=x+\nonumber y+\lambda(x^2+y^2).
 \end{align}
 
т.е. мы будем выбирать такие траектории по $x, y$, чтобы $d (x^2+y^2)=0$. Тогда найдём минимум: 

 \begin{align*}
     &\frac{\partial \tilde{f}}{\partial x} = 1 + 2\lambda x=0\;\;\frac{\partial \tilde{f}}{\partial y}=1+2\lambda y=0,\\
     &x=-\frac{1}{2\lambda},\;\;y=-\frac{1}{2\lambda}.\\
     &x^2+y^2=1\;\;\Rightarrow \frac{1}{4\lambda^2}+\frac{1}{4\lambda^2}=1\;\;\Rightarrow \lambda=\frac{1}{\sqrt{2}} \Rightarrow \\
     &\Rightarrow x=-\frac{1}{\sqrt{2}},\;\;y=-\frac{1}{\sqrt{2}}.
 \end{align*}

## Ансамбли
**Статистический ансамбль**  $\;-\;$ совокупность большого числа экземпляров физической системы частиц, каждая из которых находится в одинаковом макросостоянии. Однако, микросостоянии могут быть разными. Концепт ансамблей активно используется в статфизике.

**Микроканонический ансамбль**  $\;-\;$ ансамбль, когда фиксированные энергия изолированной системы и число частиц в ней. 

**Канонический ансамбль**  $\;-\;$ ансамбль, обменивающийся энергией с окружающей средой, но не обменивающийся с ней веществом, находящийся с ней в тепловом равновесии (т. е. сохраняется значение средней энергии $\left\langle E\right\rangle\;=\;\sum E_i\rho_i=\text{Tr}H\rho$ или просто $\text{Tr}\hat{H}\hat{\rho}$).

**Большой канонический ансамбль (БКА)**  $\;-\;$ ансамбль, обменивающийся не только энергией, но и частицами с окружающей средой, т. е. со временем устанавливается не только тепловое равновесие (среднее значение энергии), но и равновесие по составу (среднее значение частиц: $\left\langle N\right\rangle\;=\;\text{Tr}\rho N\;=\;\sum w_i N_i$).

В квантовой статфизике используют термин **статистического оператора** $\hat{\rho}$, а само усреднение определяют **оператором $Tr$ (trace)**.

$$S\;=\;-Tr(\hat{\rho} \ln{\hat{\rho}}).$$
### Микроканонический ансамбль
Посмотрим, чему будет равен статистический оператор в случае микроканонического ансамбля. Так как вероятность состояния определяется энергией, то все состояния ансамбля таких систем равновероятны. Необходимо учесть нормировочное условие:

\begin{align}
    &\sum_i{\rho} (i)= Tr \rho = 1\nonumber.
\end{align}
$\Rightarrow$ должна быть равна нулю производная от $Tr(\rho \ln{\rho)} + \alpha Tr \rho$:

\begin{align}
    &Tr(1 + \ln{\rho} + \alpha) d \rho = 0\\ 
    &\Rightarrow \texttt{p  $\;-\;$ const    } (\ln{\rho} = - (1 + \alpha)),\\
    &\rho = \exp{- (1 + \alpha)},\\
    &\sum_i {\rho_i} = 1 = \Gamma \exp{- (1 + \alpha)},\\
    &\alpha = - \ln{{\frac{1}{\Gamma}}} - 1.\\
\end{align}

### Канонический ансамбль
Теперь рассмотрим канонический ансамбль и найдём минимум энтропии, используя метод множителей Лагранжа.

\begin{align}
    &Tr\rho \ln \rho + \lambda Tr \rho + \beta Tr H\rho \Rightarrow \\
    &\Rightarrow Tr(1+\ln \rho + \lambda + \beta H)\delta \rho=0  (d\rho = \triangle \rho).
\end{align}

Или:

\begin{align}
    &\sum_i (1+\ln w_i+\lambda+\beta E_i)d w_i=0 \Rightarrow \\
    &\Rightarrow \rho=w_i=\exp{-(1+\lambda)-\beta E_i}.
\end{align}

Вспоминаем про условия: 

\begin{align}
    &1) \sum w_i=1=\exp{-(1+\lambda)}\sum_i \exp{-\beta E_i}=1 \Rightarrow \\
    &\Rightarrow \boxed{\rho = \frac{1}{Z}\exp{-\beta H}},\;\;\boxed{w_i=\frac{1}{Z} \exp{-\beta E_i}}. \\
    &2) \left\langle E\right\rangle=Tr \rho H=\sum_i \frac{E_i \exp{-\beta E_i}}{Z}=-\frac{\partial \ln Z}{\partial \beta} = \\
    &=\frac{-1}{Z} \sum_i \frac{\partial \exp{- \beta E_i}}{\partial \beta},
\end{align}
где $Z$ - **статсумма**. $Z\;=\;\sum\exp{-\beta E_i}$

### Большой канонический ансамбль
Теперь рассмотрим ансамбль, у которого мы зафиксировали среднее число частиц и среднюю энергию (БКА):

\begin{align}
    \left\langle  N \right\rangle\;=\;\text{Tr}\rho N\;=\;\sum\limits_iw_iN_i.
\end{align}

Новый функционал:

\begin{align}
    &\implies\;\text{Tr}\rho\ln{\rho}\;+\;\lambda\text{Tr}\rho\;+\;\beta\text{Tr}\rho H\;+\;\gamma\text{Tr}\rho N\\
    &\implies\;\text{Tr}\left(1\;+\;\ln{\rho}\;+\;\lambda\;+\;\beta H\;+\;\gamma N\right)\delta\rho\;=\;0\\
    &\text{Или:}\nonumber\\
    &\sum\limits_i\left(1\;+\;\ln{w_i}\;+\;\lambda\;+\;\beta E_i\;+\;\gamma N_i\right)d w_i\;=\;0,
\end{align}
где каждая $i$ означает состояние, т. е. происходит перебор не только по энергиям, но и по числу частиц.

\begin{align}
    &\implies \rho_i\;=\;\exp{-\left(\lambda\;+\;1\right)-\beta E_i-\gamma N_i},\\
    &\hat{\rho}\;=\;\exp{-\left(\lambda\;+\;1\right)\;-\;\beta\hat{H}\;-\;\gamma\hat{N}}\\
    &w_i\;=\;\exp{-\left(\lambda\;+\;1\right)\;-\;\beta E_i\;-\;\gamma N_i}.
\end{align}

Теперь:

\begin{align}
    &\sum w_i\;=\;\exp{-\left(\lambda\;+\;1\right)}\underbrace{\sum\limits_i\exp{-\beta H-\gamma N}}_{Z}\;=\;1\\
    &\boxed{\hat{\rho}\;=\;\dfrac{\exp{-\beta\hat{H}\;-\;\gamma\hat{N}}}{Z}}.
\end{align}

Добавим:

\begin{align}
\left\langle  N \right\rangle\;=\;\text{Tr}\hat{\rho}\hat{N}\;=\;-\dfrac{\partial}{\partial\gamma}\ln{Z}.
\end{align}

## Связь между термодинамикой и статфизикой
Связь между термодинамикой и статфизикой можно определить через $\left\langle  N \right\rangle$ и $\left\langle  E \right\rangle$. В микроканоническом ансамбле у нас были $E, N, V$ (экстенсивные параметры). В каноническом и большом каноническом мы имели дело уже с $\gamma, \beta$.

\begin{align}
    &d U\equiv d E=Td S - Pd V.\nonumber
\end{align}

Это в случае фиксированного числа частиц. Мы помним, что энергия, вносимая в систему при внесении в неё одной частицы, называется химическим потенциалом. Поэтому, в общем случае:

\begin{align}
    &d E\;=\;Td S\;-\;Pd V\;+\;\mu d N.\\
    &\underbrace{T}_{\text{инт.}}\;=\;\left(\dfrac{\partial E}{\partial S}\right)_{V,N},\;\underbrace{P}_{\text{инт.}}\;=\;-\left(\dfrac{\partial E}{\partial V}\right)_{S,N},\;\underbrace{\mu}_{\text{инт.}}\;=\;\left(\dfrac{\partial E}{\partial N}\right)_{V,S}.
\end{align}

Аналогично можно расписать другие потенциалы $F, H, G$, а также ввести новый  $\;-\;$ **большой термодинамический потенциал (потенциал Ландау)**:

\begin{align}
    &\Omega\;=\;F\;-\;\mu N,\\
    &d\Omega\;=\;-Sd T\;-Pd V\;-\;Nd\mu.
\end{align}

Теперь установим соответствие:

* Канонический ансамбль:

\begin{align}
    &\hat{\rho}\;=\;\dfrac{1}{Z}\exp{-\beta\hat{H}},\\
    &S\;=\;-\text{Tr}\hat{\rho}\ln{\hat\rho}\;\underset{\text{убираем}\;\wedge}{=}\;-\sum\limits_i\dfrac{1}{Z}\exp{-\beta E_i}\left(-\ln{Z}\;-\;\beta E_i\right)=\\
    &=\ln{Z}\;+\;\beta U,\qquad U = \left\langle  E\right\rangle\;=\;E.\\
    &U\;=\;F\;+\;TS\;=\;E\implies S\;=\;\dfrac{E-F}{T}\\
    &k\ln{Z}\;+\;k\beta U\;=\; S\\
    &A\;=\;-kT\ln{Z}, k\beta\;=\;\dfrac{1}{T}\\
    &\left(d S\;=\;d\ln{Z}\;+\;\beta U\;+\; Ud\beta\right).\\
    &\dfrac{1}{\beta}\;=\;\left(\dfrac{\partial U}{\partial S}\right)_{V,N}\;=\;T\\
    &\implies Z\;=\;\exp{-\dfrac{F}{kT}},\\
    &\implies \boxed{\rho\;=\;\dfrac{1}{Z}\exp{-\beta E}\;=\;\exp{\dfrac{F-E}{kT}}}.
\end{align}

Получили решение Гиббса для канонического ансамбля.

* Распределение в случае БКА.

\begin{align}
    &T\ln{Z}\;=\;TS-U+\mu N,\\
    &-T\ln{Z}\;=\;\underbrace{\underbrace{U-TS}_{F}-\mu N}_{\Omega}\\
    &\implies\boxed{\Omega\;=\;-kT\ln{Z}}\implies Z\;=\;\exp{-\dfrac{\Omega}{kT}},\\
    &\rho\;=\;\dfrac{1}{Z}\exp{-\beta H-\gamma N}\;=\;\dfrac{1}{Z}\exp{-\dfrac{H}{T}\;+\;\dfrac{\mu N}{T}}=\\
    &=\boxed{\exp{-\dfrac{H-\mu N-\Omega}{T}}}.
\end{align}

## Простые примеры
Рассмотрим ансамбль одинаковых частиц идеального газа с их фиксированным числом.

\begin{align}
    &\hat{\rho} = \dfrac{1}{Z}\exp{-\beta\hat{H}}= \dfrac{\exp{-\dfrac{E_i}{kT}}}{\sum\limits_i\exp{-\dfrac{E_i}{kT}}}.
\end{align}

Взаимодействия между молекулами идеального газа нет, следовательно: 
$$E_i = \sum\limits_j \varepsilon_j\;=\;\sum\limits_j\dfrac{p^2_j}{2m_j}\;=\;\sum\limits_j\dfrac{p^2_j}{2m}$$
В это выражение также может входить ещё какая-нибудь энергия, описывающая внутренние степени свободы.

Учтём неизменность состояния при перестановке частиц и непрерывность импульса.

\begin{align}
    &Z = \dfrac{1}{N!}\sum\limits_{\{i\}}\exp{-\dfrac{E_i}{kT}}=\sum\limits_{\{i\}}\dfrac{\exp{-\dfrac{1}{kT}\sum\limits_j\dfrac{p_{ij}^2}{2m}}}{N!}.
\end{align}

У каждой частицы ещё есть координата:

\begin{align}
    &Z = \dfrac{1}{N!}\sum\limits_{\{i\}}\exp{-\dfrac{1}{kT}\sum\limits_j\dfrac{p_{ij}^2}{2m}} \dfrac{\Delta p_{x1}\Delta p_{y1}\Delta p_{z1}}{\Delta p_{x1}\Delta p_{y1}\Delta p_{z1}}\dfrac{\Delta x_1 \Delta y_1 \Delta z_1}{\Delta x_1 \Delta y_1 \Delta z_1}= \nonumber \\
    &=\dfrac{1}{N!}\dfrac{1}{(2\pi\hslash)^{3N}}\int \left(\prod\limits_j^Nd\vec{p}_jd\vec{r}_j\right)\exp{-\dfrac{1}{T}\sum\limits_j\dfrac{p_j^2}{2m}}=\nonumber \\
    &=\dfrac{1}{N!(2\pi\hslash)^{3N}}V^N\int\left(\prod_j^Nd \vec{p}_j\right)\exp{-\dfrac{1}{T}\sum\limits_j\dfrac{p_j^2}{2m}}=\nonumber \\
    &=\dfrac{V^N}{N!(2\pi\hslash)^{3N}}\int \prod_j \left(d \vec{p}_j \; \exp{-\dfrac{1}{T}\dfrac{p_j^2}{2m}}\right). \\
    &\int d \vec{p} \; \exp{-\dfrac{p^2}{2mT}}=\left(\int d p_x \; \exp{-\dfrac{p_x^2}{2mT}}\right)^3 =\\
    &=\left(\sqrt{2mT}\int\limits_{-\infty}^\infty d y \;\exp{-y^2}\right)^3 = (2mT\pi)^{3/2}\\
    &\implies Z= \dfrac{V^N(2m\pi T)^{\frac{3N}{2}}}{N!(2\pi\hslash)^{3N}}.
\end{align}

Находим термодинамику:

\begin{align}
    &U = - \left(\dfrac{\partial}{\partial \beta} \ln Z\right)_{V,N}=T^2\dfrac{\partial}{\partial T}\ln Z = \dfrac{3}{2} NT. \\
    &F = -kT\ln Z = - kTN\left(\ln \dfrac{VT^\frac{3}{2}}{N}+\dots\right). \\
    &N!=\sqrt{2\pi N}\left(\dfrac{N}{e}\right)^N \sim N^{N+\frac{1}{2}} = N^{N(1+\frac{1}{2N})},\nonumber \\
    &P = - \Big(\frac{\partial F}{\partial V} \Big)_{T, N} =\dfrac{kTN}{V}=\dfrac{kTN_a\nu}{V},\nonumber \\
    &PV = \nu RT. \\
    &F = -kT\ln\left[\dfrac{eV}{N} \left(\dfrac{mT}{2\pi\hslash^2}\right)^{3/2}\right].\\
    &S = - \Big(\frac{\partial F}{\partial T} \Big)_{V, N}=kN\ln\left[\dfrac{eV}{N} \left(\dfrac{mT}{2\pi\hslash^2}\right)^{3/2}\right]+ \dfrac{3}{2}kN.\\
    &C_V =T \Big(\frac{\partial S}{\partial T} \Big)_{V, N} = \dfrac{3}{2}kN\label{eq:cv}.
\end{align}

## Двухуровневая система
Рассмотрим **«двухуровневую» систему** $\;-\;$ такую систему, что каждая её конкретная частица имеет энергию $+ \Delta$ или $- \Delta$. Обозначая энергию $j$-той частицы $\varepsilon_j$, получаем для $N$ частиц

\begin{align}
    & H = \sum_j \varepsilon_j, \quad \varepsilon_j = \pm \Delta.
\end{align}

Посчитаем статсумму. У нас будет два суммирования: по состояниям системы по индексу $i$ и в каждом состоянии по частицам системы по индексу $j$. Обозначим $N^+$ частицы с энергией $+ \Delta$, а $N^-$ частицы с энергией $- \Delta$. 

\begin{align}
    & Z = \sum_i \exp{- \sum_j \dfrac{\varepsilon_{ij}}{kT}} = \sum_i \exp{- \dfrac{N^+_i \Delta + N^-_i (- \Delta)}{kT}} \\
    & = \sum_i \left(\exp{- \dfrac{\Delta}{kT}}\right)^{N^+_i} \left(\exp{- \dfrac{(- \Delta)}{kT}}\right)^{N^-_i} \\
    & = \sum_i \left(\exp{- \dfrac{\Delta}{kT}}\right)^{N^+_i} \left(\exp{- \dfrac{(- \Delta)}{kT}}\right)^{N - N^+_i}.
\end{align}

Такими трюками мы пришли к чему-то, что похоже на бином Ньютона, не хватает лишь нужного коэффициента. Чтобы заполучить коэффициент, перейдём от суммы по состояниям к сумме по количеству частиц с энергией $+ \Delta$. Существует $C^{N^+}_N$ способов выбрать какие-то $N^+$ частиц из набора $N$ частиц. Значит,

\begin{align}
    & Z = \sum_{N^+ = 0}^N C^{N^+}_N \left(\exp{- \dfrac{\Delta}{kT}}\right)^{N^+_i} \left(\exp{- \dfrac{(- \Delta)}{kT}}\right)^{N - N^+_i} \\
    & = \left( \exp{-\dfrac{\Delta}{kT}} + \exp{ \dfrac{\Delta}{kT}} \right)^N = \left( 2 ch \dfrac{\Delta}{kT} \right)^N.
\end{align}

Благодаря мощи статистики автоматически получим отсюда значение свободной энергии и энтропию:

\begin{align}
    & F = - kT \ln Z = - kT \ln \left( 2 ch \dfrac{\Delta}{kT}\right)^N, \\
    & F = - N kT \ln \left( 2 ch \dfrac{\Delta}{kT} \right). \\
    & d F = - S d T - P d V + \mu d N, \\
    &S = -\Big(\frac{\partial F}{\partial T} \Big)_{V, N}. \\
    & S = N k \ln \left( 2 ch \dfrac{\Delta}{kT} \right) + N kT \; \dfrac{sh \dfrac{\Delta}{kT}}{ch \dfrac{\Delta}{kT}} \cdot \left( - \dfrac{\Delta}{kT^2} \right). \\
    & C = T \Big(\frac{\partial S}{\partial T} \Big)_{V, N} = \boxed{N \cdot \left( \dfrac{\Delta}{T} \right)^2 \cdot \dfrac{1}{ch^2 \left( \dfrac{\Delta}{T} \right)}}.
\end{align}

Также интересно посмотреть на предельные случаи ― энтропию при малой и большой температуре:

\begin{align}
    &T \to + \infty:\;\;\;S = k \ln 2^N=k\ln G, \\
    &T \to 0:\;\;\;S=\ln \left( 2 ch{\Delta}{kT} \right) - T \tanh \dfrac{\Delta}{kT} \to \dfrac{\Delta}{kT} - \dfrac{\Delta}{kT} = 0.
\end{align}

Первый предел согласуется с интуитивным представлением об $N$ частицах в одном из двух случайных расстояний, а второе не противоречит третьему началу термодинамики.
