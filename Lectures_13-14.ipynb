## Фазовые переходы I, II рода

**I рода**

Такие фазовые переходы сопровождаются скачком первых производных от термодинамических потенциалов (по интенсивным параметрам).

**Например:** $s\;=\;- \left( \dfrac{\partial g}{\partial T} \right)_{P}; \; v\;=\; \left( \dfrac{\partial g}{\partial P} \right)_{T}.$

Следовательно, сопровождаются скачком уд. энтропии и уд. объема, т.е. у нас есть скрытая теплота фазового перехода.

$$g\;=\;T(S_2\;-\;S_1).$$

**II рода**

Такие фазовые переходы сопровождаются скачком 2-ых производных от уд. термодинамических потенциалов.

Такие переходы не сопровождаются поглощением или выделением тепла (а также изменением удельного объема вещества).

**Например:**

Удельная теплоемкость:
$$(\dfrac{\partial^2 g}{\partial {T}^2})_{P} = - ( \dfrac{\partial S}{\partial T} )_{P} = - \dfrac{c_{P}}{T}.$$

Коэф. теплового расширения:
$$\dfrac{1}{v_0}\cdot \dfrac{\partial^2 g}{\partial P \partial T} \;=\; \dfrac{1}{v_0} ( \dfrac{\partial v}{\partial T} )_{P}\;=\; \alpha.$$

Коэф. сжатия:

$$-\dfrac{1}{v} \cdot (\dfrac{\partial^2 q}{\partial P^2})_{T} \;=\; - \dfrac{1}{v} \cdot ( \dfrac{\partial v}{\partial P} )_{T}\;=\; \gamma.$$

# Вероятности. Распределение Максвелла

## Простейшие определения из теории вероятностей

В макрофизике мы имеем дело с усредненными величинами (усреднение по микросостояниями). Такую процедуру можно реализовать из-за огромного число молекул (атомов). Такие процессы (и наблюдаемые закономерности) называются **вероятностными**. 

Под вероятностными мы будем подразумевать такие процессы, результаты которых являются **непредсказуемыми** (можно сказать только о вероятности того или иного исхода).
<br>
Пример с монеткой: бросаем (как распределяются результаты? 1/2).

На самом деле, за всем этим, по крайней мере в рамках классического рассмотрения, стоят динамические законы (вероятностный характер описания связан лишь с большим количеством частиц). Но мы будем применять \textbf{теорию вероятностей} к физике.

**Событие** \$-\$ являение, относительно которого имеет смысл говорить могут они происходить или нет.
<br>
**Случайное событие** \$-\$ если в результате испытания (опыта) событие может как произойти, так и нет. 
<br>
**Случайная величина** \$-\$ некоторая переменная, численными значениями которой являются исходы (результаты) некоторого случайного события.
\begin{align*}
    \xi(\omega),
\end{align*}
<br>
где $\xi$ \$-\$ случайная величина, $\omega$ \$-\$ случайное событие.

Например, $\xi$ \$-\$ это номер шара, который мы вытащили из мешка.

**Сумма событий: $A$ и $B$:** ($A+B$) (происходит либо $A$, либо $B$).

**Произведение событий $A$ и $B$:** ($A\cdot B$) (происходит как $A$, так и $B$).
<br>
$A_1, A_2, \ldots, A_n$ \$-\$ единственно возможные события, если при данном испытании одни из них обязательно должно произойти.
<br>
Два события называются **равновероятными**, если нет никаких оснований ожидать, что при испытании одно из них будет появляться чаще, чем другое.

**Вероятность** \$-\$ количественная мера ожидаемой возможности какого-то случайного события.

Рассмотрим $n$ единственно возможных, несовместимых и равновероятностных событий: $A_1, \ldots, A_n$. Тогда вероятность каждого из них:
\begin{align}
    P\;=\;\dfrac{1}{n}.
\end{align}
<br>
Переходим к более сложному случаю. Пусть у нас есть опять $n$ единственно возможных событий. Но теперь они не равновероятны! Предположим, что мы знаем как разложить $m$-ое событие на равновероятные (таких **благоприятных** $N_m$ штук): $A_{m_1}, \ldots, A_{m_{N_m}}$. Тогда:
\begin{align}
    P_{A_m}\;=\; \dfrac{N_m}{N_1+N_2+\ldots+N_n}.
\end{align}

Какова вероятность того, что в вагоне есть латыш?
Пусть $A_1$ \$-\$ русские (10), $A_2$ \$-\$ белорусы (10), $A_3$ \$-\$ латыши (2).
\begin{align*}
    P_{A_3}\;=\;\dfrac{2}{10+10+2}\;=\;\dfrac{1}{11}.    
\end{align*}

Здесь мы ещё до испытания знали, какое распределение у вероятностей.

**Теорема о сложении вероятностей:** вероятность суммы несовместимых событий равна сумме их вероятностей.

Пусть $B\;=\;A_1\;+\;A_2\;+\;\ldots\;+A_i$, тогда:
\begin{align}
    P(B)\;=\;P(A_1)\;+\;P(A_2)\;+\;\ldots\;+\;P(A_i).
\end{align}

Какова веряотность того, что выпадет четное число у игрального кубика?

\begin{align*}
    P(\text{четное})\;=\;\dfrac{1}{6}\;+\;\dfrac{1}{6}\;+\;\dfrac{1}{6}\;=\;\dfrac{1}{2}.
\end{align*}

**Условие нормировки:**
\begin{align}
    P_1\;+\;P_2\;+\;\ldots\;+\;P_n\;=\;1.
\end{align}

**Теорема умножения вероятностей:**
\begin{align}
    P(A\cdot B)\;=\;P(A)\cdot P(B|A).
\end{align}

Пусть у нас 10 черных и 8 белых шаров с четными и нечетными номерами.
$A$ \$-\$ достать черный; $B$ \$-\$ достать четный номер; $AB$ \$-\$ достать черный с четным номером.
$$P(AB)\;=\;P(A)P(B|A),$$
$$P(A)\;=\;\dfrac{10}{10+8}\;=\;\dfrac{5}{9},$$
$$P(B|A)\;=\;\dfrac{5}{5+5}\;=\;\dfrac{1}{2},$$
$$\implies P(AB)\;=\;\dfrac{5}{18}.$$
Если $A$ и $B$ независимы, т.е. произойдёт событие $A$ или никак не влияет на событие $B$, то $P(AB)\;=\;P(A)\;\cdot\;P(B).$
\begin{align}
    P\;=\;\lim\limits_{T\to\infty}\dfrac{+}{T}.
\end{align}

## Математическая статистика##
Рассмотрим случайную величину $a$. Пусть было произведено $N$ измерений:
$$n_1\;+\;n_2\;+\;n_3\;+\;\ldots\;+\;n_m\;=\;N,$$
где $n_1$ \$-\$ $a_1$ раз, $n_2$ \$-\$ $a_2$ раз, $\ldots$ Тогда:
\begin{align}
    \langle a\rangle\;=\;\dfrac{n_1a_1+n_2a_2+\ldots+n_ma_m}{N},
\end{align}
где $\langle a \rangle$ \$-\$ **среднее значение**.

Пусть $a_1,\ldots, a_m$\$-\$- единственно возможные значения и мы решили провести очень много опытов, доводя частоты до предела, тогда:

\begin{align}
    M(a)\;=\;P_1a_1\;+\ldots\;+P_{m} a_m.
\end{align}
**Ошибка** \$-\$ разность между измерененными значениями и истинным
\begin{align}
    x_i\;=\;a_i-a
\end{align}
**Среднеквадратичное отклонение:**
\begin{align}
    \Delta_{\text{кв}}\;=\;\sqrt{\langle x^2\rangle}\;=\;\sqrt{\dfrac{x_1^2\;+\;x_2^2\;+\ldots\;+\;x_N^2}{N}}.
\end{align}
**Дисперсия:**
\begin{align}
    \langle x^2\rangle\;=\;\dfrac{x_1^2\;+\;x_2^2\;+\ldots\;+\;x_N^2}{N}.
\end{align}
<br>
Рассмотрим ситуацию посложнее. Пусть мы бросаем кубик $n$ раз, а выпадает какая-то грань $n_i$ раз. тогда $\nu_1 = n_1/n$ \$-\$ частота выпадания первой грани. Будем увеличивать количество бросков. $\nu$ стремится к определённому пределу :<br>
\begin{align}
    P_1 = \lim \limits_{n\to \infty} \nu_1 = \lim \dfrac{n_1}{n}.
\end{align}
Вероятность \$-\$ предел частоты выпадания.
<br>
Другой пример: есть сосуд, пусть в нём летает огромное количество молекул. Вероятность найти молекулу в объёме $v$: $v/V$. <br>
Либо же можем измерить количество молекул в объёме $v$: $P = \lim\limits_{N\to\infty}N_1/N$. С другой стороны, предположим мы можем засечь время нахождения этой молекулы в сосуде $t$ (всё врея -- $T$). 
<br>
До этого момента мы работали с дискретными распределениями. А что, если величина принимает непрерывный спектр значений? Например, $a \in [x_1, \; x_2]$

В таком случае имеет смысл говорить о вероятности, что $a$ лежит в диапазоне от $y$ до $y + d y$ (поскольку вероятность того, что величина примет конкретное значение \$-\$ 0). Эта вероятность будет $d P$:
\begin{align}
    d P = \rho (a) d y.
\end{align}
$\rho(a)$ \$-\$ **плотность вероятности**.
<br>
**Условие нормировки**: 
\begin{align}
    \int\limits_{x_1}^{x_2} \rho(a) d a = 1.
\end{align}
Вероятность того, что случайная величина окажется в интервале $(y_1, y_2)$:
\begin{align}
    P(a\in(y_1, y_2)) = \int\limits_{y_1}^{y_2} \rho(a) d a.
\end{align}
$\langle a\rangle$ или $M(a)=$:
\begin{align}
    =\int\limits_{y_1}^{y_2}a \rho(a) d a.
\end{align}
**Распределением случайной величины** называют: 
\begin{align}
    F_a(x) = P(a \leqslant x) = \int \limits_{-\infty}^x \rho(a) d a.
\end{align}
<br>
В более общем случае это закон, описывающий область значений случайной величины и соответствующей вероятности.
## Распределение Максвелла
В статистическом равновесии все направления скоростей равновероятны. Однако что с абсолютными скоростями? Рассмотрим скоростное пространство:

Рассмотрим $N$ частиц. Мы хотим получить полную картину в скоростном пространстве. Динамически не решить, поскольку слишком много степеней свободы. Рассмотрим элементарный объём: кубик с объёмом $d v_x\cdot d v_y\cdot d v_z$ в окрестности $(v_x, v_y, v_z)$. Пусть число скоростных точек в нём $d N$, из этого объёма постоянно улетают и прилетают новые частицы, поскольку при абсолютно упругых соударениях частицы обмениваются скоростями.
<br>
Мы рассматриваем равновесное значение системы с огромным числом частиц. В таком случае можно работать со средними значениями чисел молекул в объёмах. Отклонения $\langle dN \rangle$ от постоянного значения очень малы. Следовательно, $\langle dN \rangle/N$ -- вероятность того, что частица будет иметь скорость в окрестности $[(v_x, v_x + d v_x), (v_y, v_y + d v_y), (v_z, v_z + d v_z)]$. Как мы понимаем, такая вероятность из-за непрерывного характера распределения должна быть:
<br>

\begin{align}
    \dfrac{\langle dN \rangle}{N} = f(\vec{v}) d v_x d v_y d v_z, 
\end{align}
где $f(\vec{v})$ \$-\$ функция распределения молекул по скоростям.

До этого момента обсуждалась, скорее, постановка задачи. Перейдём к рассуждениям, которые проводил Максвелл.\\
Какова вероятность, что молекула имеет компоненту скорости от $v_x$ до $v_x + d v_x$, а остальные компоненты \$-\$ любые? Такая вероятность составляет $\varphi(v_x)d v_x$. То же самое можно сказать и про другие направления: $\varphi(v_y) d v_y$, $\varphi(v_z) d v_z$. Максвелл выдвинул предположение, что эти вероятности независимы:
\begin{align}
    f(v_x,v_y,v_z) = \varphi(v_x) \varphi(v_y) \varphi(v_z).
\end{align}
Помимо этого, мы знаем что от выбора координат зависеть распределение не должно, и следовательно:
\begin{align}
    \varphi(v_x) = \varphi(-v_x) \implies \varphi = \varphi (v^2).
\end{align}
Значит, можно сказать, что 
\begin{align}
    & f(v_x^2 + v_y^2 + v_z^2) = \varphi(v_x^2) \, \varphi(v_y^2) \, \varphi(v_z^2).
\end{align}
Рассмотрим конкретную поверхность в нашем пространстве скоростей такую, \\что $v_x^2 + v_y^2 = \text{const}$ и $v_z^2 = \text{const}$:

\begin{align}
    & \varphi(v_x^2) \, \varphi(v_y^2) = \text{const}, \\
    & \ln \varphi(v_x^2) + \ln \varphi(v_y^2) = \text{const}. \\
    & \dfrac{d \varphi(v_x^2)}{\varphi(v_x^2)} + \dfrac{d \varphi(v_y^2)}{\varphi(v_y^2)} = 0, \\
    & \dfrac{\partial \varphi(v_x^2)}{\partial v_x^2} \cdot \dfrac{2 v_x d v_x}{\varphi(v_x^2)} = - \dfrac{\partial \varphi(v_y^2)}{\partial v_y^2} \cdot \dfrac{2 v_y d v_y}{\varphi(v_y^2)}. \label{eq:maxwell_distribution_1}
\end{align}
<br>
Посмотрим на $d v_x$ и $d v_y$:
<br>
\begin{align}
    & v_x^2 + v_y^2 = \text{const}, \\
    & 2 v_x d v_x + 2 v_y d v_y = 0, \\
    & 2 v_x d v_x = - 2 v_y d v_y. \label{eq:maxwell_distribution_2}
\end{align}
Подставим одно в другое.
$$\require{enclose}$$
\begin{align}
    & \dfrac{\partial \varphi(v_x^2)}{\partial v_x^2} \cdot \dfrac{\enclose{updiagonalstrike}{2 v_x d v_x}}{\varphi(v_x^2)} = \dfrac{\partial \varphi(v_y^2)}{\partial v_y^2} \cdot \dfrac{\enclose{updiagonalstrike}{2 v_x d v_x}}{\varphi(v_y^2)}, \\
    & \dfrac{\partial \varphi(v_x^2)}{\partial v_x^2} \cdot \dfrac{1}{\varphi(v_x^2)} = \dfrac{\partial \varphi(v_y^2)}{\partial v_y^2} \cdot \dfrac{1}{\varphi(v_y^2)}.
\end{align}
Правая часть не зависит от $v_x$, значит не зависит и левая, равная ей. При этом левая часть не зависит от $v_y$, значит не зависит и правая. Оба выражения не зависят ни от каких переменных, встречающихся в них, значит они постоянны!
\begin{align}
    & \dfrac{\partial \varphi(v_x^2)}{\partial v_x^2} \cdot \dfrac{1}{\varphi(v_x^2)} = \dfrac{\partial \varphi(v_y^2)}{\partial v_y^2} \cdot \dfrac{1}{\varphi(v_y^2)} = \text{const}, \\
    & \dfrac{d \varphi}{\varphi} = - \alpha d v_x^2. \\
    & \int \dfrac{d \varphi}{\varphi} = - \alpha \int d v_x^2, \\
    & \ln \dfrac{\varphi}{A} = -\alpha v_x^2, \\
    & \boxed{\varphi(v_x^2) = A \; \exp \left( - \alpha v_x^2 \right)}, \\
    & \boxed{f(v^2) = A^3 \exp \left( - \alpha v^2 \right)}.
\end{align}
Тут $\alpha$ и $A$ ― некие константы. 

Наложим условие нормировки:
\begin{align}
    & 1 = \int \limits^{+ \infty}_{- \infty} f(v^2) \, d v_x \, d v_y \, d v_z = \int \limits^{\infty}_0 f(v^2) \, v^2 \cdot 4 \pi \, d v \\
    & = \int \limits^{\infty}_0 4 \pi \, v^2 \, A^3 \, \exp \left( - \alpha v^2 \right) d v = - \dfrac{\partial}{\partial \alpha} \left( \int \limits^{\infty}_0 4 \pi \, A^3 \, \exp \left( - \alpha v^2 \right) d v \right), \\
    & \int \limits^{\infty}_0 \exp \left( - \beta x^2 \right) d x = \dfrac{1}{2} \sqrt{\dfrac{\pi}{\beta}}, \\
    & 1 = - \dfrac{\partial}{\partial \alpha} \left( 4 \pi A^3 \, \dfrac{1}{2} \sqrt{\dfrac{\pi}{\alpha}} \right) = \pi A^3 \, \dfrac{\sqrt{\pi}}{\alpha^{3/2}}, \\
    & A = \sqrt{\dfrac{\alpha}{\pi}}.
\end{align}
Из «прошлой жизни» помним, что $\left\langle \dfrac{m v^2}{2} \right\rangle = \dfrac{3}{2}k T$, то есть $\langle v^2 \rangle = \dfrac{3 k T}{m}$.
\begin{align}
    & \int \limits^{\infty}_0 v^2 \cdot f(v^2) \, v^2 \, 4 \pi d v = \dfrac{3 k T}{m}, \\
    & \dfrac{3}{2} \dfrac{\pi \sqrt{\pi} A^3}{\alpha^{5/2}} = \dfrac{3 kT}{m}, \\
