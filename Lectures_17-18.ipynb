Рассмотрим идеальный газ в состоянии равновесия в поле силы тяжести.

Равновесие:
* Одинаковая температура во всей толще газа (которую анализируем);
* Механическое равновесие.

\begin{align}
    &(P_1\;-\; P_2)S \;=\; nmS d z g;\\
    &P_1\;-\; P_2 \;=\; -\dfrac{d P}{d z}d z \;=\; P(z)\;-\; P(z\;+\; d z);\\
    &P\;=\; nkT \; \Rightarrow \; d P \;=\; d n kT; \\
    &- \dfrac{d n kT}{d z}\;=\;nmg; \\
    &kT \dfrac{d n}{n}\;=\;-mg d z; \\
    &kT \dfrac{d n}{n}\;=\; m \vec{g} d \vec{r}.
\end{align}

Необходима консервативность, т.к. $\oint\dfrac{d n}{n} = 0$, но $\oint\vec{g} d \vec{r}= 0$ только в случае консервативных полей.

$$kT \dfrac{d n}{n}\;=\; - d U;$$
$$kT \ln{\dfrac{n(\vec{r})}{n_0}}\;=\; -(U(r)\;-\;U_0);$$
$$\boxed{n(\vec{r})\;=\;n_0 \exp{-\dfrac{U(r)}{kT}}}.$$

Эквивалентно: $P\;=\;P_0 \exp{-\dfrac{U}{kT}}$.

В случае действия силы тяжести:
\begin{align}
    &u\;=\;mgz\; \Rightarrow \; n\;=\;n_0 \exp{-\dfrac{mgz}{kT}}\;=\;n_0 \exp{-\dfrac{\mu gz}{RT}}.
\end{align}

Вывод хорош только для плотной смеси.
* с точки зрения гидростатики: плотность должна быть больше;
* но с точки зрения идеального газа (один газ не знает про другой по закону Дальтона).

## Равномерное распределение кинетической энергии по степеням свободы
Вернемся к распределению кинетической энергии по степеням свободы. Ранее получали следующий вывод через поршень: $\left\langle  E_k\right\rangle=1/2kT$. Очевидно, что наличие поршня не может сказаться на распределении энергии между газами, находящимся по разные стороны. Если убрать перегородку то обмен энергиями будет осуществляться за счет столкновения напрямую. Давайте теперь учтем тот факт, что поршень сам состоит из молекул. 

Скорость поршня равна скорости центра масс всех молекул поршня: 
\begin{align}
    &  u=\frac{1}{M}\sum m_i u_i \Rightarrow\\
    &  \Rightarrow \frac{1}{2}Mu^2=\frac{1}{2}\frac{1}{M}\sum_{ij}m_iu_im_ju_j \Rightarrow \\
    &  \Rightarrow \frac{1}{2}M\left\langle u^2\right\rangle=\frac{1}{2M}\sum_{ij}m_im_j\left\langle u_iu_j\right\rangle =\\
    & \left\langle u_i u_j\right\rangle=\left\langle u_i\right\rangle\left\langle u_j\right\rangle, \text{ если } i\neq j \left\langle u_i\right\rangle=0, \left\langle u_j\right\rangle=0. \\
    & = \frac{1}{2}=\sum_{ij} \delta_{ij}m_im_j \left\langle u_i^2\right\rangle= \\
    & =\frac{1}{2M}\sum_{i}m_i^2\left\langle u_i^2\right\rangle=\frac{1}{2}kT .
\end{align}

Для простоты будем рассматривать однородный поршень: 
\begin{align}
     &\Rightarrow \frac{M}{m} = N\\
     &\Rightarrow  \frac{1}{2N}m_i\left\langle u_i^2\right\rangle N=\frac{1}{2}kT\\
     &\Rightarrow \frac{m_i\left\langle u_i^2\right\rangle}{2}=\frac{1}{2}kT.
\end{align}
<br>
Мы рассматривали одну поступательную степень свободы. Таким образом мы можем говорить не только про газ, но и вообще про любое тело: что на молекулу, что на целое макротело. 
<br>
Получить аналогичное соотношение можно и на вращательную степень свободы: 
\begin{align}
    & \omega \nonumber= \frac{\sum m_i r_i u_i}{\sum m_i r_i^2}\\
    &\Rightarrow \left\langle \nonumber\omega^2\right\rangle=\frac{1}{I^2}\sum_{ij} m_i m_j \times \left\langle r_i u_i r_j u_j\right\rangle=\nonumber\\
    &=\frac{1}{I^2} \sum m_i^2 r_i^2 \left\langle u_i^2\right\rangle. \\
    & \frac{I\left\langle \omega^2\right\rangle}{2}=\frac{1}{2I}\sum m_i^2 r_i^2 \left\langle u_i^2\right\rangle = \nonumber\\
    & = \frac{2}{2I}\sum_i m_i r_i^2 \frac{\left\langle m_i u_i^2\right\rangle}{2}=\frac{1}{2}kT.
\end{align}

## Броуновское движение

Броун рассматривал частицы цветочной пыльцы в воде. Оказалось, что всему причина непревное хаотическое дрожание молекул. Проанализируем это непрерывное движение. В одной и той же жидкости интенсивность движения возрастает с ростом $T$ и уменьшением размеров. Движение обусловлено толчками со стороны молекул жидкости. А так как частица мала, то толчки не удастся скомпенсировать. Получим, что для каждой частицы (взвесим):
\begin{align}
    \left\langle \dfrac{mv^2}{2}\right\rangle\;=\;\dfrac{3}{2}kT.
\end{align}
Квадрат скорости в эксперименте измерить трудно (хотим вообще-то формулу проверить).

Рассмотрим движение вдоль оси, где $F_x$ \$-\$ сила толчков, $6\pi\eta r\dot{x}$ \$-\$ закон Стокса:
\begin{align}
    &m\ddot{x}\;=\;F_x\;-\;6\pi\eta r\dot{x},\\
    &xm\ddot{x}\;=\;x\cdot F_x\;-\;6\pi\eta rx\dot{x},\\
    &\left[x\dfrac{d x}{d t}\;=\;\dfrac{1}{2}\dfrac{d}{d t}x^2\right]\\
    &xm\dfrac{d^2x}{d t^2}\;=\;x\cdot F_x\;-\;6\pi\eta r\dfrac{1}{2}\dfrac{d}{d t}x^2,\\
    &\ddot{x}x\;=\;\dfrac{1}{2}\dfrac{d^2}{d t^2}x^2\;-\;\left(\dfrac{d x}{d t}\right)^2,\\
    &\left(\dfrac{1}{2}\dfrac{d^2}{d t^2}x^2\;=\;\dfrac{1}{2}\dfrac{d}{d t}2x\dfrac{d x}{d t}\;=\;\dfrac{1}{2}2\left(\dfrac{d x}{d t}\right)^2\;+\;\dfrac{1}{2}2x\dfrac{d^2x}{d t^2}\right),\\
    &\implies m\cdot\left(\dfrac{1}{2}\dfrac{d^2}{d t^2}\left(x^2\right)\;-\;\left(\dfrac{d x}{d t}\right)^2\right)\;=\;xF_x\;-\;3\pi\eta r\dfrac{d}{d t}x^2.
\end{align}
Усредним: $d/d t$ и $\left\langle\;\right\rangle$ коммутируют.
\begin{align}
    &m\cdot\left(\dfrac{1}{2}\dfrac{d^2}{d t^2}\left\langle  x^2\right\rangle\;-\;\left\langle \left(\dfrac{d x}{d t}\right)^2\right\rangle\right)\;=\;\underset{\left\langle  x \right\rangle = 0; \left\langle  F_x\right\rangle = 0}{\left\langle  x\cdot F_x\right\rangle}\;-\;3\pi\eta r\dfrac{d}{d t}\left\langle  x^2\right\rangle.\\
    &\sqsupset z\;=\;\dfrac{d}{d t}\left\langle  x^2\right\rangle\implies\\
    &\implies \dfrac{m}{2}\dfrac{d z}{d t}\;-\;kT\;=\;-3\pi\eta r\cdot z,\\
    &\dfrac{m}{2}\dfrac{d z}{d t}\;+\;3\pi\eta rz\;=\;kT.\\
    &\text{Решение: }z\;=\;c\cdot \exp{-\dfrac{6\pi\eta rz}{m}t}\;+\;\dfrac{kT}{3\pi\eta r},\\
    &\text{Или: $z:$} \dfrac{m}{2}\dfrac{d z}{kT\;-\;3\pi\eta rz}\;=\;d t,\\
    &-\dfrac{1}{3\pi\eta r}\dfrac{m}{2}\ln{\big|kT\;-\;3\pi\eta rz\big|}\big |^z_{z(0)}\;=\;t,\\
    &z(0)=0 \implies z\;=\;\dfrac{kT}{3\pi\eta r}\left(1\;-\;c\cdot\exp{-\dfrac{6\pi\eta r}{m}t}\right).
\end{align}
Предельный случай (за большой промежуток времени):
\begin{align}
    \boxed{\dfrac{d}{d t}\left\langle  x^2\right\rangle\;=\;\dfrac{kT}{3\eta\pi r}},
\end{align}
$r\;\sim\;1$\;мкм,\;$\eta\;\approx\;9\cdot10^{-4}\;$Па$\cdot$с,\;$m\;=\;10^{-12}\;$г.
\begin{align}
    &\implies\left\langle  x^2\right\rangle\;=\;\dfrac{kT}{3\eta\pi r}\tau,\qquad\tau\;\gg\;\dfrac{m}{6\pi\eta r},\\
    &\implies\sqrt{\overline{\Delta^2}}\;=\;\left(\dfrac{kT}{3\eta\pi r}\right)^{1/2}\tau^{1/2}.
\end{align}
Значит, что средний квадрат отклонения пропорционален $\sqrt{\tau}$.
## Классическая теория теплоёмкости идеальных газов
Применяем закон классической механики к молекулам и атомам.
Уже получали, что $C_V\;=\;R/(\gamma-1), C_P\;=\;\gamma R/(\gamma -1)$. Т. е. надо определить $\gamma$.

Внутренняя энергия определяется: кинетическая энергия поступательного, вращательного и внутреннего движений. Для идеального газа потенциальной энергией пренебрегаем. Одноатомный газ (1 моль):
$$U\;=\;{3}/{2}NkT\;=\;{3}/{2}N_AkT\;=\;{3}/{2}RT$$
$$C_V\;=\;\dfrac{d U}{d T}\;=\;{3}/{2}R$$
$$C_P\;=\;{5}/{2}R$$
$$\implies\;\gamma\;=\;{5}/{3}$$

Для двухатомного газа 6 степеней - 1 = 5 (1 связь).
Из аналогичных рассуждений получим, что $\gamma = {7}/{5}$.

Многоатомная молекула: $\gamma\;=\;{4}/{3}.$

## Энтропия, вероятность, формула Больцмана

С точки зрения термодинамики, все процессы в замкнутой системе протекают с возрастанием энтропии. Равновесие характеризуется максимумом энтропии. При максимальном её значении никакие процессы не происходят. Однако, с точки зрения микроскопики, равновесие динамическое, а не статическое. В качестве примера рассмотрим сосуд из двух одинаковых по объёму частей. В среднем молекул в левой и в правой части одинаковое количество (кто-то вылетает слева направо, кто-то вылетает справа налево), равновесие динамическое. \textbf{Флуктуациями} называют отклонения от средних значений. Они, в свою очередь, сильно падают с ростом числа частиц.

Вероятность того, что все $N$ штук частиц находятся в одной половине составляет $(1/2)^N$. Если же хотим узнать о нахождении частиц в каком-то объёме, отличном от половину сосуда, то вероятность составляет $(V/V_0)^N$. Если систему вывести из положения равновесия, то она будет стремиться вернуться в состояния равновесия, то есть будет идти в состояние с большей вероятностью. На фоне этих рассуждений задумываемся, а нет ли связи энтропии с вероятностью? $S=S(P)$?

Рассмотрим две макроскопические подсистемы. Пусть вероятности первого и второго состояния это $P_1$ и $P_2$ соответственно. Энтропии этих систем: $S_1=f(P_1)$, $S_2=f(P_2)$. Рассмотрим эти системы как единую. Вероятность такого состояния $P_{12}=P_1\cdot P_2$ (системы независимые).
\begin{align}
    &S(P_{12})= S(P_1\cdot P_2) = f(P_1\cdot P_2) = f(P_{12}).
\end{align}
Хотим, чтобы новое определение энтропии состыковывалось со старым, то есть, чтобы новая энтропия также была аддитивной величиной, и тогда
\begin{align}
    f(P_1\cdot P_2) = f(P_1) + f(P_2).
\end{align}
Если это верно для всех соотношений между $P_1$ и $P_2$. Так, в частности,  это верно и когда $P_1\cdot P_2=const$. 
\begin{align}
    &d P_1P_2 = p_1 d P_2 + P_2 d P_1 = 0 \implies \dfrac{d P_1}{P_1} = - \dfrac{d P_2}{P_2}, \label{iht} \\
    &d f(P_1P_2) = d f(P_1) + d f(P_2) = 0 \implies d f(P_1) = -d f(P_2)\label{nggrs}.
\end{align}
Объединяя  получаем
\begin{align}
    P_1 \dfrac{d f(P_1)}{d P_1} = P_2 \dfrac{f(P_2)}{d P_2}.\label{rlhtngrs}
\end{align}
Варьируя значения $P_1$ и $P_2$ понимаем, что левые и правые части сохраняются. Таким образом, можем объявить их константой $k$ (ещё не знаем, что это постоянная Больцмана).
\begin{align}
    &P\dfrac{d f(P)}{d P} = k, \nonumber \\
    &f(P) = k\ln P + const.
\end{align}
Используя соотношение получаем:
\begin{align}
    &f(P_1P_2) = k\ln{P_1P_2} + const = k \ln P_1 + k\ln P_2 + 2\cdot const,\nonumber \\
    &const = 2\cdot const \implies const = 0.
\end{align}
Таким образом, окончательно:
\begin{align}
    \boxed{S=k\ln P}\label{eq:formula_b}.
\end{align}
Выражение называется формулой Больцмана.Вычислим энтропию, зная вероятность $P$:
\begin{align}
    \Delta S = k \ln{\left(\dfrac{V_2}{V_0}\right)^N} - k\ln{\left(\dfrac{V_1}{V_0}\right)^N} = kN\ln{\dfrac{V_2}{V_1}}.\label{s_1}
\end{align}
Вычислим энтропию идеального газа из термодинамики:
\begin{align}
    \Delta S = \nu R \ln{\dfrac{V_2}{V_1}} \quad \text{(температуры $T_1$ и $T_2$ одинаковы)}.\label{s_2}
\end{align}
Сравнивая получаем:
\begin{align}
    kN = \nu R = \dfrac{N}{N_\text{а}} R \implies \boxed{k = \dfrac{R}{N_a}}.
\end{align}
Таким образом мы получили константу Больцмана.

Здесь, в отличие от термодинамики, нет проблем с определением термодинамики в неравновесном состоянии. Также есть важный момент: возрастание в среднем (за большой промежуток времени): за состоянием идёт более вероятное состояние в подавляющем большинстве случаев.

Замечание: на практике вероятности определены относительно (3 раза в час, 2 раза в час). Все варианты не найти. То есть вероятность определена с точностью до умножения на константу.
\begin{align}
    \Delta S = k \ln\dfrac{P_2\alpha}{P_1\alpha} = \ln\dfrac{P_2}{P_1},
\end{align}
что позволяет выбрать удобную нормировку. Пусть вероятность выражается удельными числами, которые показывают число реализаций данного макросостояния.
\begin{align}
    P= \alpha \cdot G,
\end{align}
где величина $G$ называется **статистическим весом** или **статвесом**.

Рассмотрим пример: разобъём сосуд на конечные кусочки объёма: $V_1, V_2 \dots V_m$. Посчитаем количество частиц в каждом из них (этот набор чисел и будет макросостоянием): $N_1$ в $V_1$, $N_2$ в $V_2$, ..., $N_m$ в $V_m$. Микросостояниями будет нахождение конкретных молекул в конкретных объёмах.

Обозначим вероятность нахождения частицы в объёме $V_j$ как $p_j$. Вероятность нахождения молекул в конкретном микросостоянии тогда $\displaystyle P_{\mu} = p_1^{N_1} \, p_2^{N_2} \! ... \, p_m^{N_m}$. Посчитаем, сколько микросостояний соответствуют одному макросостоянию. Верхняя оценка ― $N!$ ― соответствует всем возможным перестановкам частиц. Мы, конечно, обсчитались: чтобы учесть, что перестановки частиц внутри одного и того же объёма не влияют на макросостояния, осталось поделить все возможные перестановки на перестановки внутри каждого из объёмов, $N_1!$, $N_2!$, ..., $N_m!$. Тогда вероятность макросостояния
\begin{align}
    & P = \dfrac{N!}{N_1! \, N_2! \, ... \, N_m!} P_{\mu} = \dfrac{N!}{N_1! \, N_2! \, ... \, N_m!} \; p_1^{N_1} \, p_2^{N_2} \! ... \, p_m^{N_m}.
\end{align}
Чтобы упростить выражение, перейдём к равным объёмам, а следовательно и вероятностям.
\begin{align}
    & p^{N_1} \, p^{N_2} \! ... \, p^{N_m} = p^{N_1 + ... + N_m} = p^N, \\
    & P = \underbrace{\dfrac{N!}{N_1! \, N_2! \, ... \, N_m!}}_G \; p^N. \\
    & \boxed{S = k \, \ln G}\label{eq:Plank},
\end{align}
где уравнение записано в формулировке Планка.

Только что мы говорили про расположение частиц в пространстве координат. Не забудем, что кроме координат у частицы есть и импульс, для чего перейдём в шестимерное фазовое пространство с осями $x$, $y$, $z$, $p_x$, $p_y$, $p_z$. Объёмы в фазовом пространстве однозначно соответствуют каким-то диапазонам энергии (забегая вперёд, в изолированной системе энергия зависит лишь от обобщённых координат и обобщённых импульсов). Дискретизируем пространство: разделим его на равные ячейки, которые пронумеруем натуральными числами. Энергию $j$-той ячейки обозначим $\varepsilon_j$. Повторяя предыдущие размышления, статвес
\begin{align}
    G = \dfrac{N!}{N_1! \, ... \, N_m!}.
\end{align}

Найдём состояние с наибольшим статвесом.
\begin{align}
    & N_1 + ... + N_m = N = \text{const}. \\
    & N_1 \varepsilon_1 + ... + N_m \varepsilon_m = E = \text{const}.
\end{align}
Считая число частиц большим, применим формулу Стирлинга:
\begin{align}
    &N! = \sqrt{2 \pi N} \left( \dfrac{N}{e} \right)^N,\label{eq:stirling}\\
    &G = \dfrac{\sqrt{2 \pi N} \left( \dfrac{N}{e} \right)^N}{\sqrt{2 \pi N_1} ... \sqrt{2 \pi N_m} \cdot \left( \dfrac{N_1}{e} \right)^{N_1} \!\! ... \left( \dfrac{N_m}{e} \right)^{N_m}} \nonumber\\
    &= \dfrac{\sqrt{2 \pi N}}{\Bigl(\sqrt{2 \pi}\Bigr)^m} \; \dfrac{N^N}{\displaystyle \prod_j N_j^{N_j + 1/2}} = \dfrac{C(N)}{\displaystyle \prod_j N_j^{N_j + 1/2}}.\\
    &S = - k \ln \prod_j N_j^{N_j + 1/2} + \text{const}(N) \nonumber\\
    &= - k \sum_j N_j \ln N_j + \text{const}(N).\\
    &\sum d N_i\;=\;0,\;\sum\varepsilon_i d \nonumber N_i\;=\;0,\;\sum\ln N_id N_i\;=\;0.\\
    &d N_i\ln N_i\;=\;\ln N_i d N_i\;+\;\dfrac{N_i}{N_i}d \nonumber N_i \approx \ln N_id N_i \implies\\
    &\implies \nonumber \sum\left(\ln N_i\;+\;\beta\;+\;\alpha\varepsilon_i\right)d N_i\;=\;0\\
    &\implies \ln N_i\;=\nonumber\;-\beta-\alpha\varepsilon_i\\
    &\left\langle N_{i}\right\rangle\;=\;\exp{-\alpha \varepsilon_i} \exp{- \beta}.
\end{align}
