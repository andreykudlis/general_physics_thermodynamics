## Термодинамическая система

**Термодинамическая система** \$-\$ в общем смысле, тело, которое обменивается с другими телами энергией и веществом. Однако, если мы имеем дело с маленьким количеством частиц (единицы, десятки и т.д.), то мы находимся в термодинамическом пределе, то есть можем оперировать не характеристиками отдельных частичек, а макропараметрами (величинами, связанной со всей системой как целокупностью):
$$T,\;P,\;V,\;m,\; \mu \;\text{и т.д.}$$

**Термоста** \$-\$ очень большая т.д. система, настолько большая, что при контакте с другими телами температура термостата почти не меняется.

Термодинамические системы делятся на три типа:

* открытые (обмениваются с другими системами и энергией, и веществом);
* закрытые (обмениваются энергией, но не веществом);
* изолированные (не обмениваются ни энергией, ни веществом).

Нормальную теорию взаимодействия атомов и молекул можно получить только в рамках квантовой механики, в рамках классической механики же мы опираемся на эмпирические приближения ― например, между молекулами взаимодействие выражаем в виде сил Ван-дер-Ваальса (1873 г.). ## Теплота с точки зрения МКТ

Теплота связана с хаотическим движением молекул (структурных единиц системы).

**Nota bene:** Термодинамика является феноменологической наукой, она базируется на наблюдаемых фактах и занимается тепловой формой движения материи.

**Что же это такое?**

Тепловая форма движения материи \$-\$ хаотическое движение атомов молекул.

Все дело в количестве структурных частичек в нашей системе. Из-за этого получаем равновероятное направление движения молекул.

**Nota bene:** Термодинамика изучает только равновесные состояния и медленные процессы (бесконечно медленные). Есть ещё кинетика \$-\$ подраздел МКТ, который изучает процессы с конечными скоростями.

## Общее начало термодинамики

Системы в разных состояниях, которые поместили в изолированную оболочку, приходят в термодинамическое равновесие.

**Nota bene:** Аналогия: жидкость - пар.

**Nota bene:** Равновесие может быть и динамическим, допускающее сложные колебания микропараметров, если макропараметры при этом постоянны во времени.

Термодинамическое равновесие подразумевает механическое и химическое равновесие между всеми частями системы.

Два тела меют одинаковую теплоту, если мы приводим в контакт (защита от механического и химического воздействия), и они остаются в состоянии равновесия.

Итак, температура: макропараметр $T$.

Нормальная точка плавления (1 атмосфера).
Тройная точка (сосуществование трёх фаз).

Что там с количественной мерою? Из термоскопа надо сделать термометр (проградуировать).

$T=f(a)$, a \$-\$ термометрическая величина.

Нужны реперные точки:

1954 г. \$-\$ нормальная точка кипения, нормальная точка плавления (воды).

\begin{align}
    &\Rightarrow \; \sqsupset f(a_k) - f(a_{\text{пл}}) = 100
    \\
    &\sphericalangle \; \; T = \alpha a
    \\
    &\Rightarrow \boxed{\alpha = \dfrac{T_k - T_{\text{пл}}}{a_k - a_{\text{пл}}}} \; .
\end{align}

Оказалось, что не очень удобно воспроизводить: решено строить по тройной точке воды: $T_{\text{тр}} = \alpha a_{\text{тр}}\;\;\overset{def}{\Rightarrow}\;\;(273.16)/{a_{\text{тр}}} = \alpha $.


(Идеально) Газовая шкала температур.

На основе закона: Бойля-Мариотта

Если мы имеем дело с разреженным газом, то $P, V, T$ (для фиксированного количества вещества) должно удовлетворять:
\begin{align}
    P\cdot V = C \cdot T.
\end{align}

(Слабая зависимость от природы газа) 

При $T = 0 \Rightarrow F=0$, то есть абсолютный ноль. Однако, это только догадка.

В теле прекращается тепловое движение, остаётся только тепловое движение частиц, котрое связано с нулевыми колебаниями.

Высокие и низкие температуры не могут быть обработаны при помощи газового термометра.

Нужна термическая шкала температур.

**Идеальный газ** \$-\$ газ, который строго подчиняется закону Бойля-Мариотта.
\begin{align}
    \boxed{PV = CT} \; .
\end{align}

Первый пример уравнения состояния

$PV = CT = \varphi (P, V, T) = 0.$


Очевидно, что состояние газа зависит от его массы и химической природы. Введём понятие количества вещества:
\begin{align*}
    \nu\;=\;\dfrac{m}{\mu},\;\;\left[\nu\right]\;=\;\text{моль},
\end{align*}
1 моль вещества содержит в себе столько молекул (атомов), сколько содержит в себе 12 г изотопа углерода $C^{12}$.

В идеальных газах в одинаковых объёмах содержится одинаковое количество молекул при одинаковых значениях температуры $T$ и давления $P$ (из вида уравнения).

Удобно использовать для измерения 1 моля в граммах.
Тогда запишем уравнение Менделеева-Клапейрона:
\begin{align}
    \boxed{PV\;=\;\nu RT}\;,
\end{align}
где $R$ \$-\$ универсальная газовая постоянная. $R$=8.31;Дж/(К$\cdot$моль).

Рассмотрим смесь идеальных газов.
\begin{align*}
    PV\;=\;\nu RT.
\end{align*}
т. к. идеальный газ\$-\$ сильно разреженный газ, то:
\begin{align}
    &P_1V\;=\;\nu_1RT\nonumber\\
    &+\nonumber\\
    &P_2V\;=\;\nu_2RT\nonumber\\
    &+\nonumber\\
    &\ldots\nonumber\\
    &\left(P_1\;+\;P_2\;+\;\ldots\;+\;P_n\right)V\;=\;\left(\nu_1\;+\;\nu_2\;+\;\ldots\;+\;\nu_n\right)RT,
\end{align}
$V$ не меняется, так как считаем, что молекулы не занимают объёма. Таким образом, мы записали \textbf{закон Дальтона}.

## Уравнения состояния
Запишем уравнения состояния.
Пусть есть такое уравнение состояния, что:
\begin{align}
    \varphi\left(P,V,T\right)\;=\;0,\;\;V\;=\;V\left(P, T\right)\nonumber\\
    \mathrm{d}V\;=\;\left(\dfrac{\partial V}{\partial T}\right)_P\mathrm{d}T\;+\;\left(\dfrac{\partial V}{\partial P}\right)_T\mathrm{d}P. \label{eq:dif_main}
\end{align}

Это выражение справедливо всегда, так как число частиц фиксированное. Рассмотрим, например, процесс, при котором $\mathrm{d}V\;=\;0$:
\begin{align}
    &\left(\dfrac{\partial V}{\partial T}\right)_P\cdot\;\left(\dfrac{\mathrm{d}T}{\mathrm{d}P}\right)_V+\;\left(\dfrac{\partial V}{\partial P}\right)_T\;=\;0\;\implies\\
    &\implies -1\;=\;\left(\dfrac{\partial V}{\partial T}\right)_P\left(\dfrac{\partial P}{\partial V}\right)_T\left(\dfrac{\partial T}{\partial P}\right)_V.
\end{align}
## Коэффициенты изменения макропараметров
### Коэффициент теплового расширения:
\begin{align}
    \alpha\;=\;\dfrac{1}{V_0}\left(\dfrac{\partial V}{\partial T}\right)_P.
\end{align}
### Термический коэффициент давления:
\begin{align}
    \beta\;=\;\dfrac{1}{P_0}\left(\dfrac{\partial P}{\partial T}\right)_V.
\end{align}
### Модуль всестороннего сжатия:
\begin{align}
    K\;=\;-V\left(\dfrac{\partial P}{\partial V}\right)_T.
\end{align}
Нетрудно заметить, что:
\begin{align}
    &\dfrac{\alpha K}{\beta}\;=\;\dfrac{P_0}{V_0}\dfrac{\left(\dfrac{\partial V}{\partial T}\right)_P\left(-V\left(\dfrac{\partial P}{\partial V}\right)_T\right)}{\left(\dfrac{\partial P}{\partial T}\right)_V}\\
    &-\dfrac{\alpha K}{\beta}\dfrac{V_0}{P_0V}\;=\;\enclose{updiagonalstrike}{\dfrac{\partial V,P}{\partial T,P}}\;\enclose{updiagonalstrike}{\dfrac{\partial P,T}{\partial V,T}}\;\enclose{updiagonalstrike}{\dfrac{\partial T,V}{\partial P,V}} 
    &\implies \boxed{\dfrac{\alpha K}{\beta} \dfrac{V_0}{P_0V}\;=\;1}.
\end{align}

## Макропараметры
Такие величины, как $V, T, P, \rho$, определяют макросостояние. Например, в термодинамическом равновесии состояние газа может быть полностью описано массой, давлением, температурой и химической природой.

В будущем, когда мы обоснуем термодинамические законы не только эксприментально, но и при помощи методов статистической физики, макропараметры внезапно окажутся средними значениями каких-то величин ― однако долгое время учёные изучали термодинамику без этих соображений, опираясь исключительно на эмпирические наблюдения.

\begin{align}
    &A_{12} (\text{работа газа}) = (U_{i,1} + U_{ii,1}) - (U_{i,2} + U_{ii, 2}), \nonumber \\
    &A_{12} = -\underbrace{(U_{i,2} - U_{i,1})}_{\Delta U_1} + \underbrace{(U_{ii,1} - U_{ii,2})}_Q.\nonumber \\
    &\implies \Delta U_1 + A_{12} = Q - \text{убыль внутренней энергии}.\nonumber \\
    & \boxed{\delta Q = \mathrm{d}U + \delta A}.
\end{align}

Пояснение: $\mathrm{d}U$ - дифференциал, поскольку внутренняя энергия есть функция состояния, $\delta A$ - лишь элементарное приращение (обозначаемое дельтой), поскольку работа определяется тем, как её совершают, а следовательно и элементарная теплота ($\delta Q$) «зависит от траектории процесса».
А теперь частные случаи:
1) Постоянный объем, $\mathrm{d}V = 0$:
\begin{align}
    &\int\delta Q = \int \mathrm{d}U + P\mathrm{d}V = \int \mathrm{d}U = \Delta U.
\end{align}
2) Постоянное давление, тогда можем внести $P$ под дифференциал:
\begin{align}
    &\int \delta Q = \int d U + P\mathrm{d}V = \int \mathrm{d}(U + PV) = \Delta(U + PV).
\end{align}
Величину $U + PV$ назовем **энтальпией**.
\begin{align}
    &I = U + PV.
\end{align}
Откуда делаем вывод, что при изобарном процессе изменение теплоты равно изменению энтальпии.

# Теплоёмкость

Для описания конкретного термодинамического процесса удобно ввести теплоёмкость ― величину, которая связывает количество теплоты с изменением температуры.
\begin{align}
    &C \overset{\text{def}}{=} \dfrac{\delta Q}{\mathrm{d} T} \text{ ― теплоёмкость}, 
    &c_m \overset{\text{def}}{=} \dfrac{C}{m} \text{ ― удельная теплоёмкость}, 
    &c_{\nu} \overset{\text{def}}{=} \dfrac{C}{\nu} \text{ ― молярная теплоёмкость}.
\end{align}
Стоит заметить, что теплоёмкость не является функцией состояния!

Рассмотрим теплоёмкость равновесного процесса:
\begin{align}
    &\mathrm{d}U = \left( \dfrac{\partial U}{\partial T} \right)_V \mathrm{d}T + \left( \dfrac{\partial U}{\partial V} \right)_T \mathrm{d}V. \\
    &C = \dfrac{\mathrm{d}U + p\mathrm{d}V}{\mathrm{d}T} = \left( \dfrac{\partial U}{\partial T} \right)_V + \left( \dfrac{\partial U}{\partial V} \right)_T \dfrac{\mathrm{d}V}{\mathrm{d}T} + P \dfrac{\mathrm{d}V}{\mathrm{d}T}, \\
    &C = \left( \dfrac{\partial U}{\partial T} \right)_V + \left[\left( \dfrac{\partial U}{\partial V} \right)_T + P \right] \dfrac{\mathrm{d}V}{\mathrm{d}T}.
\end{align}

Применяя это соотношение к изохорному и изобарическому процессам, можем вывести их теплоёмкости.

При $V = \text{const}$:
\begin{align}
    &C_V = \left( \dfrac{\partial U}{\partial T} \right)_V. 
\end{align}

При $p = \text{const}$:
\begin{align}
    &C_p = C_V + \left[\left( \dfrac{\partial U}{\partial V} \right)_T + P \right] \left( \dfrac{\partial V}{\partial T} \right)_p.
\end{align}

Представим себе газ, который изохорно нагревается от абсолютного нуля температур до некой температуры $T$. Скажем, что в абсолютном нуле внутренняя энергия тоже нулевая.
\begin{align}
    &V = \text{const}, \quad \dfrac{\mathrm{d}U}{\mathrm{d}T} = \left( \dfrac{\partial U}{\partial T} \right)_V = C_V. 
    &\int \limits^U_0 \mathrm{d}\tilde{U} = \int \limits^T_0 C_V \mathrm{d} \tilde{T}, \\
    &\boxed{U = C_V T = \nu c_V T} .
\end{align}
Таким образом, внутренняя энергия газа однозначно зависит от его температуры, без привлечения иных велчин.

### Уравнение Роберта Майера
Рассмотрим молярные теплоёмкости изобарного и изохорного процессов, а именно их разность в случае идеального газа.
$$\require{enclose}$$
\begin{align}
    &c_p - c_V = \enclose{updiagonalstrike}{c_V} + \dfrac{1}{\nu} \left[\left( \dfrac{\partial U}{\partial V} \right)_T + P \right] \left( \dfrac{\partial V}{\partial T} \right)_p - \enclose{updiagonalstrike}{c_V} \nonumber\\
    &= \dfrac{1}{\nu} \cdot p \left( \dfrac{\partial V}{\partial T} \right)_p = \dfrac{1}{\nu} \cdot p \dfrac{\partial}{\partial T} \left( \dfrac{\nu R T}{p} \right)_p \nonumber\\
    &= \dfrac{1}{\nu} \cdot p \cdot \dfrac{\nu R}{p} = R, \\
    &\boxed{c_p - c_V = R} .
\end{align}

# Эффект Джоуля-Томсона
Рассмотрим теплоизолированный от внешнего мира газ, который под действием поршня пропускают через пористую перегородку ― из левой части объёмом $V_1$ в правую объёмом $V_2$.
\begin{align}
    &0 = \int \limits^2_1 \delta Q = U_2 - U_1 + p_2 V_2 - p_1 V_1 = (U_2 + p_2 V_2) - (U_1 + p_1 V_1) = I_2 - I_1, 
    &I = \text{const}.
\end{align}
Эксперименты показали, что чем более разрежен газ, тем меньше изменяется температура, а значит и энтальпия (так как $pV = \nu R T$). Значит, $U_2 - U_1 = 0$.

# Уравнение политропы
Политропный процесс ― процесс, при котором теплоёмкость постоянна. Мы уже знакомы со многими политропными процессами: изохорный, изобарный, изотермический ($C = + \infty$), адиабатический ― но бывают и иные политропные процессы. Выведем уравнение, характеризующее произвольный политропный процесс.

\begin{align}
    &C = \dfrac{\mathrm{d}U + p\mathrm{d}V}{\mathrm{d}T} = C_V + \dfrac{p\mathrm{d}V}{\mathrm{d}T}, \\
    &C - C_V = p \dfrac{\mathrm{d}V}{\mathrm{d}T}. \\
    &C_p = \left( \dfrac{\partial U}{\partial T} \right)_V + p \left( \dfrac{\partial V}{\partial T} \right)_p = C_v + p \left( \dfrac{\partial V}{\partial T} \right)_p, \\
    &C - C_p = p \dfrac{\mathrm{d}V}{\mathrm{d}T} - p \left( \dfrac{\partial V}{\partial T} \right)_p. \\
    &\mathrm{d}V = \left(\dfrac{\partial V}{\partial T} \right)_p \mathrm{d} T + \left(\dfrac{\partial V}{\partial p} \right)_T \mathrm{d}p. \\
    &C - C_p = p \left[ \left(\dfrac{\partial V}{\partial T} \right)_p + \left(\dfrac{\partial V}{\partial p} \right)_T \dfrac{\mathrm{d}p}{\mathrm{d}T} - \left(\dfrac{\partial V}{\partial T} \right)_p  \right] = p \left(\dfrac{\partial V}{\partial p} \right)_T \dfrac{\mathrm{d}p}{\mathrm{d}T}. \\
    &\dfrac{C - C_p}{C - C_v} = \enclose{updiagonalstrike}{p} \left(\dfrac{\partial V}{\partial p} \right)_T \dfrac{\mathrm{d}p}{\enclose{updiagonalstrike}{\mathrm{d}T}} \cdot \dfrac{\enclose{updiagonalstrike}{\mathrm{d}T}}{\enclose{updiagonalstrike}{p} \mathrm{d}V} = \left(\dfrac{\partial V}{\partial p} \right)_T \dfrac{\mathrm{d}p}{\mathrm{d}V}. 
    &\left(\dfrac{\partial V}{\partial p} \right)_T = \dfrac{\partial}{\partial p} \left( \dfrac{\nu R T}{p} \right)_T = - \dfrac{\nu R T}{p^2} = - \dfrac{p V}{p^2} = - \dfrac{V}{p}, \\
    &\dfrac{C - C_p}{C - C_v} = - \dfrac{V}{p} \dfrac{\mathrm{d}P}{\mathrm{d}V}.
\end{align}
Введём показатель политропы ― величину $n = \dfrac{C - C_p}{C - C_v}$, которая будет характеризовать конкретный политропный процесс.
\begin{align}
    &n = - \dfrac{V}{p} \dfrac{\mathrm{d}p}{\mathrm{d}V}, \\
    &n \ \dfrac{\mathrm{d}V}{V} = - \dfrac{\mathrm{d}p}{p}, \\
    &n \int \limits^V_{V_0} \dfrac{\mathrm{d}\tilde{V}}{\tilde{V}} = - \int \limits^p_{p_0} \dfrac{\mathrm{d}\tilde{p}}{\tilde{p}}, \\
    &n \ln \dfrac{V}{V_0} = - \ln \dfrac{p}{p_0}, \\
    &\dfrac{V^n}{V_0^n} = \dfrac{p_0}{p}, \\
    &p V^n = p_0 V_0^n, \\
    &\boxed{p V^n = \text{const}} .
\end{align}
