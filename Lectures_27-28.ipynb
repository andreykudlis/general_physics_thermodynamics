## Нестационарные задачи
Допускаем зависимость от времени. Сама же среда пусть будет однородна, тогда $ae=\text{const}$, более того, $\rho,\;c_v=\text{const}$.
\begin{align*}
    \implies \rho c_v\frac{\partial T}{\partial t}\;=\;ae\dfrac{\partial^2T}{\partial x^2}+q(x,t).
\end{align*}
Чтобы решить задачу нам надо определить начальные и граничные условия. С граничными мы уже знакомились.

Уравнение первого порядка по времени: необходимо задать распределение температуры в зависимости от координаты в начальный момент времени $t=0$:

\begin{align*}
    T(x)\Big |_{t=0}\;=\;f(x),
\end{align*}
где $f(x)$ $\;-\;$ известная функция от координаты.

С граничными условиями по координате будет больше функций. Например, в случае пластинки длиной $l$:
\begin{align*}
    T(t)\Big |_{x=0}\;=\;\varphi_1(t),\\
    T(t)\Big |_{x=l}\;=\;\varphi_2(t).\\
\end{align*}

В математике есть две задачи: доказать существование решения и его едиственность. Поэтому при решении задач такого рода возникают эти же вопросы.

Пусть уравнение:
\begin{align}
    \frac{\partial T}{\partial t}\;=\;\psi\dfrac{\partial^2T}{\partial x^2}\;+\;\dfrac{q(x,t)}{c_v\rho}    
\end{align}
имеет два решения: $T_1(x,t)$ и $T_2(x,t)$. Тогда:
\begin{align}
    &\theta(x,t)\;=\;T_1(x,t)-T_2(x,t),\\
    &\implies\boxed{\frac{\partial \theta}{\partial t}\;=\;\chi\frac{\partial^2 \theta}{\partial x^2}}.
\end{align}
Получившееся уравнение линейное. Заметим, что во всех граничных случаях $\theta$ = 0. Для доказательства единственности решения введём $I$:
\begin{align}
    I(t)\;=\;\int\limits^l_0\theta^2(x,t)d{x},
\end{align}
тогда:
\begin{eqnarray*}
    \dfrac{d{I(t)}}{d{t}}&&=\;\int\limits^l_02\theta(x,t)\frac{\partial \theta}{\partial t}d{x}\;=\;2\chi\int\limits^l_0\theta(x,t)\frac{\partial \theta}{\partial x}d{x}=\\
    &&=2\psi\theta(x,t)\frac{\partial \theta}{\partial x}\Big |^l_0-2\chi\int\limits^l_0\underset{\geqslant\;0}{\left(\frac{\partial \theta}{\partial x}\right)^2}d{x}.
\end{eqnarray*}
Значит, что функция со временем не возрастает. С учётом того, что её граничные условия равны нулю, то:
\begin{align}
    &\begin{cases}
    \dfrac{d{I}}{d{t}}\leqslant0,\\
    I(0)\;=\;0,\\
    I(t)\;=\;\int\limits^l_0\theta^2(x,t)d{x}\geqslant0
    \end{cases}\implies\\
    &\implies I\;=\;\text{const}\;=\;I(0)\;=\;0\\
    &\implies \theta(x,t)\equiv0\\
    &\implies T_2 \equiv T_1.
\end{align}

## Уравнение Ланжевена (на лекции ещё не было)
Применяется для описания эволюции набора степеней свободы, подобных тем, которые мы рассматривали в случае броуновского движения. Должен подчиняться марковской динамике (без памяти).

Уравнения Ланжевена содержит в себе как силы трения, так и случайные силы. Например:
\begin{align}
    &dot{x} = - \dfrac{1}{mB}\dot{x} + \dfrac{\xi}{m},\\ 
\end{align}
где $\xi$ $\;-\;$ тепловой шум (случайная сила).
\begin{align}
    &\dot{v} = - \dfrac{1}{mB}v  + \dfrac{\xi}{m}.
\end{align}
Решение такого уравнения имеет вид:
\begin{align}
    &v(t)=v_0\exp{-\lambda t} + \exp{-\lambda t}\int\limits_0^t \xi(\tau)\exp{\lambda \tau} d \tau. \label{lan}
\end{align}

Шум распределён независимо по времени:
\begin{align}
    &<\xi(t)> = 0, \nonumber \\
    &<\xi(t)\xi(t\;')> = \delta (t-t\;').
\end{align}

Если сделать усреднение уравнения 16.71 по шуму, то получим:
\begin{align}
    <v(t)> = v_o\exp{-\lambda t}.
\end{align}
Делаем что-то:
\begin{align}
    &<v(t_1)v(t_2)>\!\!\mid_{t_1=t_2} = v_0^2 \exp{-2\lambda t_1} + \dfrac{\exp{-2\lambda t_1}}{m^2} \cdot \\ \nonumber 
    &\cdot \int\limits_0^{t_1} <\xi(\tau_1)\exp{\lambda \tau_1} d \tau_1 \int\limits_0^{t_1} \xi(\tau_2)>\exp{\lambda \tau_2} d \tau_2 = \nonumber \\
    &=v_0^{-2\lambda t_1} + \dfrac{b}{2m^2\lambda}(1 - \exp{-2\lambda t_1}) \to \dfrac{b}{2 \lambda m^2}.
\end{align}
А если хотим сделать ещё что-то:
\begin{align}
    <v^2> = \dfrac{kT^2}{2m} = \dfrac{kT}{m} = \dfrac{b}{2\lambda m^2} \implies b = 2\lambda m k T = 2 \dfrac{1}{mB}mkT = \dfrac{2kT}{B}.
\end{align}

## Элементы механики жидкостей
Жидкости обладают только объёмной упругостью. По сравнению с газами, у жидкостей очень маленькая сжимаемость. Жидкость, в которой сжимаемостью можно принебречь, называют \textbf{несжимаемой}.

Как мы уже знаем, в жидкостях есть силы трения и вязкости. Жидкость, в которой при всяких движениях не возникают силы трения, называется \textbf{идеальной}.

Силы всегда можно разделить на массовые (объёмные) и поверхностные (нормальное или касательное напряжение). Пусть есть только силы нормального давления (для идеальной жидкости).

Рассмотрим элемент объёма и определим равнодействующую сил давления:
РИСУНОК
\begin{align}
    &F= [p(x)-p(x+d x)]S = -\frac{\partial p}{\partial x}d x S = \underbrace{-\frac{\partial p}{\partial x}}_{S_x}d V.\\
    \text{В других направлениях:}
        &S_y = - \frac{\partial p}{\partial y}; \quad S_z = -\frac{\partial p}{\partial z}; \quad S_x = -\frac{\partial p}{\partial x}.
\end{align}
$S_x$, $S_y$, $S_z$ $\;-\;$ проекции силы, вызванной разностью нормального давленя, на различные оси, действующей на единицу объёма жидкости $d V$. Так, можно состряпать вектор:
\begin{align}
    \boxed{\vec{S} = -grad p}.
\end{align}
$\vec{S}$ $\;-\;$ результирующая сил давления.

В состоянии равновесия эта сила должна быть уравновешена массовой силой $\vec{f}$:
\begin{align}
    &(\vec{S} + \vec{f})d V = 0,\\
    &-grad p + \vec{f} = 0.\\ 
    \text{Получаем основное уравнение гидростатики:}
    \implies &\boxed{\vec{f} = grad p}.
\end{align}

Напишем основное уравнение гидродинамики идеальной жидкости из второго закона Ньютона (уравнение Эйлера):
\begin{align}
    &m\dfrac{d \vec{v}}{d t} = \vec{F} - grad p V,\\
    &\boxed{\rho \dfrac{d \vec{v}}{d t} = \vec{f} - grad p}.
\end{align}

Так, например, в поле силы тяжести
\begin{align}
    & \vec{f} = \rho \vec{g}, \\
    & \dfrac{\partial p}{\partial x} = \dfrac{\partial \rho}{\partial y} = 0, \\
    & \dfrac{\partial \rho}{\partial z} = - \rho g, \\
    &p = p_0 - \rho g z.
\end{align}

Рассмотри объём жидкости в состоянии равновесия с остальной жидкостью. Допустим, это всё находится в неком потенциальном поле. Обозначая вес жидкости в объёме за $\vec{Q}$, а сумму сил, связанных с давлением, как $\vec{F}$, получаем
\begin{align}
    & \vec{Q} + \vec{F} = 0.
\end{align}
Если заменить данный объём жидкости на твёрдое тело, то в случае равновесия ситуация должна оказаться точно такой же (силы гидростатического далвения те же). Значит, \textbf{в равновесии давление жидкости на тело равно весу жидкости, вытесненному этим телом}.

## Кинематическое описание движения жидкости

Перейдём от рассмотрения отдельных частиц к изучению свойств некой точки в пространстве ― так, например, рассмотрим скорости всех частиц, проходящих через эту точку пространства. Получится векторное поле, сопоставляющее каждой точке в каждый момент времени вектор скорости частицы в этой точке.

Если поле скоростей не зависит от времени ― $\vec{v}(\vec{r})$ ― течение называется стационарным. Иначе ― $\vec{v}(\vec{r}, t)$ ― будем говорить о нестационарном течении.

Введём понятие линии тока ― линии, касательная к которой в каждой точке соответствует направлению скорости в ней. Такие линии, проведённые от данной поверхности, образуют трубку тока.

Масса жидкости, протекающей через поперечное сечение трубки ток в единицу времени, очевидно, составит
\begin{align}
    & d m = \rho v S d t.
\end{align}
Понятно, что при стационарном течении жидкость не может выходить за боковую поверхность трубки тока. Масса, проходящая через сечения трубки тока сохраняется:
\begin{align}
    & \rho_1 v_1 S_1 = \rho_2 v_2 S_2, \\
    & \text{а если жидкость несжимаема, то } \rho_1 = \rho_2, \\
    & \dfrac{v_1}{v_2} = \dfrac{S_1}{S_2}.
\end{align}
Последнее уравнение очень активно употрелбяют сантехники при расчёте трубопроводов и прочей инфраструктуры.

## Уравнение Бернулли
Рассмотрим идеальную (есть только силы нормального давлления) несжимаемую жидкость. Примем течение стационарным в потенциальном поле силы тяжести.

За некоторое время $M N D C$ перейдёт в $M' N' D' C'$. Рассмотрим работу сил давления с обоих концов (знаки различаются из-за того, что в первом случае сила давления сонаправлена с движением, а во втором противонаправлена):
\begin{align}
    & A_1 = + P_1 S_1 l_1 = P_1 \Delta V_1 = P_1 \; \dfrac{\Delta m)1}{\rho_1}, \\
    & A_2 = - P_2 S_2 l_2 = P_2 \Delta V_2 = P_2 \; \dfrac{\Delta m_2}{\rho_2}. \\
    & \text{Из стационарности течения } \Delta m_1 = \Delta m_2 = \Delta m. \\
    & A = A_1 + A_2 = \left(\dfrac{P_1}{\rho_1} - \dfrac{P_2}{\rho_2} \right) \Delta m. \\
\end{align}
Работа сил давления идёт на изменение энергии. Изначально выразим энергии в пересчёте на единицу массы:
\begin{align}
    & A = \varepsilon_2 \Delta m - \varepsilon_1 \Delta m, \\
    & \dfrac{P_1}{\rho_1} + \varepsilon_1 = \dfrac{P_2}{\rho_2} + \varepsilon_2. \\
    & \boxed{\dfrac{P}{\rho} + \varepsilon = \text{const}} \text{ ― уравнение Бернулли}.
\end{align}
Если же энергия не зависит от сжатия, то можно раскрыть выражение для удельной энергии, получая
\begin{align}
    & \boxed{\dfrac{v^2}{2} + g h + \dfrac{P}{\rho} = \text{const}}. 
\end{align}
