## Адиабатический процесс

В прошлый раз мы остановились на политропическом процессе
<br>

$$PV^{n}\;=\; const,\; n\;=\; \dfrac{c\;-\;c_{p}}{c\;-\;c_{V}}$$

При n = 1 $\;-\;$ изотермический процесс,

n = 0 $\;-\;$ изобарный процесс $P = const$,

$n\;=\; \infty$ $\;-\;$ изохорный процесс, легче понять если:
$$PV^{n}\;=\; P \cdot \left ( \dfrac{RT}{P} \right )^{n} \;=\; P^{1-n}\;T^{n}\;=\;const$$
$$\Rightarrow P^{\frac{1}{n}\;-\;1}\;T\;=\; const \Rightarrow \dfrac{T}{P}\;=\; const.$$

$n\;=\; \gamma$ $\;-\;$ адиабатный процесс.
$$c\;=\; 0 \Rightarrow n\;=\; \gamma \;=\; \dfrac{c_{p}}{c_{V}}.$$

Уравнение $PV^\gamma \;=\; const$ $\;-\;$ это уравнение Пуассона (1781 $\;-\;$ 1840).

**Что можно сказать о поведении газа при ад. процессе?**

$$\gamma \;>\; 1 \Rightarrow T V^{\gamma \;-\; 1} \;=\; const \Rightarrow \uparrow V \Rightarrow \downarrow T$$
(т.е. работа над газом меняет температуру)

**Nota Bene:** Мы смотрели только квазиравновесный процесс!!!

$$\rho Sd x\;=\;d m$$

С другой стороны масса должна быть такой же.

<br>

$$\rho\;=\; \rho_0\;+\; \rho(x),$$
\begin{align}
    \rho_0 Sd x\;=\; (\rho_0\;+\;\rho(x)) \left [ x\;+\; d x \;+\; U(x\;+\;d x)\;-\; x\;-\;U(x) \right] S, 
\end{align}
\begin{align}
    \rho_0 Sd x\;=\; (\rho_0\;+\;\rho(x)) \left [ x\;+\; d x \;+\; U(x)\;+\; \frac{\partial U}{\partial x} d x\;-\; x\;-\;U(x) \right] S,
\end{align}
\begin{align}
    \rho(x)\;=\; -(\rho_0 \;+\; \rho(x)) \frac{\partial U}{\partial x}, \\
    \frac{\partial \rho}{\partial x}\;=\;-\rho_0 \frac{\partial^2 U}{\partial x^2} \; \approx \; -\rho \frac{\partial^2 U}{\partial x^2}.
\end{align}

<br>

Второй закон Ньютона:

<br>

\begin{align}
    S \rho d x \frac{\partial^2 U}{\partial t^2} \;=\; S p(x,t)\;-\;Sp(x\;+\;d x,t)\;=\; S \left [ - \frac{\partial p}{\partial x} d x \right] 
\end{align}
\begin{align}
    \Rightarrow \; \rho d x \frac{\partial^2 U}{\partial t^2} \;=\; - \frac{\partial p}{\partial x} d x.
\end{align}

\begin{align}
\begin{cases}
    \rho \frac{\partial^2 U}{\partial t^2} \;=\; - \frac{\partial p}{\partial x};\\
    \frac{\partial \rho}{\partial x}\;=\; -\rho \frac{\partial^2 U}{\partial x^2}.
\end{cases}
\end{align}

<br>

**Nota bene:**
$$\rho \frac{\partial^2 U}{\partial t^2}\;=\; -\frac{\partial p}{\partial \rho} \frac{\partial \rho}{\partial x}\;=\; \frac{\partial p}{\partial \rho} \rho \frac{\partial^2 U}{\partial x^2}.$$

\begin{align}
    \Rightarrow \frac{\partial^2 U}{\partial t^2}\;=\; \frac{\partial p}{\partial \rho} \frac{\partial^2 U}{\partial x^2}.
\end{align}

\begin{align}
    \frac{\partial^2 U}{\partial x^2} \;-\; \dfrac{1}{\left( \sqrt{\frac{\partial p}{\partial \rho}} \right)^2} \frac{\partial^2 U}{\partial t^2}\;=\; 0.
\end{align}

$$\Rightarrow \; c\;=\; \sqrt{\frac{\partial p}{\partial \rho}}.$$

<br>

**Процесс какой?**

Если предположить, что температура между сгущениями одинакова (успевает выравниваться)

$$\Rightarrow \text{процесс р-ия изотермический}$$

<br>

\begin{align}
    \Rightarrow \sqrt{(\frac{\partial p}{\partial \rho})_{T}}\;=\; \sqrt{\dfrac{P}{\rho}}.
\end{align}

<br>

Для воздуха:

$$\approx 281 \text{м/с, однако опыт: 330 м/с}.$$

**Что не так?**

Колебания плотности происходят очень быстро (теплопроводность мала). Т.е. температура не успевает выравниваться, следовательно, процесс происходит быстро. Теплообмен не играет роли, значит, адиабатический процесс.

<br>

\begin{align}
    &\Rightarrow PV^\gamma = const 
    \\
    &PV = \dfrac{m}{M} RT
    \\
    &m= V\rho
    \\
    &\Rightarrow \dfrac{P}{\rho^{\gamma}} = const
    \\
    &\Rightarrow d\Bigg[ \dfrac{P}{\rho^\gamma} \Bigg] = 0 \Rightarrow \dfrac{1}{\rho^\gamma}dP + Pd\dfrac{1}{\rho^{\gamma}} = 0
    \\
    &\dfrac{dP}{\rho^{\gamma}} + (-\gamma)\rho^{-\gamma -1} Pd\rho = 0
    \\
    &\rho dP - \gamma Pd\rho = 0
    \\
    &\dfrac{dP}{d\rho} = \gamma \dfrac{P}{\rho} \text{ $\;-\;$ для адиабаты.}
    \\
    \\
    & \frac{\partial p}{\partial \rho}_{\text{ад}} = \gamma \dfrac{P}{\rho} \Rightarrow C_{\text{ск. зв.}} = \sqrt{\gamma \dfrac{P}{\rho}}
    \\
    &\text{Для воздуха:}\;\gamma \approx \dfrac{7}{5} = \dfrac{i+2}{i} \; \; |i=5
\end{align}

<br>

**Замечания:**

Соотношение $\dfrac{(\frac{\partial p}{\partial \rho})_{\text{ад.}}}{(\frac{\partial p}{\partial \rho})_{T}}=\dfrac{C_P}{C_V}$ справедливо для любого изотропного однородного вещества.

Получим его адиабатический процесс:

<br>

\begin{align}
    &\delta Q = 0 \Rightarrow dU + PdV = 0
    \\
    &dU = (\frac{\partial U}{\partial V})_pdV + (\frac{\partial U}{\partial P})_VdP
    \\
    &\Rightarrow (\frac{\partial U}{\partial V})_pdV + (\frac{\partial U}{\partial P})_VdP + PdV = 0
    \\
    &(\frac{\partial U}{\partial V})_p + (\frac{\partial U}{\partial P})_V (\frac{\partial P}{\partial V})_{\text{ад.}} + P = 0
    \\
    &\Rightarrow (\frac{\partial P}{\partial V})_{\text{ад.}} = - \Bigg[ \dfrac{(\frac{\partial U}{\partial V})_p + P}{(\frac{\partial U}{\partial P})_V}\Bigg] \circeq
    \\
    &(\frac{\partial U}{\partial P})_V = (\frac{\partial U}{\partial T})_V (\frac{\partial T}{\partial P})_V = C_v (\frac{\partial T}{\partial P})_V
    \\
    &(\frac{\partial U}{\partial V})_p + P = (\frac{\partial U}{\partial T})_p (\frac{\partial T}{\partial V})_p + P = C_p (\frac{\partial T}{\partial V})_p
\end{align}
\begin{align}
    & \circeq - \dfrac{C_p (\frac{\partial T}{\partial V})_p}{C_V (\frac{\partial T}{\partial P})_V} = -\dfrac{C_p}{C_V} \cdot \dfrac{\partial (T,P) \; \partial (P,V)}{\partial (V,P) \; \partial (T,V)} = \dfrac{C_p}{C_V} \cdot (\frac{\partial P}{\partial V})_T
    \\
    &\Rightarrow (\frac{\partial P}{\partial V})_{\text{ад.}} = \dfrac{C_p}{C_V} \cdot (\frac{\partial P}{\partial V})_T
    \\
    &\Rightarrow (\frac{\partial P}{\partial V})_{\text{ад.}} = \dfrac{C_p}{C_V} \cdot (\frac{\partial P}{\partial V})_{\rho}
\end{align}


<br>

Рассмотрим смесь:

Для каждого газа: $P_{i}V = \nu_i RT$

Молярная масса смеси: $M = \dfrac{m}{\nu} = \dfrac{\sum \mu_i}{\sum \nu_i}$

Внутренняя энергия: 

<br>

\begin{align}
    &U_i = \nu_i C_{V_i} T = \dfrac{\nu_i}{\gamma_i - 1} RT 
    \\
    &\gamma_i = \dfrac{C_{V_i} +R}{C_{V_i}} = 1 + \dfrac{R}{C_{V_i}} \; \Rightarrow \; C_{V_i} = \dfrac{R}{\gamma_i - 1}
    \\
    &\begin{cases}
        &U = \sum U_i
        \\
        &U = \dfrac{\nu}{\gamma - 1} RT
    \end{cases}
    \\
    &\dfrac{\nu}{\gamma - 1} RT = \sum \dfrac{\nu_i}{\gamma_i - 1} RT
    \\
    &\Rightarrow \dfrac{\sum \nu_i}{\sum \dfrac{\nu_i}{\gamma_i - 1}} = \gamma -1
    \\
    &\boxed{\gamma = 1+\dfrac{\sum \nu_i}{\sum\dfrac{\nu_i}{\gamma_i - 1}}} \; .
\end{align}

## Второе начало термодинамики

1. Первое начало $\;-\;$ это ЗСЭ. Он не говорит о том, *как* протекают процессы.
2. Второе начало говорит о *направлении* процесса.
3. Второе начало позволяет ввести *количественную меру температуры*.

Первая формулировка второго начала термодинамики была дана в 1850 году Рудольфом Клаузиусом и Вильямом Томсоном.

1. Начинаем греть $\implies$ расширяется с нагревом:
\begin{align*}
    Q_1\;=\;\Delta U\;+\;A_{1a2}.
\end{align*}
2. Возращаем поршень (приводим в контакт с холодильником):
\begin{align*}
    -Q_2\;=\;-\Delta U\;-\;A_{1b_2},
\end{align*}
где знак минус появляется, если говорим про отданное тепло холодильнику.
$$\implies\;Q_1\;-\;Q_2\;=\;A_{1a2}\;-\;A_{1b2}\;=\;Q\;=A.$$
$Q$ $\;-\;$ тепло, которое ушло на совершение работы. 

С технической точки зрения, нам хотелось бы, чтобы весь $Q_1$ шел на работу.

<br>

\begin{align}
    \eta\;=\;\dfrac{Q_1\;-\;Q_2}{Q_1}\;=\;\dfrac{A}{Q_1}.
\end{align}

<br>

$\cdot$ **Формулировка Томсона:** невозможен такой циклический процесс, единственным результатом которого была бы передача всего тепла на совершение работы.

$\cdot$ **Формулировка Клаузиуса** теплота не может самопроизвольно переходить от тела менее нагретого к телу более нагретому. То есть нельзя забрать тепло от менее нагретого тела, целиком передать его более нагретому и притом так, чтобы в окружающем мире ничего не поменялось бы.

Два рассуждения:

1) Если невозможен процесс по Томсону, то невозможен по Клаузиусу. Пусть по Томсону невозможен, а по Клаузиусу возможен $\implies$ совершаем цикл $A = Q_1 - Q_2$ $\implies$ забираем у холодильника $Q_2$ и отдаём нагревателю. $\implies Q_1 - Q_2$ $\;-\;$ теплота, которую забрали у нагревателя, и работа $A = Q_1 - Q_2 $. Противоречие.

2) Если невозможен процесс по Клаузиусу, то невозможен и по Томсону. Пусть невозможен по Клаузиусу, но возможен по Томсону. Отнимаем у холодильника тепло $Q$. Пусть $-A_\text{внешняя}=Q$, то есть совершаем работу. Этой работой начинаем греть более нагретое тело, но такой процесс невозможен.

## Цикл Карно

<br>

<figure>
    <center>
    <img src="images/3-4,32,2.png" width="400" align="center"/>

<br>

Рассмотрим следующий круговой процесс: два тепловых резервуара $\underset{\text{Н}}{T_1},\; \underset{\text{Х}}{T_2}, \; T_1 >T_2$.

1) Тепловой контакт с $T_1$ (сама система имеет $T_1$)

2) Квазистатическое равновесие (бесконечно уменьшать внешнее давление)

3) За 1-2 она забирает тепло $Q_1$ и совершает работу $A_{12}$

4) Адиабатическая оболочка (дают возможность расширяться)

5) Её температура достигает $T_2$ (совершена работа $A_{23}$)

6) Приводят в тепловой контакт с $T_2$ (увеличением давления сжимают, совершлили работу, или система совершает работу $A_{34}$ и отдаёт тепло $Q_2$)

7) Адиабатически сжимают систему до исходного состояния.

Таким образом, работа за цикл $A = A_{12} + A_{23} + A_{34} + A_{41} = Q_1 - Q_2$,
\begin{align}
    \eta = \dfrac{Q_1 - Q_2}{Q_1} = 1 - \dfrac{Q_2}{Q_1}. 
\end{align}
Теорема Карно: коэффициент полезного действия тепловой машины, работающей по циклу Карно, зависит только от температур $T_1$ и $T_2$ нагревателя и холодильника и не зависит от устройства машины и вида рабочего вещества.

## Термодинамическая шкала температур
Теоремой Карно можно воспользоваться для построения температурной шкалы, которая не зависит от особенностей термометра. Пусть $T_1$ и $T_2$ - измеренные термометром температуры (эмпирические).

<br>

\begin{align}
\eta = \dfrac{Q_1 - Q_2}{Q_1} = f(T_1,T_2),
\end{align}

<br>

Это универсальная функция, согласно теореме Карно. 

Как построить шкалу?

Пусть $\varphi(T_1,T_2) = Q_1/Q_2, \ \ 1 - 1/\varphi = f$. 

<br>

<figure>
    <center>
    <img src="images/3-4,34.png" width="400" align="center"/>
        <figcaption>Рис. 2.7: Построение термодинамической шкалы температур (два цикла Карно друг на друге).</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &\dfrac{Q_1}{Q_2} = \varphi(T_1,T_2), \nonumber \\
    &\dfrac{Q_2}{Q_3} = \varphi(T_2,T_3), \nonumber \\
    &\dfrac{Q_1}{Q_3} = \varphi(T_1,T_2)\cdot\varphi(T_2,T_3) = \varphi(T_1, T_3)\nonumber \\
    &\implies\varphi(T_1,T_2) = \dfrac{\varphi(T_1, T_3)}{\varphi(T_2,T_3)}\nonumber \\
    &\implies\dfrac{Q_1}{Q_2} = \dfrac{\varphi(T_1, T_3)}{\varphi(T_2,T_3)} \ \text{― для любого} \ T_3\\
    &\implies \dfrac{Q_1}{Q_2}\;=\;\dfrac{\theta\left(T_1\right)}{\theta\left(T_2\right)},\\
    &\implies \varphi\left(T_1,\;T_2\right)\;=\;\dfrac{\theta\left(T_1\right)}{\theta\left(T_2\right)}.
\end{align}

<br>

Так как в данном случае отношение функций не зависит от $T_3$, то можно перейти к термодинамической температуре $\theta(T)$.

<br>

\begin{align}
    &\dfrac{Q_1}{Q_2} = \dfrac{\theta(T_1)}{\theta(T_2)}.
\end{align}

<br>

$\implies\;\theta\;$ $\;-\;$ есть функция температуры, так как ир-ия ещё нет о температуре, то почему бы самой этой функции не стать температурой.

<br>

\begin{align}
    &\boxed{\dfrac{Q_1}{Q_2} = \dfrac{T_1}{T_2}}.
\end{align}

<br>

Итак, мы получили $\theta$ $\;-\;$ **абсолютную термодинамическую температуру (шкалу)**.

Осталось согласовать реперные точки двух шкал, что позволяет закрыть глаза на разницу между термодинамической и газовой шкалами температуры:

Варианты её определения: либо фиксируем **две точки** разных агрератных состояний воды, и говорим, что разница между ними равна 100; либо берём одну точку $\;-\;$ **тройную точку воды** (строго определённые значения температуры и давления, при которых вода может быть в виде любой из трёх фаз, а именно: 273.16 К $\left[0.01 ° \text{C}\right]$ и давление 611.7 Па).

<br>

\begin{align*}
    \dfrac{Q_\text{к}}{Q_\text{п}}\;=\;\dfrac{\theta_\text{к}}{\theta_\text{п}}; \qquad \theta_\text{к}\;-\;\theta_\text{п}\;=\;100,\qquad \theta_\text{к}\;=\;273.15\;\text{К},\;\;\theta_\text{п}\;=\;373.15\;\text{К}.
\end{align*}

<br>

## Эквивалентность термодинамической и идеально-газовой шкал температур
Осуществим цикл Карно над идеальным газом: 

<br>

<figure>
    <center>
    <img src="images/3-4.Carnot.png" width="400" align="center"/>
        <figcaption>Рис. 2.8: Цикл Карно. 1-2 и 3-4: изотермы; 2-3 и 4-1: адиабаты.</figcaption>
    </center>
</figure>

<br>

\begin{align}
    &PV\;=\;RT\;=\;\text{const},\nonumber\\
    &Q_{12}\;=\;A_{12}\;=\;\int\dfrac{d VRT_1}{V}\;=\;RT_1\ln\left(\dfrac{V_2}{V_1}\right);\\
    &Q_{23}\;=\;0,\;PV^\gamma\;=\;\text{const}\;\implies\;TV^{\gamma-1}\;=\;\text{const};\nonumber\\
    &\implies\;T_1V_2^{\gamma-1}\;=\;T_2V_3^{\gamma-1}.\\
    &Q_{34}\;=\;A_{34}\;=\;RT_2\ln\left(\dfrac{V_4}{V_3}\right),\\
    &Q_{41}\;=\;0\;\implies\;T_2V_4^{\gamma-1}\;=\;T_1V_1^{\gamma-1}.
\end{align}

<br>

Если нам требуется посчитать отданное тепло, то:

<br>

\begin{align}
    &Q_2\;=\;-Q_{34}\;=\;RT_2\ln\left(\dfrac{V_3}{V_4}\right),\nonumber\\
    &\implies\dfrac{Q_2}{Q_1}\;=\;\dfrac{RT_2\ln\left({V_3}/{V_4}\right)}{RT_1\ln\left(V_2/V_1\right)},
\end{align}
где: $V_1/V_4\;=\;V_2/V_3\;\implies\;V_3/V_4\;=\;V_2/V_1$.

## Неравенство Клаузиуса. Вторая теорема Карно

Рассмотрим две тепловые машины, одна из которых работает по циклу Карно, и два тепловых резервуара (источника теплоты) $R_1$ и $R_2$ температуры $T_1$ и $T_2$ соответственно.

<br>

<figure>
    <center>
    <img src="images/2lecture4_37.png" width="400" align="center"/>
        <figcaption>Рис. 2.12: Две машины и два резервуара.</figcaption>
    </center>
</figure>

<br>

Пусть первая тепловая машина получает теплоту $Q_1$ от $R_1$ и теплоту $Q_2$ от $R_2$, совершая за цикл работу $Q_1 + Q_2$ (где какое-то из количеств теплоты отрицательно). 

Вторая машина работает по циклу Карно ― варьируя её параметры, мы можем подобрать различные значения теплоты $Q_1'$, $Q_2'$, забираемые ей из резервуаров. При этом, как установлено ранее, выполняется условие

<br>

\begin{align}
    &-\dfrac{Q_1'}{Q_2'} = \dfrac{|Q_1'|}{|Q_2'|} = \dfrac{T_1}{T_2},
\end{align}
где минус возникает из-за того, что какое-то количество теплоты отрицательное.

Общая работа, совершаемая обоими тепловыми машинами:

<br>

\begin{align}
    &A = Q_1 + Q_1' + Q_2 + Q_2'.
\end{align}

<br>

Подберём параметры цикла Карно таким образом, что $Q_1' = -Q_1$.

<br>

\begin{align}
    &A = Q_1 - Q_1 + Q_2 + Q_2' = Q_2 + Q_2' = Q_2 + \dfrac{T_2}{T_1} Q_1, \\
    &A = T_2 \left( \dfrac{Q_2}{T_2} + \dfrac{Q_1}{T_1} \right).
\end{align}

<br>

Если бы работа была положительной, то тепловая машина бы превращала в работу всю теплоту $Q_2 + Q_2'$ от теплового резервуара $R_2$, что противоречит второму началу термодинамики (в формулировке Томсона): значит, работа не больше нуля.

<br>

\begin{align}
    &A \leq 0, \\
    &T_2 \left( \dfrac{Q_2}{T_2} + \dfrac{Q_1}{T_1} \right) \leq 0.
\end{align}

<br>

Заметив, что температура $T_2$ всегда положительна, приходим к неравенству Клаузиуса

<br>

\begin{align}
    \boxed{\dfrac{Q_2}{T_2} + \dfrac{Q_1}{T_1} \leq 0}.
\end{align}

<br>

Из неравенства Клаузиуса прямо следует вторая теорема Карно: для удобства рассмотрим величину теплоты, полученной от холодильника $|Q_1| = - Q_1$.

<br>

\begin{align}
    &\dfrac{Q_2}{T_2} - \dfrac{|Q_1|}{T_1} \leq 0, \\
    &\dfrac{Q_2}{T_2} \leq \dfrac{|Q_1|}{T_1}, \\
    &\dfrac{T_1}{T_2} \leq \dfrac{|Q_1|}{Q_2}, \\
    &1 - \dfrac{|Q_1|}{Q_2} \leq 1 - \dfrac{T_1}{T_2}, \\
    &\boxed{\dfrac{Q_2 - |Q_1|}{Q_2} \leq \dfrac{T_2 - T_1}{T_2}}.
\end{align}

<br>

Из первой теоремы Карно следует, что равенство во второй теореме Карно достигается при использовании цикла Карно (или эквивалентногого ему цикла Стирлинга), что даёт явное выражение для КПД цикла Карно:

<br>

\begin{align}
    \boxed{\eta_{\text{к}} = \dfrac{T_2 - T_1}{T_2}},
\end{align}
где $T_2$ ― температура нагревателя, $T_1$ ― температура холодильника.
