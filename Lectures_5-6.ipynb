- Неравенство Клаузиуса (частный случай)

<br>

Для произвольной тепловой машины:
$$\boxed{\dfrac{Q_1}{T_1}\;+\; \dfrac{Q_2}{T_2}\; \leqslant \; 0} .$$

$\Rightarrow$ вторая теорема Карно 

<br>

КПД любой тепловой машины не может быть больше КПД идеальной машины, работающей по циклу Карно с теми же $T_1$ и $T_2$.

<br>

Обобщим неравенство Клаузиуса.

<br>

Работа за цикл:
\begin{align}
    S: A = Q_1 + ... + Q_n
\end{align}

<br>

Для каждого цикла Карно:
\begin{align}
    &\dfrac{Q_{0i}}{T_0} + \dfrac{Q'_i}{T_i} = 0,
    \\
    &Q_0 = \sum Q_{0i} = -T_0 \sum \dfrac{Q'_i}{T_i}.
\end{align}

**1 большой цикл:**
<br>

- тепло резервуара $R_0$: $\sum Q_{0i} = Q_0$ ("туда" или "обратно")
- тепло резервуаров $i=1, ..., n \; \sum (Q_i + Q'_i)$
<br>

$\Rightarrow$ Работа: $A = Q_0 + (Q_1 + Q'_1) + ... + (Q_n + Q'_n)$.

Настроим все машины Карно так, чтобы $Q_i + Q'_i = 0$.

<br>

$\Rightarrow \; R_1, ..., R_n$ \$-\$ тепло не отдадут и не получат, то есть из-за $\sum (Q_i + Q'_i) = 0$
\begin{align}
    \boxed{A = Q_0} - \text{то есть работа засчёт одного резервуара, а так НЕ может быть!!}
\end{align}

<br>

$\Rightarrow A$ \$-\$ должна быть неположительной

<br>

\begin{align}
&\Rightarrow \; -T_0 \sum \dfrac{Q'_i}{T_i} \leqslant 0,
\\
& T_0 \sum \dfrac{Q_i}{T_i} \leqslant 0,
\\
& \Rightarrow \boxed{\sum \dfrac{Q_i}{T_i} \leqslant 0} - \text{круговой процесс произвольной системы}.
\end{align}

<br>

Если же устремить $Q_i$ к бесконечно малой величине:
\begin{align}
    \Rightarrow \; \boxed{\oint \dfrac{\delta Q}{T} \leqslant 0} - \text{неравенство Клаузиуса.}
\end{align}

<br>

Надо понимать в самом общем смысле обмен с окружающим миром.

<br>

Можно сделать ограничение на кпд:
\begin{align}
    &\int \dfrac{\delta Q_1}{T_1} - \int \dfrac{\delta Q_2}{T_2} \leqslant 0,\\
    & \nonumber T_{1 max} - \text{максимальная температура нагревателя,}\\
    & \nonumber T_{2 min} - \text{минимальная температура холодильника.}\\
    & \int \dfrac{\delta Q_1}{T_{1 max}} - \int \dfrac{\delta Q_2}{T_{2 min}} \leqslant 0,\\
    & \dfrac{Q_1}{T_{1 max}} - \dfrac{Q_2}{T_{2 min}} \leqslant 0,\\
    &\dfrac{Q_2}{Q_1} \geqslant \dfrac{T_{2 min}}{T_{1 max}} \Rightarrow -\dfrac{Q_2}{Q_1} \leqslant -\dfrac{T_{2 min}}{T_{1 max}},\\
    & \nu \underset{def}{=} \dfrac{Q_1 -Q_2}{Q_1} \leqslant \dfrac{T_{1 max} - T_{2 min}}{T_{1 max}}.
\end{align}

## Равенство Клаузиуса

<br>


Во-первых, потребуем квазистатичности процесса. В этом случае под Т в неравенстве Клаузиуса можно понимать температуру самой системы: Т.

<br>



Мы обсуждали, что квазистатический процесс обратим, тогда запускаем его в обратном направлении: 
\begin{align}
    \oint \dfrac{\delta Q}{T}.
\end{align}

<br>


Процесс проходит через ту же последовательность, но в обратном направлении: $\delta Q\;'=-\delta Q\;\implies$
\begin{align}
    &\implies\;\begin{cases}
    \oint\dfrac{\delta Q}{T}\;\geqslant\;0\nonumber\\
    \oint\dfrac{\delta Q}{T}\;\leqslant\;0\nonumber
    \end{cases}\nonumber\\
    &\implies\boxed{\oint\dfrac{\delta Q}{T}\;=\;0}\label{eq:сlausius_equality}.
\end{align}
Равенство называется **равенством Клаузиуса**.

Теперь рассмотрим произвольный равновесный процесс:

<br>

<figure>
    <center>
    <img src="images/43.png" width="350" align="center"/>
        <figcaption>Рис. 2.1: Произвольный равновесный процесс</figcaption>
    </center>
</figure>

<br>

\begin{align*}
    &\int\limits_{1a2}\dfrac{\delta Q}{T}\;+\;\int\limits_{2b1}\dfrac{\delta Q}{T}\;=\;0,\\
    &\int\limits_{1a2}\dfrac{\delta Q}{T}\;=\;\int\limits_{1b2}\dfrac{\delta Q}{T} \implies \dfrac{\delta Q}{T}\text{приведенное количество элементарного тепла}
\end{align*}

<br>

Это тепло не зависит от пути! Тогда можно ввести полный дифференциал (**энтропию**):
\begin{align}
    dS\;=\;\dfrac{\delta Q}{T} \implies S_2\;-\;S_1\;=\;\int\limits_{1\to2}\dfrac{\delta Q}{T}.
\end{align}

<br>


## Энтропия идеального газа

<br>

\begin{align}
    &\delta Q\;=\;C_VdT\;+\;Pd{V}\;=\;C_Vd{T}\;+\;RT\dfrac{d{V}}{V}, \nonumber\\
    d{S}\;=\;\dfrac{\delta Q}{T}\;=\;C_V\dfrac{d{T}}{T}\;+\;R\dfrac{d{V}}{V}, \nonumber\\
    &S\;=\;\int C_V\dfrac{dT}{T}\;+\;R\ln{V}.
\end{align}

<br>


Если $C_V\neq C_V(T)\;\implies\;const$
\begin{align}
    S\;=\;C_V\ln{T}\;+\;R\ln{V}\;+\;const.\nonumber\\
\end{align}
Внимание! До этого мы работали с 1 молем. Сейчас добавим $\nu$.\\
\begin{align*}
    &S\;=\;\nu C_V\ln{T}\;+\;\nu R\ln{V}\;+\;const(\nu).
\end{align*}

<br>


Заметьте! $const\;=\;const(\nu)$. Почему? Потому что энтропия -- аддитивная величина. Представьте себе две камеры. В каждой было по $\nu$ вещества, каждая объёмом $V$. Тогда, при объединении, энтропия просто сложится и станет $2S$, $\nu\to 2\nu$, $V\to 2V$. Очевидно, что нужно прийти к такому виду:

<br>

\begin{align}
    &S\;=\;C_V\nu\ln{T}\;+\;\nu R\ln{\dfrac{V}{\nu}}\;+\widetilde{const},\nonumber\\
    &\boxed{S\;=\;\nu\left[C_V\ln{T}\;+\;R\ln{\dfrac{V}{\nu}}\;+\;\widetilde{\widetilde{const}}\right]}.
\end{align}

<br>


Равновесный адиабатический \$-\$ **изоэнтропийный**!!!

$0=\delta Q\implies dS=0$.
## Закон возрастания энтропии

\begin{align*}
    &\oint\limits_{1a2b1}\dfrac{\delta Q}{T}\leqslant0, \\
    &\int\limits_{1a2}\dfrac{\delta Q}{T}\;+\;\int\limits_{2b1}\dfrac{\delta Q}{T}\leqslant0, \\
    &\int\limits_{1a2}\dfrac{\delta Q}{T}+S_1-S_2\leqslant0, \\
    &S_2-S_1\geqslant\int\limits_{1a2}\dfrac{\delta Q}{T}.
\end{align*}
<br>

<br>


Если  процесс адиабатический, то $\delta Q=0 \implies$\\

<br>


\begin{align}
    \boxed{S_2\geqslant S_1}.
\end{align}

<br>


Значит, происходит неубывание энтропии в адиабатически изолированной системе.

<br>


Таким образом, второе начало говорит нам о направлении протекания процессов.

## Термодинамические функции

### Преобразования Лежандра

<br>

В общем случае термодинамическая функция $L$ естественных независимых переменных $x,\;y,\;z\dots$ имеет следующий полный дифференциал:

<br>

\begin{align}
    dL = \left(\dfrac{\partial L}{\partial x}\right)_{y,z\dots} \!\!\!\!\!\!\!\!\! dx + \left(\dfrac{\partial L}{\partial y}\right)_{x,z\dots} \!\!\!\!\!\!\!\!\!dy + \dots = Xdx + Ydy + Zdz + \dots 
\end{align}

<br>

Преобразованием Лежандра по паре $x, X$ называется следующее преобразование функции $L$:

<br>

\begin{align}
    &\overline{L} = L - x \cdot X \nonumber \\
    &\implies d\overline{L} = Xdx + Ydy +Z dz + \dots - Xdx - xdX \; = \nonumber \\
    &= -xdX + Ydy + Z dz + \dots
\end{align}

<br>

Итак, системы в положении равновесия характеризуются термодинамическими функциями. Все термодинамические величины могут быть получены как частные производные от термодинамических функций. Каждая термодинамическая функция зависит от набора независимых переменных ― естественных параметров.

<br>

### Внутренняя энергия

<br>

\begin{align}
    &\delta Q = TdS =dU + \delta A = dU + PdV, \nonumber 
    &\Rightarrow dU = TdS - PdV.
\end{align}

<br>

Следовательно, внутренняя энергия есть функция $S$ и $V$.

<br>

\begin{align}
    &\left(pdv{U}{S}\right)_V\!\!\! = T, 
    &\left(pdv{U}{V}\right)_S\!\!\! = -P.
\end{align}

<br>

Из математики знаем, что 

<br>

\begin{align}
    \dfrac{\partial^{\,2} U}{\partial S \partial V} = \dfrac{\partial^{\,2} U}{\partial V \partial S}.
\end{align}

<br>

Посчитаем и приравняем соответствующие вторые производные:

<br>

\begin{align}
    &\dfrac{\partial}{\partial S} \dfrac{\partial U}{\partial V} = - \left(pdv{P}{S}\right)_V\!\!\! = \dfrac{\partial}{\partial V} \dfrac{\partial U}{\partial S} = \left(pdv{T}{V}\right)_S, \nonumber \\
    &\implies \boxed{\left(pdv{T}{V}\right)_S\!\!\! = - \left(pdv{P}{S}\right)_{V \;\;}\!\!\!}.
\end{align}

<br>


### Энтальпия
<br>

Энтальпия \$-\$ преобразования Лежандра внутренней энергии относительно $(-P), \, V$:
\begin{align}
    &H \equiv I = U - (-P)V  = U + PV, \nonumber \\
    dH = dU + PdV + VdP = TdS + VdP.
\end{align}

<br>

Следовательно, энтальпия есть функция от её естественных параметров $S$ и $P$.

<br>

\begin{align}
    &\left(pdv{H}{S}\right)_P\!\!\!= T, \\
    &\left(pdv{H}{P}\right)_S\!\!\! = V, \\
    &\dfrac{\partial}{\partial S}\dfrac{\partial H}{\partial P} =  \left(pdv{V}{S}\right)_P\!\!\! = \dfrac{\partial}{\partial P}\dfrac{\partial H}{\partial S} = \left(pdv{T}{P}\right)_S\!\!\!, \nonumber \\
    &\implies \boxed{\left(pdv{V}{S}\right)_P\!\!\! = \left(pdv{T}{P}\right)_S}.
\end{align}

<br>

### Свободная энергия Гельмгольца
Свободная энергия Гельмгольца \$-\$ преобразования Лежандра внутренней энергии относительно $T,\, S$:

<br>

 \begin{align}
     &F = U - TS, \nonumber \\
     dF = - SdT - PdV. \label{dF}
 \end{align}
 
 <br>

 Следовательно, свободная энергия Гельмгольца \$-\$ функция от $T$ и $V$.
 
 <br>

 \begin{align}
     &\left(pdv{F}{T}\right)_V\!\!\! = -S, \\
     &\left(pdv{F}{V}\right)_T\!\!\! = -P, \\
     &\dfrac{\partial}{\partial T}\dfrac{\partial F}{\partial V} = -\left(pdv{P}{T}\right)_V\!\!\! = \dfrac{\partial}{\partial V}\dfrac{\partial F}{\partial T} = - \left(pdv{S}{V}\right)_T\!\!\!,  \nonumber \\
     &\implies \boxed{\left(pdv{P}{T}\right)_V\!\!\! = \left(pdv{S}{V}\right)_T}.
 \end{align}
 
 <br>

 Также можно сказать, что при изотермическом процессе ($d T=0$) выполняется следующее:
 
 <br>

 \begin{align}
     \boxed{dF = - \delta A}.
 \end{align}
 
 <br>

 ### Термодинамический потенциал Гиббса
 
 <br>

 Термодинамический потенциал Гиббса \$-\$ преобразования Лежандра энтальпии относительно $T, \, S$:
 
 <br>

 \begin{align}
     &G = I - TS = U + PV - TS = F + PV , \nonumber 
     dG = - SdT + VdP.
 \end{align}
 
 <br>

 Следовательно, термодинамический потенциал Гиббса \$-\$ функция от $T$ и $P$.
 
 <br>

 \begin{align}
     &\left(pdv{G}{T}\right)_P\!\!\! = -S, 
     &\left(pdv{G}{P}\right)_T\!\!\! = V,
 \end{align}
 
 <br>

  \begin{align}
     &\dfrac{\partial}{\partial T} \dfrac{\partial G}{\partial P} = \left(pdv{V}{T}\right)_P\!\!\! = \dfrac{\partial}{\partial P}\dfrac{\partial G}{\partial T} = -\left(pdv{S}{P}\right)_T\!\!\!, \nonumber \\
     &\implies \boxed{\left(pdv{V}{T}\right)_P\!\!\! = - \left(pdv{S}{P}\right)_T}.
 \end{align}

### Случай многих степеней свободы

До этого мы работали с системами двух степенй свободы, однако в случае большего их количества термодинамичекие потенциалы можно обобщить:

<br>

\begin{align}
    &\delta Q = TdS = dU + \sum_i A_i da_i, \\
    &dU = TdS - \sum_i A_i da_i, \\
    &dF = - SdT - \sum_i A_i da_i, \\
    &dH = TdS + \sum_i a_i dA_i, \\
    &dG = - SdT + \sum_i a_i dA_i.
\end{align}

## Про метод термодинамических функций

<br>

Для функции состояния $f(x, y)$ можно записать полный дифференциал
\begin{align}
    &df = X(x, y) dx + Y(x, y) dy, \\
    &\left( \dfrac{\partial X}{\partial y} \right)_x = \left( \dfrac{\partial Y}{\partial x} \right)_y,
\end{align}

<br>

что позволяет выводить различные термодинамические тождества.

<br>

Используем это, чтобы доказать, что $Q$ ― не функция состояния. Допустим, $dQ = \delta Q$ не зависит от траектории:
\begin{align}
    &dQ = dU + PdV = dH - VdP\; = \nonumber \\
    &= \left( \dfrac{\partial H}{\partial T} \right)_PdT + \left( \dfrac{\partial H}{\partial P} \right)_TdP - VdP \; = \nonumber \\
    &= \left( \dfrac{\partial H}{\partial T} \right)_PdT + \left[ \left( \dfrac{\partial H}{\partial P} \right)_T - V \right] dP. \\
    & \dfrac{\partial^2 Q}{\partial T \partial P} = \dfrac{\partial^2 H}{\partial P \partial T} = \dfrac{\partial^2 H}{\partial T \partial P} - \left( \dfrac{\partial V}{\partial T} \right)_P, 
    &\left( \dfrac{\partial V}{\partial T} \right)_P = 0.
\end{align}

<br>

Подобное уравнение весьма очевидно противоречит эмпирическим наблюдениям, из чего следует, что $Q$ нельзя считать функцией состояния.

## Уравнения Гиббса-Гельмгольца

<br>


\begin{align}
    &U = F + TS = F - T \left( \dfrac{\partial F}{\partial T} \right)_V, \\
    &H = G + TS = G - T \left( \dfrac{\partial G}{\partial T} \right)_P.
\end{align}

Выразим работу термодинамической системы из первого начала термодинамики:
\begin{align}
    A = Q - \Delta U.
\end{align}

<br>


Она очевидным образом зависит от траектории протекания процесса, однако интересно посмотреть, есть ли у совершённой системой работы верхний предел, считая температуру окружающей среды постоянной.
\begin{align}
    &\oint \dfrac{\delta Q}{T} \leqslant 0,
\end{align}

<br>


Неравенство Клаузиуса (интегрирование это суммирование),

<br>

\begin{align}
    &S_2 - S_1 \geqslant \int \limits_{1 \to 2} \dfrac{\delta Q}{T_0}, \\
    &S_2 - S_1 \geqslant \dfrac{Q}{T_0} = \dfrac{A + \Delta U}{T_0}, \\
    &T_0 S_2 - T_0 S_1 \geqslant A + \Delta U, \\
    &-U_2 + T_0 S_2 + U_1 - T_0 S_1 \geqslant A, \\
    &A \leqslant \underbrace{(U_1 - T_0 S_1)}_{F_1} - \underbrace{(U_2 - T_0 S_2)}_{F_2}. \\
    &\boxed{A \leqslant F_1 - F_2}.
\end{align}

<br>


При этом стоит заметить, что равенство достигается только в случае обратимости рассматриваемого процесса!


<br>


Рассмотрим такой процесс:

<br>

\begin{align}
    &A_{\text{max}} = F_1 - F_2, \\
    &F = U - T_0 S = U + T_0 \left( \dfrac{\partial F}{\partial T} \right)_V, \\
    &A_{\text{max}} = U_1 + T_0 \left( \dfrac{\partial F_1}{\partial T} \right)_V - U_2 - T_0 \left( \dfrac{\partial F_2}{\partial T} \right)_V \\
    &= U_1 - U_2 + T_0 \left( \dfrac{\partial (F_1 - F_2)}{\partial T} \right)_V,
\end{align}
\begin{align}
    \boxed{A_{\text{max}} = U_1 - U_2 + T_0 \left( \dfrac{\partial A_{\text{max}}}{\partial T} \right)_V}.
\end{align}

<br>

Иногда вводится вспомогательное понятие полезной работы: для поддержания постоянной температуры необходим контакт со средой с постоянным давлением $P_0$. Против неё совершается работа, однако никакого полезного эффекта эта работа $P_0 (V_2 - V_1)$ не несёт.

<br>

\begin{align}
    &A_{\text{п. max}} = F_1 - F_2 - P_0 V_2 + P_0 V_1 = \underbrace{(F_1 + P_0 V_1)}_{G_1} - \underbrace{(F_2 + P_0 V_2)}_{G_2},
\end{align}

<br>

\begin{align}
    \boxed{A_{\text{п. max}} = G_1 - G_2}.
\end{align}

<br>


Аналогично первому случаю приходим к равенству
\begin{align}
    &\boxed{A_{\text{п. max}} = H_1 - H_2 + T_0 \left( \dfrac{\partial A_{\text{п. max}}}{\partial T} \right)_P }.
\end{align}
