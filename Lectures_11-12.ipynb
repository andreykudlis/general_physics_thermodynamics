# Поверхностное натяжение

Если рассмотреть молекулу жидкости в объеме, то силы, действующие со стороны других молекул, в среднем уравновешивают друг друга. Однако, если молекула в приграничном слое, то на нее действует сила (внутрь). Следовательно, для перемещения молекулы на поверхность жидкости необходимо затратить работу (совершить). 

**Определение:** Работа, которую необходимо затратить, чтобы изотермически и квазистатически увеличить поверхность жидкости на единицу при сохранении объема неизменным, называется коэффициентом поверхностного натяжения или просто поверхностным натяжением.

$$d F \;=\; -Sd T\;-\; \delta A, $$
$$\delta Q \;=\; T d S \;=\; \delta A \;+\; d U \;=\; \delta A \;+\; d F \;+\; Td S \;+\; Sd T,$$
$$U\;=\;F\;+\;TS, \; d U\;=\; d F\;+\;Td S\;+\;Sd T.$$
$$0\;=\;\delta A \;+\; d F \;+\; Sd T.$$
Изотерма: $d F \;=\; - \delta A.$

Введем представление: $F\;=\; F_{\text{об.}}\;+\; F_{\text{пов.}}$
$$F_{\text{об.}} \; \sim \; V, \; F_{\text{пов.}} \; \sim S, \; \left( \dfrac{V}{S} \right)^{-1}\; \rightarrow_{S\;\rightarrow \; \infty}\;0.$$
$$F_{\text{пов.}}\;=\; \sigma \; \cdot \; S.$$

Рассмотрим палку: опыт показывает, что пленка старается уменьшить свою площадь. Приложим силу ($f$), чтобы уравновесить систему.

Сделаем бесконечно малое увеличение силы, следовательно, пленка начнет расти, а сила ($f$) совершит работу.

$$A\;=\; f \; \cdot \; \Delta x.$$

А $S$ увеличилась на $\Delta x$.

По определению: $A\;=\; \sigma  l \Delta x \;=\; f \Delta x$
$$\Rightarrow \; \boxed{\sigma\;=\; \dfrac{f}{l}}.$$

Получается, что пленка находится в натяжении, следовательно, при разрезе вообще любой пленки на линию разреза с обеих сторон будет действовать сила, касательная к поверхности и перпендикулярной к линии разреза.

Аналогия с резиновым мешком:
1. при растяжении р.м. силы вдоль растяжения, в жидости же  $\;-\;$ внутрь;
2. чем больше р.м. растянут, тем большая сила нужна.


При определении поверхностного натяжения необходимо добавлять инф. о том, с какой пов-тью гр. жидкость. Приграничный слой тоже будет вытягивать молекулы жидкости из жидкости.


В состоянии равновесия свободная энергия системы должна быть минимальной. Если на систему не действуют внешние силы (имеем дело со свободной жидкостью) $\Rightarrow$ будет принимать форму шара (При фиксированном объеме и )

**Замечание.** В условиях невесомости любая капля жидкости будет иметь сферическую форму.

## Термодинамика поверхностного натяжения

$$\partial Q=d U+\partial A=d U-\sigma d B.$$

$B$  $\;-\;$ площадь, $\sigma d B$  $\;-\;$ работа внешних сил.

$$d U=TdS+\sigma d B.$$

$$F=U-TS\Rightarrow d F=-Sd T+\sigma d B.$$

$$\Rightarrow S=-\Big(\frac{\partial F}{\partial T}\Big)_B.$$

$$\boxed{F=U+T\Big(\frac{\partial F}{\partial T}\Big)_B\;\;},$$ $$F=\sigma B \Rightarrow \sigma B=U+T\Big(\frac{\partial \sigma B}{\partial T}\Big)_B$$

$$\Rightarrow U=B(\sigma - T \frac{d  \sigma}{d T})$$

При изотермическом расширении:

$$\int \partial Q=\int d U-\int\sigma d B,$$
$$Q=\triangle U-\sigma\triangle B,$$
$$\triangle U=\sigma \triangle B - T\triangle B \frac{d \sigma}{d T} \Rightarrow,$$
$$\Rightarrow Q=-T\frac{d \sigma}{d T}\triangle B \Rightarrow$$
$$\Rightarrow \boxed{q=-T\frac{d \sigma}{d T}}.$$

$$\sigma\text{(опыт)}\downarrow c \uparrow T$$
$q$  $\;-\;$ тепло, которое сообщается единице поверхности пленки при изотермическом расширении.

## Формула Лапласа
Если имеем кривую поверхность жидкости, то давления по разные стороны должны быть разными.
Пусть $P_1$  $\;-\;$ внутреннее давление, тогда:

<br>

\begin{align}
    &F_{\sigma x}\;=\;-2b\sigma\sin{\dfrac{\varphi}{2}}\;\approx\;-b\sigma\varphi,\\
    &F_p\;=\;\left(P_2\;-\;P_1\right)S,\\
    &P_2\;-\;P_1\;-\underbrace{b\sigma\varphi}_{\Delta P}\;=\;0,\\
    &\dfrac{\sigma S}{RS}\;=\;\dfrac{ba\sigma}{RS}\;=\;\boxed{P_2\;-\;P_1\;=\;\Delta P\;=\;\dfrac{b\sigma\varphi}{S}}.
\end{align}

<br>

Рассмотрим случай, когда жидкость ограничена поверхностью второй кривизны:

<br>

\begin{align}
    &F_1\;=\;\dfrac{\sigma}{R_1}S,\\
    &F_2\;=\;\dfrac \sigma{R_2}S\\
    &\implies\;F\;=\;F_1\;+\;F_2\;=\;S\sigma\left(\dfrac{1}{R_1}\;+\;\dfrac{1}{R_2}\right)\\
    &\implies\;\Delta P\;=\;P_2\;-\;P_1\;=\;\\

# Фазовые переходы
Под  **фазой** будем подразумевать макроскопическую физически однородную часть вещества, отделённую от остальных частей системы границами раздела.

Пример: вода, лёд и насыщенный пар образуют 3 фазы. Отдельно отметим, что парообразных фаз всегда 1 (газы смешиваются).

Условие равновесия фаз складывается из теплового и механического равновесия. Также необходимо постоянство масс.

Механическое  $\;-\;$ постоянство давлений, тепловое  $\;-\;$ равество температур и равновесие по отношению к взаимным превращениям.

Немного терминологии:

**Конденсированное состояние**  $\;-\;$ сосуществование твёрдого и жидкого состояния.

**Сублимация** (возгонка)  $\;-\;$ процесс перехода вещества из твёрдого в газообразное состояние, минуя жидкое. Обратный процесс  $\;-\;$ **десублимация** (из газа в твёрдое, минуя жидкое).

**Испарение**  $\;-\;$ процесс перехода вещества из жидкого состояния в газообразное. Наоборот  $\;-\;$ **конденсация**.

**Плавление**  $\;-\;$ процесс перехода вещества из твёрдого состояния в жидкое. Наоборот  $\;-\;$ **кристаллизация** (отвердевание).
## Химически однородные вещества
Итак, в нашем процессе давление и температура поддерживаются постоянными. Следовательно, энергия Гиббса при фазовом переходе будет принимать своё минимальное значение, поскольку $G = G(P,T)$. Можно понять так:

<br>

\begin{align}
    &d G = - Sd T + V d P, \nonumber \\
    &d T = d P = 0,\nonumber \\
    &d G = 0.
\end{align}

<br>

Рассмотрим систему двух фаз. $m_1$, $m_2$  $\;-\;$ массы вещества в первой и второй фазе, $g_1$, $g_2$  $\;-\;$ удельные термодинамические потенциалы соответствующих фаз. Отдельно отметим, что наши удельные потенциалы не зависят от массы вещества и являются только лишь функциями $P$ и $T$ и не зависят от массы.

<br>

\begin{align}
    &G = G_1 + G_2 = m_1g_1 + m_2g_2, \nonumber \\
    &m = m_1 + m_2
\end{align}

<br>

Теперь положим, что в ходе фазового перехода получили $g_1>g_2$. Тогда любой переход $m_1$ в $m_2$ будет сопровождаться уменьшением $G$. Так и будет происходить, пока вся фаза 1 не перейдёт в фазу 2.

Можем провести аналогичные рассуждения в обратном направлении. Тогда $g_2>g_1$ и $m_2$ переходит в $m_1$.

Исходя из проделанных нами рассуждений, едиственно возможный вариант фазового равновесия:

<br>

\begin{align}
    g_1(P,T) = g_2(P,T).
\end{align}

<br>

**Nota bene:** Также вместо удельного термодинамического потенциала можно говорить про химический потенциал:

<br>

\begin{align}
    \mu = \dfrac{G}{N}, \ \text{где} \ N \ \text{ $\;-\;$ количество частиц}.
\end{align}

<br>

Вернёмся. Если равны потенциалы, то равны и их производные:

<br>

\begin{align}
    d g = -s d T + v d P, \;  s = \dfrac{S}{M}, \; \; v = \dfrac{V}{M},
\end{align}
Где $s, \; v$  $\;-\;$ удельные энтропия и объём, а $M$  $\;-\;$ общая масса вещества.

<br>

Итак, равновесие фаз происходит при равенстве удельных потенциалов. Следовательно, фазовые превращения происходят всегда при постоянных (и равных) значениях $g_1$ и $g_2$.

Стоит понимать, что при любых изменениях состояния вещества его удельный термодинамический потенциал меняется непрерывно.

Рассмотрим изотерму газа Ван-дер-Ваальса при температуре ниже критической. Поскольку двигаемся по изотерме, $d T=0$ и $d g = v d P$. Также мы уже знаем правило Максвелла и то, как происходит переход фаз.

Перейдём от координат $P,V$ к координатам $V,P$, поскольку удельный термодинамический потенциал зависит от давления.

\begin{align}
    \int d g = \int v d P.
\end{align}

<br>

Распишем промежутки убывания и возрастания нашего удельного потенциала:\\
1) $EA$  $\;-\;$ возрастает

2) $AB$  $\;-\;$ убывает

3) $BD$  $\;-\;$ возрастает.

Таким образом, найдётся точка на $BD$, в которой функция $g$ примет то же значение, что и в точке G:

<br>

\begin{align}
    g_G=g_L.
\end{align}

<br>

В свою очередь точки $G$ и $L$  $\;-\;$ это уже точки разных фаз.

Мы знаем, что $g$  $\;-\;$ функция состояния, и $\oint d g = 0$. Распишем:

<br>

\begin{align}
    &\oint d g = 0 =\int \limits_{GACBL} vd P + \int \limits_L^G vd P.
\end{align}

<br>

Интеграл по траектории $LG$ равен нулю по смыслу интеграла, и тогда

<br>

\begin{align}
    &0 = \int \limits_{GACBL} vd P = \int \limits_{GAC} vd P + \int \limits_{CBL} vd P = \int \limits_{GAC} vd P - \int \limits_{LBC} vd P, \nonumber \\
    &\implies \int \limits_{GAC} vd P = \int \limits_{LBC} vd P
\end{align}

<br>

Следовательно, равны заштрихованные площади. Получили правило Максвелла для описания равновесия фаз. Более того:

<br>

\begin{align}
    &g_M = g_G - \int \limits_{P_M}^{P_G} vd P, \nonumber \\
    &g_N = g_L - \int \limits_{P_N}^{P_L} vd P.
\end{align}

<br>

Причём в первом случае ($g_M$) $v$  $\;-\;$ это удельный объём газа, а во втором  $\;-\;$ удельный объём жидкости. Также известно, что $v_\text{газа} \gg v_\text{жидкости}$. И тогда $g_M < g_N$. Следовательно, газ  $\;-\;$ более устойчивое состояние.

<br>

Рассмотрим простейшую фазовую диаграмму. При равновесии фаз наблюдается $g_1 (P, T) = g_2 (P, T)$. Запишем условие на движение по кривой $P(T)$ сосуществования фаз:

<br>

\begin{align}
    & d g_1 = d g_2. \\
    & d g = -s d T + v d P, \\
    & -s_1 d T + v_1 d P = - s_2 d T + v_2 d P, \\
    & \dfrac{d P}{d T} = \dfrac{s_2 - s_1}{v_2 - v_1}.
\end{align}

<br>

Осталось перейти от разницы энтропий к какой-то величине, которую легче измерить в ходе эксперимента. Очевидный выход ― измерить удельную теплоту, которая выделяется в ходе фазового перехода ― обозначим эту \emph{удельную теплоту фазового перехода} как $q = T(s_2 - s_1)$.

<br>

\begin{align}
    & \boxed{\dfrac{d P}{d T} = \dfrac{q}{T(v_2 - v_1)}} \text{ ― уравнение Клапейрона-Клаузиуса}.
\end{align}

<br>

### Давление насыщенного пара


Рассмотрим переход от идеального газа к жидкости при условии нескольких допущений:
* Удельная теплота фазового перехода не зависит от температуры.
* Удельным объёмом жидкости можно пренебречь по сравнению с удельным объёмом газа (плотность жидкости сильно больше плотности газа).
* Выполняется уравнение состояния идеального газа $PV = \nu R T$, которое можно переформулировать как $Pv = \dfrac{RT}{\mu}$.


Уравнение Клапейрона-Клаузиуса примет следующий вид ($v$ ― удельный объём газа):

<br>

\begin{align}
    & \dfrac{d P}{d T} = \dfrac{q}{Tv} = \dfrac{q \cdot P \mu}{T \cdot RT} = \dfrac{\mu q}{R} \dfrac{P}{T^2}, \\
    & \dfrac{d P}{P} = \dfrac{\mu q}{R} \dfrac{d T}{T^2}. \\
    & \ln \dfrac{P}{P_0} = - \dfrac{\mu q}{R} \left( \dfrac{1}{T} - \dfrac{1}{T_0} \right), \\
    & P = P_0 \; \exp \left( \dfrac{- \mu q}{R} \left( \dfrac{1}{T} - \dfrac{1}{T_0} \right) \right).
\end{align}

## Химический потенциал
Стоит заметить, что в предыдущих разделах вместо удельных энергий Гиббса $g_1$ и $g_2$ можно было прибегнуть к понятию химического потенциала $\mu$, который в случае постоянного количества частиц $N$ определяется как 

<br>

\begin{align}
    & \mu = \dfrac{G}{N}.
\end{align}

<br>

Если же количество частиц непостояннно, то привычное по первому и второму началам термодинамики выражение для дифференциала внутренней энергии принимает вид

<br>

\begin{align}
    & \boxed{d U = T d S - P d V + \mu d N}.
\end{align}

<br>

В этом случае физический смысл химического потенциала $\mu$ ― количество энергии, привносимое адиабатическим добавлением одной частицы.

## Теплоёмкость насыщенного пара
Предположим, что мы имеем дело с процессом, где при изменении температуры объём меняется таким образом, чтобы пар всё ещё оставался насыщенным. $\dfrac{\delta Q}{d T}$ в таком процессе называют теплоёмкостью насыщенного пара.

Обозначив как $h$ удельную энтальпию (и переиспользовав $Q$ для удельной теплоты), можем записать

<br>

\begin{align}
    & \delta Q = d h - v d P, \\
    & \text{причём } d h = c_p d T. \\
    & c = \dfrac{\delta Q}{d T} = c_p - v \dfrac{d P}{d T}, \\
    & \dfrac{d P}{d T} = \dfrac{q}{v T}, \\
    & c = c_p - \dfrac{q}{T}.
\end{align}

<br>

## Тройные точки


Тройная точка ― точка сосуществования трёх различных фаз одной и той же системы: $g_1 (P, T) = g_2 (P, T) = g_3 (P, T)$. Например, для воды тройная точка располагается в 4.58 мм. рт. ст. и $0{,}008 \;  °$C. 

Во время фазового перехода температура постоянна, поэтому

<br>

\begin{align}
    &q_{13} = T (s_1 - s_3) = T (s_1 - s_2 + s_2 - s_3) = T (s_1 - s_2) + T (s_2 - s_3) = q_{12} + q_{23}.
\end{align}

<br>

Примем $v_2 \ll v_1$ и $v_3 \ll v_1$, тогда уравнения Клапейрона-Клаузиуса запишутся как

<br>

\begin{align}
    & \dfrac{d P_{12}}{d T} = \dfrac{q_{12}}{T (v_1 - v_2)} \approx \dfrac{q_{12}}{T v_1}, \\
    & \dfrac{d P_{13}}{d T} = \dfrac{q_{13}}{T (v_1 - v_3)} \approx \dfrac{q_{13}}{T v_1} = \dfrac{q_{12} + q_{23}}{T v_1}, \\
    & \text{то есть }\dfrac{d P_{13}}{d T} > \dfrac{d P_{12}}{d T}.
\end{align}
